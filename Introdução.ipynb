## üß† O que √© Probabilidade?

A probabilidade √© uma medida que expressa a chance de um determinado evento acontecer. Ela varia de 0 a 1, onde:

- **0**: O evento nunca acontece.
- **1**: O evento sempre acontece.

## ‚öñÔ∏è Exemplo Te√≥rico

Um experimento aleat√≥rio √© uma a√ß√£o ou processo que pode ser repetido v√°rias vezes, sob as mesmas condi√ß√µes, e cujos resultados n√£o podem ser previstos com certeza. Em outras palavras, √© um procedimento no qual o resultado √© incerto e determinado pelo acaso.

üìù Caracter√≠sticas de um Experimento Aleat√≥rio:
* Repetibilidade: Pode ser realizado v√°rias vezes nas mesmas condi√ß√µes.
* Resultados Poss√≠veis: O conjunto de todos os resultados poss√≠veis √© chamado de espa√ßo amostral.
* Incerteza: N√£o √© poss√≠vel prever com exatid√£o qual ser√° o resultado de uma √∫nica execu√ß√£o.
* Aleatoriedade: O resultado de cada execu√ß√£o √© determinado pelo acaso.


Exemplos Cl√°ssicos de Experimentos Aleat√≥rios: Lan√ßamento de uma moeda:
* Espa√ßo amostral: {Cara, Coroa}
* N√£o √© poss√≠vel prever qual lado cair√° em um lan√ßamento espec√≠fico.


```python
# Espa√ßo amostral para o lan√ßamento de uma moeda:
espaco_amostral = ['cara', 'coroa']

# N√∫mero de resultados poss√≠veis:
num_resultados_possiveis = len(espaco_amostral)

# N√∫mero de resultados favor√°veis (sair cara):
num_resultados_favoraveis = 1  # 'cara' √© o resultado desejado

# Probabilidade de sair cara:
probabilidade_cara = num_resultados_favoraveis / num_resultados_possiveis

print(f"A probabilidade de sair cara √©: {probabilidade_cara}")
```

## üîÑ Abordagem Frequentista

Em teoria, os eventos simples s√£o equiprov√°veis. No entanto, na maioria das situa√ß√µes pr√°ticas, os elementos do espa√ßo amostral n√£o s√£o igualmente prov√°veis e, deste modo, a probabilidade dos eventos deve ser calculada utilizando-se da no√ß√£o de freq√º√™ncia relativa

A abordagem frequentista interpreta a probabilidade como a frequ√™ncia relativa de um evento ao longo de um grande n√∫mero de repeti√ß√µes. Quanto mais observa√ß√µes, mais precisa ser√° a estimativa.

**Exemplo de Sa√∫de:**
Vamos simular um ensaio cl√≠nico com 1000 pacientes, onde a chance de complica√ß√µes √© de 20%:


```python
import numpy as np

# Definindo o n√∫mero de pacientes na simula√ß√£o
n_pacientes = 1000

# Gerando complica√ß√µes com distribui√ß√£o binomial (20% de chance)
# A fun√ß√£o `np.random.binomial` gera um array bin√°rio (0 ou 1) de tamanho `n_pacientes`.
complicacoes = np.random.binomial(1, 0.20, n_pacientes)

# Calculando a probabilidade estimada de complica√ß√µes
#A m√©dia dos valores resulta na probabilidade estimada de complica√ß√µes.
probabilidade_real = np.mean(complicacoes)

print(f"Probabilidade estimada de complica√ß√µes: {probabilidade_real * 100:.2f}%")
```

```python
import numpy as np

# Aumentando o n√∫mero de pacientes para um bilh√£o
n_pacientes = 1000000000

# Simulando complica√ß√µes novamente com probabilidade de 20%
complicacoes = np.random.binomial(1, 0.20, n_pacientes)
probabilidade_real = np.mean(complicacoes)
print(f"Probabilidade estimada de complica√ß√µes: {probabilidade_real * 100:.2f}%")
```

### üé∞ A Fal√°cia do Apostador
Acreditar que um evento futuro √© influenciado por eventos passados, como pensar que ap√≥s cinco resultados iguais, o pr√≥ximo ser√° diferente.

Na sa√∫de, isso pode ocorrer ao supor que, ap√≥s v√°rios pacientes terem resposta positiva a um tratamento, o pr√≥ximo ter√° menos chance de responder da mesma forma. Isso n√£o faz sentido se os eventos s√£o independentes.

### üéÇ **Paradoxo do Anivers√°rio**
Qual a probabilidade de que, em uma sala com **23 pessoas**, pelo menos duas compartilhem a mesma data de anivers√°rio? Este paradoxo √© interessante porque a probabilidade √© maior do que intuitivamente pensamos.

```python
import math

#Definimos uma fun√ß√£o que calcula a probabilidade de nenhuma coincid√™ncia de anivers√°rio em um grupo de `n` pessoas

def prob_compartilhar_aniversario(n):
    prob = 1.0
    for i in range(n):
        prob *= (365 - i) / 365
    return 1 - prob

n = 23
prob = prob_compartilhar_aniversario(n)
print(f'Probabilidade de pelo menos duas pessoas compartilharem anivers√°rio em um grupo de {n}: {prob:.2%}')
```
