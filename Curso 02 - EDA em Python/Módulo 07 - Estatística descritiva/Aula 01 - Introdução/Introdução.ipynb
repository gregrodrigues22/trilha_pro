#🧠 O que é Estatística Descritiva?

A estatística descritiva é o conjunto de métodos usados para **organizar, resumir e apresentar dados**.

Ela responde perguntas como:

📌 Qual é o valor típico de uma variável?

📌 Os valores são muito diferentes entre si?

📌 Existem valores extremos?

📌 Como os dados estão distribuídos?

Essas perguntas são fundamentais em qualquer análise de dados.

#📊 Exemplos em saúde

Imagine que queremos analisar os dados de pacientes de uma unidade de saúde.

Algumas perguntas que a estatística descritiva ajuda a responder:

📌 Qual é a idade média dos pacientes?

📌 Qual é a distribuição da pressão arterial?

📌 Existe grande variação no IMC da população?

📌 Existem pacientes com valores muito extremos?

Essas perguntas são o primeiro passo para compreender um conjunto de dados.

## Exemplo: Mudança nas taxas de mortalidade


Existem muitas estatísticas descritivas que podemos calcular a partir de dados de saúde.

Para ilustrar isso, imagine que queremos analisar se **a mortalidade hospitalar para um determinado procedimento tem diminuído ao longo do tempo**.

Abaixo está um pequeno conjunto de dados com taxas de mortalidade observadas em diferentes anos.

```python
import pandas as pd

dados = pd.DataFrame({
    "Ano": [2000,2002,2004,2006,2008,2010,2012,2014],
    "Taxa_Mortalidade": [8.2,7.9,7.5,6.8,6.4,6.1,5.9,5.7]
})

dados
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(dados["Ano"], dados["Taxa_Mortalidade"], marker="o")

for i in range(len(dados)):
    plt.text(dados["Ano"][i], dados["Taxa_Mortalidade"][i]+0.05,
             dados["Taxa_Mortalidade"][i])

plt.title("Taxa de mortalidade ao longo do tempo")
plt.xlabel("Ano")
plt.ylabel("Taxa de mortalidade (%)")

plt.grid(True)

plt.show()
```

O gráfico permite visualizar a tendência das taxas de mortalidade ao longo do tempo.

Observa-se uma redução gradual da taxa de mortalidade entre os anos analisados.

Esse tipo de gráfico é um exemplo clássico de estatística descritiva, pois nos permite observar padrões e tendências nos dados.

No entanto, o gráfico não nos diz por que a mortalidade diminuiu. Para investigar causas, seriam necessárias análises adicionais.

Para entender melhor a mudança nas taxas de mortalidade, podemos dividir os dados em dois períodos:

- Período inicial
- Período mais recente

Uma maneira simples de fazer isso é separar os anos **antes e depois de um determinado ponto no tempo**.

```python
dados.groupby(dados.Ano <= 2006)["Taxa_Mortalidade"].mean()
```

Esse cálculo mostra a **média da taxa de mortalidade em dois períodos diferentes**.

Esse tipo de análise é um exemplo de **estatística descritiva**, pois estamos apenas **resumindo e descrevendo os dados observados**.

Não estamos tentando explicar por que a mortalidade diminuiu, apenas descrevendo **o que aconteceu nos dados**.

#📊 Diferença com Estatística Inferencial

Na análise descritiva, o foco está em descrever e resumir os dados por meio de medidas numéricas e representações gráficas. Já na análise inferencial, o objetivo é tirar conclusões gerais sobre a população a partir de amostras representativas, utilizando técnicas como estimativas e testes de hipóteses.


image.png

| Comparação | Análise descritiva | Análise inferencial |
|---|---|---|
| **SIGNIFICADO** | A área da estatística é definida como estatística descritiva. Seu objetivo é fornecer uma descrição da população estudada. | Um tipo de estatística conhecida como estatística inferencial concentra-se em inferir informações sobre a população a partir de análises e descobertas de amostras. |
| **FUNÇÃO** | Para resumir a amostra, ela descreve os dados previamente conhecidos. | Faz um esforço para extrair uma amostra da população que vá além das evidências disponíveis. |
| **USO** | Explicar e descrever os dados coletados, tornando-os acessíveis para interpretação. | Estimar, prever ou testar hipóteses sobre fenômenos populacionais. |
| **MEDIDA TOMADA** | Organizar e visualizar os dados através de métricas como média, mediana, desvio padrão, entre outros. | Comparar dados, realizar testes de hipóteses e criar modelos preditivos. |
| **FORMA DE RESULTADO FINAL** | Apresentações gráficas (histogramas, tabelas, gráficos de dispersão) e numéricas (média, desvio padrão, etc.). | Resultados probabilísticos, como intervalos de confiança e significância estatística. |
