#🔍 O que queremos descobrir com Análise Bivariada?

* Existe relação entre essas duas variáveis?
* Essa relação é positiva (quanto maior a altura, maior o peso) ou negativa (quanto maior uma, menor a outra)?
* A relação é forte ou fraca?
* É linear ou de outro tipo (curvilínea, por exemplo)?

#⚒️ Ferramentas comuns

### Gráfico de dispersão (scatterplot)

* Cada ponto representa um indivíduo com suas duas medidas.
* Visualizamos se há tendência (pontos formando uma nuvem inclinada ou dispersa).

###Correlação

* Pearson: mede relação linear (requer normalidade aproximada).
* Spearman: mede relação monotônica (serve também para dados não normais ou ordinais).


Varia entre -1 e 1:
* +1 = correlação perfeita positiva.
* -1 = correlação perfeita negativa.
* 0 = ausência de correlação linear.

Realizando em Planilhas: Correl()

###Regressão linear simples

Ajusta uma equação do tipo:

𝑌 = 𝑎 + 𝑏 𝑋

* X = variável independente (ex.: altura).
* Y = variável dependente (ex.: peso).
* O coeficiente b mostra quanto Y varia em média quando X aumenta 1 unidade.

#🌳 Qual análise usar?

| Tipo de variáveis            | Associação / Exploratória      | Objetivo inferencial                                                                   | Objetivo preditivo                                                        |
| ---------------------------- | ------------------------------ | -------------------------------------------------------------------------------------- | ------------------------------------------------------------------------- |
| **Numérica × Numérica**      | Pearson, Spearman, Kendall, MI | Regressão linear                                                                       | Regressão, ML                                                             |
| **Numérica × Binária**       | Ponto-bisserial, MI            | t-test / Mann-Whitney                                                                  | Regressão logística                                                       |
| **Categórica × Binária**     | Qui-quadrado, Cramér’s V, MI   | Qui-quadrado / Fisher, **OR/RR não ajustado**                                          | Regressão logística (**OR ajustado**)                                     |
| **Categórica × Categórica**  | Qui-quadrado, Cramér’s V, MI   | Qui-quadrado / Fisher                                                                  | Modelos de classificação                                                  |
| **Categórica × Numérica**    | Eta², ANOVA, MI                | ANOVA / regressão com dummies                                                          | Regressão / classificação                                                 |
| **Binária × Binária**        | Qui-quadrado, OR, RR, MI       | **OR/RR não ajustado**, diferença de risco                                             | Regressão logística (**OR ajustado**) / Poisson robusta (**RR ajustado**) |
| **Multivariado (Y binário)** | MI, análise exploratória       | Regressão logística multivariada (**OR ajustado**) / Poisson robusta (**RR ajustado**) | ML, LASSO, árvores                                                        |


#📌 Por que não pular direto para o modelo ajustado (multivariado)?

**Exploração inicial dos dados**
* Antes de ajustar, você precisa conhecer a distribuição. Exemplo: se uma variável categórica quase não tem variação (ex.: 98% em um grupo, 2% em outro), o modelo pode travar (separação perfeita). Os testes bivariados ajudam a detectar isso.

**Transparência e comunicação científica**
* Em artigos, os revisores esperam ver: Tabela descritiva (como os grupos são diferentes). Testes bivariados (p-valores brutos). Modelo multivariado final (efeitos ajustados). Se você pula direto, pode parecer que “jogou tudo na caixa preta”.

**Comparar efeitos brutos vs ajustados**
* Exemplo: Idade tem OR bruto = 2, mas ajustado = 1,2. Isso revela confundimento (parte do efeito era explicado por outra variável). Sem ver o “bruto”, você perde essa informação.

**Controle da complexidade**
* Modelos multivariados com muitas variáveis aumentam risco de overfitting e multicolinearidade. Olhar associações simples antes ajuda a priorizar quais variáveis entram no modelo.

**Seleção de variáveis**
* Em datasets grandes, o screening bivariado ajuda a não levar lixo para o modelo. Variáveis sem associação alguma dificilmente vão ser relevantes no multivariado.
