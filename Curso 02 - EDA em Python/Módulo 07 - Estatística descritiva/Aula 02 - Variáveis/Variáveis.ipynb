#🧠 O que é uma variável

Variáveis são **características ou propriedades que podem ser observadas ou medidas** em um estudo.

Exemplos em saúde:

👤 idade do paciente  
⚧️ sexo  
🧪 nível de glicose  
🩺 número de consultas

Essas características podem variar entre indivíduos.

#🧩 Tipos de variáveis

As variáveis podem ser classificadas em duas categorias principais.

**Variáveis qualitativas**

Representam categorias.

Exemplos:

⚧️ sexo  
🩺 tipo de exame  

Podem ser:

🔹 Nominais → não possuem ordem  
🔹 Ordinais → possuem ordem



**Variáveis quantitativas**

Representam valores numéricos.

Podem ser:

🔢 Discretas → contagem  
📏 Contínuas → medidas

# 🧩 Por que categorizar as variáveis é importante?

Classificar corretamente as variáveis é uma etapa essencial em qualquer análise de dados.  
O tipo de variável influencia diretamente **como os dados serão tratados, analisados e interpretados**.

---

### 🧹 1️⃣ Preparação de dados

Saber o tipo de variável ajuda na **limpeza e preparação dos dados**, que são etapas críticas de qualquer análise.

Exemplos:

- tratar valores ausentes  
- padronizar categorias  
- converter tipos de dados  

---

### 🔎 2️⃣ Escolha da análise apropriada

Cada tipo de variável exige **técnicas estatísticas diferentes**.

Por exemplo:

- variáveis categóricas → contagens e proporções  
- variáveis numéricas → média, mediana, desvio padrão  

---

### 📊 3️⃣ Visualização de dados

O tipo de variável também determina **qual gráfico utilizar**.

Exemplos:

- variável categórica → gráfico de barras  
- variável contínua → histograma ou boxplot  

---

### 🤖 4️⃣ Modelagem estatística e aprendizado de máquina

A escolha de modelos estatísticos ou algoritmos de machine learning depende do tipo de variável.

Exemplos:

- regressão linear → variável resposta contínua  
- regressão logística → variável resposta categórica  
- árvores de decisão → podem lidar com diferentes tipos de variáveis
