# 🎯 Definição das Medidas de Frequência

Imagine que você coletou dados de satisfação de pacientes.

A pergunta não é só:
"Qual foi a resposta?"

Mas sim:
- Quantos responderam cada categoria?
- Qual proporção está satisfeita?
- Existe concentração em algum grupo?

É isso que as medidas de frequência respondem.

#❤️ Criando um dataset de exemplo em saúde

```python
import pandas as pd

df = pd.DataFrame({
    "paciente": range(1, 21),
    "satisfacao": [
        "Muito Insatisfeito", "Satisfeito", "Neutro", "Insatisfeito", "Neutro",
        "Muito Insatisfeito", "Satisfeito", "Neutro", "Muito satisfeito", "Muito satisfeito",
        "Muito satisfeito", "Insatisfeito", None, "Satisfeito", None,
        "Satisfeito", "Satisfeito", "Insatisfeito", "Neutro", "Neutro"
    ]
})

df
```

👉 Aqui já introduz um ponto importante: existem dados faltantes

# 📊 Cálculo das medidas de frequência

##### **Frequência absoluta(ni)**

A frequência absoluta indica quantas vezes cada categoria aparece.

```python
freq_abs = df["satisfacao"].value_counts(dropna=False)
freq_abs
```

👉 Interpretação: quantos pacientes estão em cada categoria

##### **Frequência relativa (fi)**

Mostra a proporção de cada categoria em relação ao total.

```python
freq_rel = df["satisfacao"].value_counts(normalize=True, dropna=False)
(freq_rel * 100).round(1)
```

##### **Percentual válido**

Considera apenas os dados válidos (exclui faltantes).

```python
df_validos = df["satisfacao"].dropna()

freq_rel_valida = df_validos.value_counts(normalize=True)
(freq_rel_valida * 100).round(1)
```

👉 Interpretação: mais fiel quando há missing

##### **Frequência acumulada**

Mostra o acúmulo das frequências ao longo das categorias.

```python
freq_ord = df_validos.value_counts().sort_index()

freq_acum = freq_ord.cumsum()
freq_rel_acum = (freq_ord / freq_ord.sum()).cumsum()

print("Frequência acumulada:")
print(freq_acum)

print("\nFrequência relativa acumulada (%):")
print((freq_rel_acum * 100).round(1))
```

##### **Tabela completa**

```python
tabela_freq = pd.DataFrame({
    "ni": df["satisfacao"].value_counts(dropna=False),
})

tabela_freq["fi (%)"] = (tabela_freq["ni"] / len(df) * 100).round(1)

tabela_freq_validos = df["satisfacao"].dropna().value_counts()

tabela_freq["Fi acumulada (%)"] = tabela_freq["fi (%)"].cumsum()

tabela_freq["fi válida (%)"] = (
    tabela_freq_validos / tabela_freq_validos.sum() * 100
).round(1)

tabela_freq
```

# 🧠 Interpretação

Como interpretar?

* Frequência absoluta → volume real
* Frequência relativa → proporção
* Percentual válido → análise sem viés de missing
* Frequência acumulada → concentração ao longo das categorias

Perguntas que você consegue responder:

* A maioria está satisfeita?
* Existe polarização?
* Há muitos dados faltantes?

# 🔗 Conexão com saúde

Medidas de frequência são usadas para:

* Distribuição de diagnósticos (CID)
* Origem dos pacientes
* Classificação de risco
* Satisfação do paciente (CSAT)
* Perfil epidemiológico

Elas são o primeiro passo antes de qualquer análise mais complexa.

# 🔥 Takeaway

Se tiver que levar uma coisa dessa aula

- Frequência é a forma mais simples e poderosa de entender dados
- Sempre verifique dados faltantes
- Percentual válido muda a interpretação
- Frequência acumulada ajuda a entender concentração

Antes de qualquer análise:
