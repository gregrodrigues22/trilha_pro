# 📊 Tipos de análises das variáveis


Análise univariada, bivariada e multivariada são termos utilizados na estatística para descrever o número de variáveis consideradas em uma análise. Esses tipos de análises têm uma relação íntima com a estatística descritiva, uma vez que cada tipo de análise pode utilizar métodos estatísticos descritivos para resumir e descrever os dados para uma ou mais variáveis em questão.


image.png



| Tipo | Definição | Objetivo |
|---|---|---|
| **Análise univariada** | A análise univariada é o tipo mais simples de análise estatística. Como o nome sugere, foca em uma única variável. | Descrever a distribuição e a natureza dessa única variável. |
| **Análise bivariada** | A análise bivariada examina a relação entre duas variáveis simultaneamente. | Identificar associações, correlações ou potenciais relações causais entre as duas variáveis. |
| **Análise multivariada** | A análise multivariada lida com três ou mais variáveis ao mesmo tempo. | Compreender as relações complexas entre várias variáveis, podendo incluir a identificação de padrões, grupos ou fatores subjacentes nos dados. |

#❤️ Criando um dataset de exemplo em saúde

Em todo caso, à título de exemplo rápido, vamos trabalhar com um pequeno conjunto de dados simulando pacientes de uma clínica, para mostrar os diferentes tipos de medidas.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "sexo": ["F", "M", "F", "F", "M", "M", "F", "M", "F", "F"],
    "idade": [23, 35, 29, 41, 50, 37, 31, 28, 46, 39],
    "consultas": [2, 5, 3, 6, 4, 7, 2, 1, 5, 3],
    "risco": ["baixo", "médio", "baixo", "alto", "médio", "alto", "baixo", "baixo", "alto", "médio"]
})

df
```

**Frequência para variável qualitativa**

```python
frequencia_sexo = df["sexo"].value_counts()
frequencia_relativa_sexo = df["sexo"].value_counts(normalize=True) * 100

print("Frequência absoluta:")
print(frequencia_sexo)

print("\nFrequência relativa (%):")
print(frequencia_relativa_sexo.round(1))
```

**Medidas de tendência central para variável quantitativa**

```python
idade = df["idade"]

print("Média:", idade.mean())
print("Mediana:", idade.median())
print("Moda:", idade.mode().tolist())
```

**Medidas de variabilidade para variável quantitativa**

```python
print("Variância:", idade.var())
print("Desvio padrão:", idade.std())
```

**Medidas de separação**

```python
print("Q1:", idade.quantile(0.25))
print("Q2 (mediana):", idade.quantile(0.50))
print("Q3:", idade.quantile(0.75))
print("Percentil 90:", idade.quantile(0.90))
```

**Medidas de dispersão para variável quantitativa**

```python
print("Mínimo:", idade.min())
print("Máximo:", idade.max())
print("Amplitude:", idade.max() - idade.min())
```

```python
print("Assimetria:", idade.skew())
print("Curtose:", idade.kurt())
```

```python
Q1 = idade.quantile(0.25)
Q3 = idade.quantile(0.75)
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"Intervalo Interquartil (IQR): {IQR}")
```
