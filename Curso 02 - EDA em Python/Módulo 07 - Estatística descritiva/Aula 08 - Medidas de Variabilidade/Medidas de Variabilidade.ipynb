#🎯 Medidas de variabilidade

As medidas de tendência central mostram onde os dados estão concentrados. Mas isso não basta. Dois conjuntos podem ter a mesma média e a mesma mediana e, ainda assim, serem muito diferentes em termos de dispersão. É para isso que servem as medidas de variabilidade: elas mostram o quanto os valores se afastam do centro.

#🧸  Exemplo conceitual inicial

```python
import pandas as pd
import numpy as np

grupo_1 = pd.Series([4, 5, 5, 5, 6])
grupo_2 = pd.Series([0, 1, 5, 9, 10])

print("Grupo 1")
print("Média:", grupo_1.mean())
print("Mediana:", grupo_1.median())
print("Variância:", grupo_1.var(ddof=0))
print("Desvio padrão:", grupo_1.std(ddof=0))

print("\nGrupo 2")
print("Média:", grupo_2.mean())
print("Mediana:", grupo_2.median())
print("Variância:", grupo_2.var(ddof=0))
print("Desvio padrão:", grupo_2.std(ddof=0))
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Box(x=grupo_1, name="Grupo 1", boxpoints="all", jitter=0.3, pointpos=0))
fig.add_trace(go.Box(x=grupo_2, name="Grupo 2", boxpoints="all", jitter=0.3, pointpos=0))

fig.update_layout(
    title="Mesma média, mesma mediana, variabilidade diferente",
    xaxis_title="Valor",
    template="plotly_white",
    height=450
)

fig.show()
```

Observe que os dois grupos têm o mesmo centro, mas o segundo é muito mais espalhado.

Isso mostra que média e mediana não são suficientes para descrever completamente um conjunto de dados.

# 📊 Conceito de variância

A variância mede o quanto os valores se afastam da média.

Ela é calculada a partir de três ideias:

1. calcular a média
2. ver a distância de cada valor em relação à média
3. elevar essas distâncias ao quadrado e tirar a média

```python
perda_peso = pd.Series([2.0, 8.0, 5.0, 10.0, 1.0])

media = perda_peso.mean()

tabela_variancia = pd.DataFrame({
    "perda_peso": perda_peso,
    "dif_media": perda_peso - media,
    "dif_media_ao_quadrado": (perda_peso - media) ** 2
})

tabela_variancia
```

```python
variancia_pop = tabela_variancia["dif_media_ao_quadrado"].mean()
desvio_pop = np.sqrt(variancia_pop)

print("Média:", round(media, 2))
print("Variância populacional:", round(variancia_pop, 2))
print("Desvio padrão populacional:", round(desvio_pop, 2))
```

👉 **Interpretação**

Quanto maior a variância, maior o espalhamento dos dados em torno da média.

A variância é útil, mas tem uma limitação:
ela fica em unidade ao quadrado.

Por isso, muitas vezes o desvio padrão é mais intuitivo.

# ⚖️ Conceito de variância

- Variância populacional: usada quando temos todos os elementos da população
- Variância amostral: usada quando temos apenas uma amostra

A diferença prática no Python está no parâmetro `ddof`

No Python:

- `ddof=0` calcula a versão populacional
- `ddof=1` calcula a versão amostral

Em análises com bases reais, costuma-se usar a versão amostral.

```python
print("Variância populacional:", perda_peso.var(ddof=0))
print("Variância amostral:", perda_peso.var(ddof=1))

print("Desvio padrão populacional:", perda_peso.std(ddof=0))
print("Desvio padrão amostral:", perda_peso.std(ddof=1))
```

#❤️ Aplicação em um dataset de exemplo em saúde

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# ---------------------------------
# 1. BASE BIMODAL CONTROLADA
# ---------------------------------

# dois grupos bem separados
grupo_1 = np.random.normal(loc=200, scale=25, size=400)
grupo_2 = np.random.normal(loc=800, scale=25, size=400)

# pequena cauda alta para assimetria
grupo_3 = np.random.normal(loc=2500, scale=400, size=40)

custos_base = np.concatenate([grupo_1, grupo_2, grupo_3])
custos_base = np.abs(custos_base)

# arredondar para múltiplos de 10
custos_base = (custos_base / 10).round() * 10

df = pd.DataFrame({"custo": custos_base})

# ---------------------------------
# 2. GARANTIR DUAS MODAS EXATAS
# ---------------------------------

# remover todos os 200 e 800 que surgiram naturalmente
df = df[~df["custo"].isin([200, 800])].copy().reset_index(drop=True)

# adicionar exatamente a mesma quantidade de 200 e 800
moda_200 = pd.DataFrame({"custo": [200] * 180})
moda_800 = pd.DataFrame({"custo": [800] * 180})

df = pd.concat([df, moda_200, moda_800], ignore_index=True)
```

# 📊 Conceito de Desvio padrão

O desvio padrão é a raiz quadrada da variância.

Ele indica, em média, o quanto os dados se afastam da média, na mesma unidade original da variável.

```python
custos = df["custo"]

media_custo = custos.mean()
desvio_custo = custos.std(ddof=1)

print("Média dos custos:", round(media_custo, 2))
print("Desvio padrão dos custos:", round(desvio_custo, 2))
```

👉 **Interpretação**

Se o desvio padrão é alto, os custos são mais heterogêneos.

Se o desvio padrão é baixo, os custos são mais homogêneos.

Ou seja:
- menor desvio padrão → maior consistência
- maior desvio padrão → maior dispersão

```python
import plotly.express as px

fig = px.histogram(
    df,
    x="custo",
    nbins=50,
    marginal="box",
    title="Distribuição dos custos e sua variabilidade"
)

fig.add_vline(
    x=media_custo,
    line_dash="dash",
    line_width=3
)

fig.add_annotation(
    x=media_custo,
    y=1.05,
    xref="x",
    yref="paper",
    text=f"Média = {media_custo:.0f}",
    showarrow=False
)

fig.update_layout(
    template="plotly_white",
    xaxis_title="Custo",
    yaxis_title="Frequência",
    height=600,
    title_x=0.5
)

fig.show()
```

👉 **Interpretação**

- caixa larga → maior dispersão
- presença de muitos pontos extremos → heterogeneidade
- distribuição mais espalhada → maior variabilidade

# 🆚 Bloco comparativo

Simular dois cenários com mesma média e desvio diferente.

```python
np.random.seed(42)

baixa_var = pd.Series(np.random.normal(loc=600, scale=40, size=300))
alta_var = pd.Series(np.random.normal(loc=600, scale=220, size=300))

comparacao = pd.DataFrame({
    "custo": pd.concat([baixa_var, alta_var], ignore_index=True),
    "grupo": ["Baixa variabilidade"] * len(baixa_var) + ["Alta variabilidade"] * len(alta_var)
})
```

```python
comparacao.groupby("grupo")["custo"].agg(["mean", "median", "var", "std"])
```

```python
fig = px.box(
    comparacao,
    x="grupo",
    y="custo",
    color="grupo",
    title="Mesmo centro, dispersões diferentes"
)

fig.update_layout(
    template="plotly_white",
    xaxis_title="",
    yaxis_title="Custo",
    height=500,
    showlegend=False,
    title_x=0.5
)

fig.show()
```

👉 **Interpretação**

Os dois grupos podem ter médias parecidas, mas comportamentos muito diferentes.

O grupo com maior desvio padrão é mais instável, menos homogêneo e mais imprevisível

# 📱 Aplicação em saúde

Por que isso importa em saúde?

Medidas de variabilidade ajudam a entender se os dados são homogêneos ou heterogêneos.

Exemplos:
- tempo de espera muito variável pode indicar instabilidade operacional
- custos muito variáveis podem indicar falta de padronização
- respostas muito variáveis em satisfação podem indicar experiência inconsistente

# 🔥 Takeaway


Se tiver que levar uma coisa desta aula, leve isso:

- média mostra o centro
- variabilidade mostra o espalhamento
- desvio padrão ajuda a entender consistência
- alta variabilidade sugere heterogeneidade
- baixa variabilidade sugere maior homogeneidade

Sem medidas de variabilidade, a leitura dos dados fica incompleta.
