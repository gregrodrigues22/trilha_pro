#Set up

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

# Função do gráfico

```python
def histograma_boxplot_plotly(
    df,
    col_valor="valor",
    titulo="Distribuição da variável",
    subtitulo=None,
    usar_log=True,
    incluir_zeros=False,
    nbins=None,
    metodo_bins="sturges",
    largura_px=1400,
    altura_px=850,
    altura_histograma=0.75,
    espacamento_vertical=0.06,
    x_label="Valor",
    y_label_hist="Número de registros",
    y_label_box="Boxplot",
    exibir_curva_normal=True,
    exibir_media=True,
    exibir_mediana=True,
    exibir_legenda_info=True,
    exibir_outliers_boxplot=True,
    posicao_texto_barras="outside",
    template_texto_barras="{contagem:,.0f} ({percentual:.1f}%)",
    casas_decimais=2,
    prefixo_valor="",
    cor_histograma="rgba(46, 107, 177, 0.85)",
    cor_boxplot_preenchimento="rgba(46, 107, 177, 0.55)",
    cor_boxplot_linha="rgba(46, 107, 177, 1.0)",
    cor_curva_normal="red",
    cor_media="black",
    cor_mediana="green",
    dash_media="dash",
    dash_mediana="dash",
    cor_fundo="white",
    cor_subtitulo="gray",
    tamanho_fonte_titulo=24,
    tamanho_fonte_subtitulo=13,
    tamanho_fonte_eixos=13,
    tamanho_fonte_ticks=10,
    tamanho_fonte_legenda=11,
    margem_esquerda=90,
    margem_direita=280,
    margem_superior=120,
    margem_inferior=80,
    posicao_legenda_x=1.02,
    posicao_legenda_y=0.98,
    ancora_legenda_x="left",
    ancora_legenda_y="top",
    fundo_legenda="rgba(255,255,255,0.90)",
    borda_legenda="rgba(0,0,0,0.15)",
    espessura_borda_legenda=1,
    posicao_subtitulo=0.948,
    exibir_figura=True,
):
    """
    Gera um gráfico combinado de distribuição contendo:
    - histograma
    - curva normal teórica opcional
    - linhas de média e mediana
    - boxplot horizontal

    A função é agnóstica e pode ser usada para qualquer variável numérica contínua.

    Parâmetros
    ----------
    df : pandas.DataFrame
        O DataFrame contendo os dados.
    col_valor : str, default 'valor'
        Nome da coluna no DataFrame que contém os valores numéricos a serem plotados.
    titulo : str, default 'Distribuição da variável'
        Título principal do gráfico.
    subtitulo : str, optional
        Subtítulo do gráfico. Se `None`, um subtítulo automático será gerado.
    usar_log : bool, default True
        Se `True`, o eixo X será em escala logarítmica. Caso contrário, será linear.
    incluir_zeros : bool, default False
        Se `True`, valores zero serão incluídos na análise. Caso contrário, serão excluídos.
    nbins : int, optional
        Número de bins para o histograma. Se `None`, será calculado automaticamente usando `metodo_bins`.
    metodo_bins : str, default 'sturges'
        Método para calcular o número de bins automaticamente ('sturges', 'sqrt', 'rice').
    largura_px : int, default 1400
        Largura total do gráfico em pixels.
    altura_px : int, default 850
        Altura total do gráfico em pixels.
    altura_histograma : float, default 0.75
        Proporção da altura total do gráfico dedicada ao histograma (entre 0 e 1).
    espacamento_vertical : float, default 0.06
        Espaçamento vertical entre o histograma e o boxplot (entre 0 e 1).
    x_label : str, default 'Valor'
        Rótulo do eixo X.
    y_label_hist : str, default 'Número de registros'
        Rótulo do eixo Y do histograma.
    y_label_box : str, default 'Boxplot'
        Rótulo do eixo Y do boxplot.
    exibir_curva_normal : bool, default True
        Se `True`, uma curva de distribuição normal teórica será exibida sobre o histograma.
    exibir_media : bool, default True
        Se `True`, uma linha indicando a média da distribuição será exibida.
    exibir_mediana : bool, default True
        Se `True`, uma linha indicando a mediana da distribuição será exibida.
    exibir_legenda_info : bool, default True
        Se `True`, informações estatísticas adicionais serão exibidas na legenda.
    exibir_outliers_boxplot : bool, default True
        Se `True`, os outliers serão exibidos no boxplot.
    posicao_texto_barras : str, default 'outside'
        Posição do texto nas barras do histograma ('inside', 'outside', 'auto', 'none').
    template_texto_barras : str, default '{contagem:,.0f} ({percentual:.1f}%)'
        Template para formatar o texto exibido nas barras do histograma.
    casas_decimais : int, default 2
        Número de casas decimais para exibir valores formatados.
    prefixo_valor : str, default ''
        Prefixo a ser adicionado aos valores formatados (ex: 'R$ ').
    cor_histograma : str, default 'rgba(46, 107, 177, 0.85)'
        Cor das barras do histograma.
    cor_boxplot_preenchimento : str, default 'rgba(46, 107, 177, 0.55)'
        Cor de preenchimento do boxplot.
    cor_boxplot_linha : str, default 'rgba(46, 107, 177, 1.0)'
        Cor da linha do boxplot.
    cor_curva_normal : str, default 'red'
        Cor da curva normal teórica.
    cor_media : str, default 'black'
        Cor da linha da média.
    cor_mediana : str, default 'green'
        Cor da linha da mediana.
    dash_media : str, default 'dash'
        Estilo da linha da média (ex: 'solid', 'dot', 'dash', 'longdash', 'dashdot', 'longdashdot').
    dash_mediana : str, default 'dash'
        Estilo da linha da mediana.
    cor_fundo : str, default 'white'
        Cor de fundo do gráfico.
    cor_subtitulo : str, default 'gray'
        Cor do texto do subtítulo.
    tamanho_fonte_titulo : int, default 24
        Tamanho da fonte do título principal.
    tamanho_fonte_subtitulo : int, default 13
        Tamanho da fonte do subtítulo.
    tamanho_fonte_eixos : int, default 13
        Tamanho da fonte dos rótulos dos eixos.
    tamanho_fonte_ticks : int, default 10
        Tamanho da fonte dos ticks dos eixos.
    tamanho_fonte_legenda : int, default 11
        Tamanho da fonte da legenda.
    margem_esquerda : int, default 90
        Margem esquerda do gráfico em pixels.
    margem_direita : int, default 280
        Margem direita do gráfico em pixels.
    margem_superior : int, default 120
        Margem superior do gráfico em pixels.
    margem_inferior : int, default 80
        Margem inferior do gráfico em pixels.
    posicao_legenda_x : float, default 1.02
        Posição X da legenda (relativo ao papel, entre 0 e 1).
    posicao_legenda_y : float, default 0.98
        Posição Y da legenda (relativo ao papel, entre 0 e 1).
    ancora_legenda_x : str, default 'left'
        Ancoragem horizontal da legenda.
    ancora_legenda_y : str, default 'top'
        Ancoragem vertical da legenda.
    fundo_legenda : str, default 'rgba(255,255,255,0.90)'
        Cor de fundo da caixa da legenda.
    borda_legenda : str, default 'rgba(0,0,0,0.15)'
        Cor da borda da caixa da legenda.
    espessura_borda_legenda : int, default 1
        Espessura da borda da caixa da legenda.
    posicao_subtitulo : float, default 0.948
        Posição Y do subtítulo (relativo ao papel, entre 0 e 1).
    exibir_figura : bool, default True
        Se `True`, a figura será exibida após a geração.

    Retorna
    -------
    tuple
        Uma tupla contendo a figura do Plotly (go.Figure) e um dicionário com um resumo estatístico.
    """

    # Importações necessárias para a função
    import numpy as np
    import pandas as pd
    from scipy import stats
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots

    # ==========================================================
    # 1. Funções auxiliares
    # ==========================================================
    def formatar_valor(v):
        """
        Formata um valor numérico para exibição, adicionando prefixo,
        casas decimais e separador de milhares/decimais conforme as configurações.
        """
        texto = f"{prefixo_valor}{v:,.{casas_decimais}f}"
        # Ajusta para formato brasileiro (vírgula como separador decimal)
        return texto.replace(",", "X").replace(".", ",").replace("X", ".")

    def formatar_inteiro(v):
        """
        Formata um valor inteiro para exibição com separador de milhares.
        """
        # Ajusta para formato brasileiro (ponto como separador de milhares)
        return f"{int(v):,}".replace(",", ".")

    def calcular_nbins_automatico(n, metodo="sturges"):
        """
        Calcula o número ideal de bins para um histograma com base em diferentes métodos.

        Parâmetros
        ----------
        n : int
            Número de observações nos dados.
        metodo : str, default 'sturges'
            Método a ser utilizado ('sturges', 'sqrt', 'rice').

        Retorna
        -------
        int
            Número de bins calculado.
        """
        metodo = str(metodo).lower()

        if n <= 1:
            return 1

        if metodo == "sturges":
            # Regra de Sturges: k = 1 + log2(n)
            bins = int(np.ceil(1 + np.log2(n)))
        elif metodo == "sqrt":
            # Regra da raiz quadrada: k = sqrt(n)
            bins = int(np.ceil(np.sqrt(n)))
        elif metodo == "rice":
            # Regra de Rice: k = 2 * n ** (1 / 3)
            bins = int(np.ceil(2 * (n ** (1 / 3))))
        else:
            raise ValueError("`metodo_bins` deve ser 'sturges', 'sqrt' ou 'rice'.")

        # Garante um número mínimo de bins para melhor visualização
        return max(5, bins)

    def montar_textos_barras(contagens, percentuais):
        """
        Monta os textos a serem exibidos nas barras do histograma,
        formatando contagem e percentual conforme o template.
        """
        textos = []
        for c, p in zip(contagens, percentuais):
            texto = template_texto_barras.format(contagem=c, percentual=p)
            # Ajusta para formato brasileiro
            texto = texto.replace(",", "X").replace(".", ",").replace("X", ".")
            textos.append(texto)
        return textos

    # ==========================================================
    # 2. Validações iniciais
    # ==========================================================
    # Verifica se a coluna de valor especificada existe no DataFrame.
    if col_valor not in df.columns:
        raise ValueError(f"A coluna '{col_valor}' não existe no DataFrame.")

    # Valida a proporção da altura do histograma.
    if not 0 < altura_histograma < 1:
        raise ValueError("`altura_histograma` deve estar entre 0 e 1.")

    # Valida o espaçamento vertical entre os subplots.
    if not 0 <= espacamento_vertical < 1:
        raise ValueError("`espacamento_vertical` deve estar entre 0 e 1.")

    # ==========================================================
    # 3. Preparação dos dados
    # ==========================================================
    # Converte a coluna de valor para numérico, trata valores não numéricos
    # como NaN e remove-os, garantindo que a série seja de floats.
    serie = pd.to_numeric(df[col_valor], errors="coerce").dropna().astype(float)

    # Remove valores zero se `incluir_zeros` for False.
    if not incluir_zeros:
        serie = serie[serie != 0]

    # Remove valores não positivos se `usar_log` for True (pois log(0) e log(negativo) são indefinidos).
    if usar_log:
        serie = serie[serie > 0]

    # Verifica se há dados válidos após as filtragens.
    if serie.empty:
        raise ValueError("Após os filtros aplicados, não restaram valores válidos para análise.")

    # Calcula o número total de registros após a limpeza.
    total = int(len(serie))

    # ==========================================================
    # 4. Número de bins
    # ==========================================================
    # Define o número de bins para o histograma. Se não for especificado,
    # calcula automaticamente usando o método definido.
    if nbins is None:
        nbins = calcular_nbins_automatico(total, metodo_bins)
    else:
        nbins = int(max(1, nbins))  # Garante que nbins seja pelo menos 1

    # ==========================================================
    # 5. Estatísticas descritivas
    # ==========================================================
    # Calcula as principais estatísticas descritivas da série de dados.
    media = float(serie.mean())
    mediana = float(serie.median())
    minimo = float(serie.min())
    maximo = float(serie.max())
    desvio_padrao = float(serie.std(ddof=0))  # ddof=0 para desvio padrão populacional

    # Calcula quartis e limites para identificação de outliers (regra 1.5 * IQR).
    q1, q3 = np.percentile(serie, [25, 75])
    iqr = q3 - q1
    fence_low = q1 - 1.5 * iqr
    fence_high = q3 + 1.5 * iqr

    # Conta o número de outliers abaixo e acima dos limites.
    out_low = int((serie < fence_low).sum())
    out_high = int((serie > fence_high).sum())

    # ==========================================================
    # 6. Histograma
    # ==========================================================
    # Calcula as contagens e as bordas dos bins para o histograma.
    counts, edges = np.histogram(serie, bins=nbins)
    # Calcula as larguras e centros de cada barra do histograma.
    widths = np.diff(edges)
    centers = (edges[:-1] + edges[1:]) / 2
    # Calcula os percentuais de cada bin em relação ao total de dados.
    percentuais = counts / total * 100

    # Gera os textos a serem exibidos nas barras do histograma.
    textos_barras = montar_textos_barras(counts, percentuais)

    # ==========================================================
    # 7. Curva normal teórica
    # ==========================================================
    # Gera pontos para plotar uma curva normal teórica com a mesma média e desvio padrão dos dados.
    x_vals = np.linspace(serie.min(), serie.max(), 400)
    # Evita divisão por zero se o desvio padrão for zero.
    desvio_efetivo = desvio_padrao if desvio_padrao > 0 else 1e-9
    # Calcula os valores Y da PDF normal, escalados para a altura do histograma.
    y_vals = stats.norm.pdf(x_vals, media, desvio_efetivo) * total * widths.mean()

    # Determina o valor máximo de Y para dimensionar o eixo do histograma corretamente.
    ymax_ref = max(np.max(counts), np.max(y_vals) if exibir_curva_normal else np.max(counts))

    # ==========================================================
    # 8. Criação da figura
    # ==========================================================
    # Cria uma figura com dois subplots (histograma e boxplot), compartilhando o eixo X.
    fig = make_subplots(
        rows=2,
        cols=1,
        shared_xaxes=True,
        row_heights=[altura_histograma, 1 - altura_histograma],  # Define a proporção de altura de cada subplot
        vertical_spacing=espacamento_vertical,
    )

    # ==========================================================
    # 9. Histograma
    # ==========================================================
    # Adiciona o histograma ao primeiro subplot.
    fig.add_trace(
        go.Bar(
            x=centers,
            y=counts,
            width=widths,
            # Exibe texto nas barras se `posicao_texto_barras` não for 'none'
            text=None if posicao_texto_barras == "none" else textos_barras,
            textposition=None if posicao_texto_barras == "none" else posicao_texto_barras,
            marker=dict(color=cor_histograma),
            name="Distribuição",
        ),
        row=1,
        col=1,
    )

    # ==========================================================
    # 10. Curva normal
    # ==========================================================
    # Adiciona a curva normal teórica se `exibir_curva_normal` for True.
    if exibir_curva_normal:
        fig.add_trace(
            go.Scatter(
                x=x_vals,
                y=y_vals,
                mode="lines",
                line=dict(color=cor_curva_normal),
                name="Curva normal",
            ),
            row=1,
            col=1,
        )

    # ==========================================================
    # 11. Média
    # ==========================================================
    # Adiciona uma linha vertical para a média se `exibir_media` for True.
    if exibir_media:
        fig.add_trace(
            go.Scatter(
                x=[media, media],
                y=[0, ymax_ref],  # Linha vai da base até o topo do histograma
                mode="lines",
                line=dict(color=cor_media, dash=dash_media),
                name=f"Média: {formatar_valor(media)}",
            ),
            row=1,
            col=1,
        )

    # ==========================================================
    # 12. Mediana
    # ==========================================================
    # Adiciona uma linha vertical para a mediana se `exibir_mediana` for True.
    if exibir_mediana:
        fig.add_trace(
            go.Scatter(
                x=[mediana, mediana],
                y=[0, ymax_ref],  # Linha vai da base até o topo do histograma
                mode="lines",
                line=dict(color=cor_mediana, dash=dash_mediana),
                name=f"Mediana: {formatar_valor(mediana)}",
            ),
            row=1,
            col=1,
        )

    # ==========================================================
    # 13. Boxplot
    # ==========================================================
    # Adiciona o boxplot horizontal ao segundo subplot.
    fig.add_trace(
        go.Box(
            x=serie,
            orientation="h",
            name="Boxplot",
            # Exibe outliers se `exibir_outliers_boxplot` for True
            boxpoints="outliers" if exibir_outliers_boxplot else False,
            marker=dict(color=cor_boxplot_preenchimento),
            line=dict(color=cor_boxplot_linha),
        ),
        row=2,
        col=1,
    )

    # ==========================================================
    # 14. Informações adicionais na legenda
    # ==========================================================
    # Adiciona informações estatísticas detalhadas à legenda como traces "fantasma".
    if exibir_legenda_info:
        legenda_info = [
            f"Menor valor: {formatar_valor(minimo)}",
            f"Maior valor: {formatar_valor(maximo)}",
            f"Q1 (25%): {formatar_valor(q1)}",
            f"Q3 (75%): {formatar_valor(q3)}",
            f"IQR (Q3-Q1): {formatar_valor(iqr)}",
            f"Limite inferior: {formatar_valor(fence_low)}",
            f"Limite superior: {formatar_valor(fence_high)}",
            f"Outliers abaixo: {formatar_inteiro(out_low)}",
            f"Outliers acima: {formatar_inteiro(out_high)}",
        ]

        for txt in legenda_info:
            fig.add_trace(
                go.Scatter(
                    x=[None],
                    y=[None],
                    mode="markers",
                    marker=dict(size=10, color="rgba(0,0,0,0)"),  # Marker transparente
                    showlegend=True,
                    name=txt,
                    hoverinfo="skip",  # Não exibe hoverinfo para esses traces
                ),
                row=1,
                col=1,
            )

    # ==========================================================
    # 15. Subtítulo automático
    # ==========================================================
    # Gera um subtítulo automático com base nas características dos dados e filtros aplicados.
    if subtitulo is None:
        subtitulo_final = (
            f"n={formatar_inteiro(total)}"
            f" | bins={nbins}"
            f" | {'com zeros' if incluir_zeros else 'sem zeros'}"
            f" | {'log' if usar_log else 'linear'}"
        )
    else:
        subtitulo_final = subtitulo

    # ==========================================================
    # 16. Layout geral
    # ==========================================================
    # Configura o layout geral da figura, incluindo título, cores de fundo, dimensões e margens.
    fig.update_layout(
        title=dict(
            text=f"<b>{titulo}</b>",  # Título em negrito
            x=0.43,  # Posição horizontal do título
            xanchor="center",
            y=0.90,
            yanchor="top",
            font=dict(size=tamanho_fonte_titulo),
        ),
        paper_bgcolor=cor_fundo,  # Cor de fundo da área do papel
        plot_bgcolor=cor_fundo,  # Cor de fundo da área de plotagem
        width=largura_px,
        height=altura_px,
        bargap=0.08,  # Espaçamento entre as barras do histograma
        margin=dict(
            l=margem_esquerda,
            r=margem_direita,
            t=margem_superior,
            b=margem_inferior,
        ),
        legend=dict(
            x=posicao_legenda_x,
            y=posicao_legenda_y,
            xanchor=ancora_legenda_x,
            yanchor=ancora_legenda_y,
            bgcolor=fundo_legenda,
            bordercolor=borda_legenda,
            borderwidth=espessura_borda_legenda,
            font=dict(size=tamanho_fonte_legenda),
        ),
    )

    # ==========================================================
    # 17. Subtítulo abaixo do título
    # ==========================================================
    # Adiciona o subtítulo como uma anotação para maior controle de posicionamento.
    fig.add_annotation(
        x=0.5,
        y=posicao_subtitulo,
        xref="paper",
        yref="paper",
        text=subtitulo_final,
        showarrow=False,
        xanchor="center",
        yanchor="top",
        font=dict(
            size=tamanho_fonte_subtitulo,
            color=cor_subtitulo
        ),
        align="center"
    )

    # ==========================================================
    # 18. Eixos
    # ==========================================================
    # Atualiza as propriedades dos eixos Y do histograma.
    fig.update_yaxes(
        title_text=y_label_hist,
        title_font=dict(size=tamanho_fonte_eixos),
        tickfont=dict(size=tamanho_fonte_ticks),
        row=1,
        col=1,
    )

    # Atualiza as propriedades dos eixos Y do boxplot (sem grid).
    fig.update_yaxes(
        title_text=y_label_box,
        title_font=dict(size=tamanho_fonte_eixos),
        tickfont=dict(size=tamanho_fonte_ticks),
        showgrid=False,
        row=2,
        col=1,
    )

    # Atualiza as propriedades dos eixos X (compartilhado).
    fig.update_xaxes(
        title_text=x_label,
        title_font=dict(size=tamanho_fonte_eixos),
        tickfont=dict(size=tamanho_fonte_ticks),
        row=2,
        col=1,
    )

    # Aplica escala logarítmica ao eixo X se `usar_log` for True.
    if usar_log:
        fig.update_xaxes(type="log", row=1, col=1)
        fig.update_xaxes(type="log", row=2, col=1)

    # ==========================================================
    # 19. Resumo estatístico
    # ==========================================================
    # Compila um dicionário com um resumo das estatísticas calculadas e parâmetros usados.
    resumo = {
        "n": total,
        "nbins": nbins,
        "media": media,
        "mediana": mediana,
        "minimo": minimo,
        "maximo": maximo,
        "desvio_padrao": desvio_padrao,
        "q1": float(q1),
        "q3": float(q3),
        "iqr": float(iqr),
        "limite_inferior_outlier": float(fence_low),
        "limite_superior_outlier": float(fence_high),
        "outliers_abaixo": out_low,
        "outliers_acima": out_high,
        "usar_log": usar_log,
        "incluir_zeros": incluir_zeros,
    }

    # ==========================================================
    # 20. Exibição
    # ==========================================================
    # Exibe a figura se `exibir_figura` for True.
    if exibir_figura:
        fig.show()

    # Retorna a figura e o dicionário de resumo.
    return fig, resumo
```

# Dados de demostração

```python
def gerar_df_teste_variavel_continua(
    n=2000,
    seed=42,
    nome_coluna_valor="valor",
    incluir_zeros=True,
    proporcao_zeros=0.12,
    proporcao_outliers=0.01,
    media_log=7.2,
    sigma_log=0.9,
):
    """
    Gera um DataFrame sintético para testar gráficos e análises
    de uma variável numérica contínua.

    Estrutura:
    - id
    - grupo
    - categoria
    - segmento
    - status
    - valor numérico principal
    - faixa_valor

    Parâmetros
    ----------
    n : int
        Número de registros.

    seed : int
        Semente para reprodutibilidade.

    nome_coluna_valor : str
        Nome da coluna numérica principal.

    incluir_zeros : bool
        Se True, insere parte dos registros com valor zero.

    proporcao_zeros : float
        Proporção de zeros.

    proporcao_outliers : float
        Proporção de outliers.

    media_log : float
        Média da distribuição log-normal.

    sigma_log : float
        Dispersão da distribuição log-normal.

    Retorno
    -------
    df : pandas.DataFrame
    """
    import numpy as np
    import pandas as pd

    np.random.seed(seed)

    # ==========================================================
    # 1. Identificador
    # ==========================================================
    ids = np.arange(1, n + 1)

    # ==========================================================
    # 2. Variáveis auxiliares
    # ==========================================================
    grupo = np.random.choice(
        ["Grupo A", "Grupo B", "Grupo C"],
        size=n,
        p=[0.40, 0.35, 0.25]
    )

    categoria = np.random.choice(
        ["Categoria 1", "Categoria 2", "Categoria 3", "Categoria 4"],
        size=n,
        p=[0.30, 0.25, 0.25, 0.20]
    )

    segmento = np.random.choice(
        ["Premium", "Intermediário", "Base"],
        size=n,
        p=[0.20, 0.35, 0.45]
    )

    status = np.random.choice(
        ["Ativo", "Inativo"],
        size=n,
        p=[0.85, 0.15]
    )

    # ==========================================================
    # 3. Variável contínua principal
    # ==========================================================
    valor = np.random.lognormal(
        mean=media_log,
        sigma=sigma_log,
        size=n
    )

    # Pequenos multiplicadores para dar mais realismo
    mult_grupo = np.select(
        [grupo == "Grupo A", grupo == "Grupo B", grupo == "Grupo C"],
        [1.00, 1.12, 0.92],
        default=1.00
    )

    mult_segmento = np.select(
        [segmento == "Premium", segmento == "Intermediário", segmento == "Base"],
        [1.25, 1.00, 0.85],
        default=1.00
    )

    valor = valor * mult_grupo * mult_segmento

    # ==========================================================
    # 4. Inclusão de zeros
    # ==========================================================
    if incluir_zeros and proporcao_zeros > 0:
        qtd_zeros = int(n * proporcao_zeros)
        idx_zeros = np.random.choice(n, qtd_zeros, replace=False)
        valor[idx_zeros] = 0

    # ==========================================================
    # 5. Inclusão de outliers
    # ==========================================================
    if proporcao_outliers > 0:
        qtd_outliers = max(1, int(n * proporcao_outliers))
        idx_outliers = np.random.choice(n, qtd_outliers, replace=False)
        valor[idx_outliers] = valor[idx_outliers] * np.random.uniform(5, 12, qtd_outliers)

    # ==========================================================
    # 6. Faixas para interpretação
    # ==========================================================
    positivos = valor[valor > 0]

    if len(positivos) > 0:
        p50 = np.percentile(positivos, 50)
        p90 = np.percentile(positivos, 90)
    else:
        p50 = 0
        p90 = 0

    faixa_valor = np.where(
        valor == 0,
        "Zero",
        np.where(
            valor < p50,
            "Baixo",
            np.where(valor < p90, "Médio", "Alto")
        )
    )

    # ==========================================================
    # 7. Montagem final
    # ==========================================================
    df = pd.DataFrame({
        "id": ids,
        "grupo": grupo,
        "categoria": categoria,
        "segmento": segmento,
        "status": status,
        nome_coluna_valor: valor,
        "faixa_valor": faixa_valor,
    })

    return df
```
