#Set up

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

# Função do gráfico

```python
def plot_boxplot_horizontal_por_categoria(
    df,
    col_categoria,
    col_valor,
    titulo="Distribuição por categoria",
    subtitulo=None,
    ordenar_por="media",
    ordem_crescente=True,
    mostrar_linha_media=True,
    mostrar_linha_referencia_global=False,
    referencia_global_tipo="media",
    referencia_global_rotulo=None,
    mostrar_resumo_estatistico=True,
    resumo_estatistico_linhas=None,
    resumo_prefixos=None,
    casas_decimais_resumo=1,
    boxpoints="outliers",
    categoria_titulo="Categoria",
    valor_titulo="Valor",
    largura=1600,
    altura=None,
    altura_por_categoria=110,
    altura_minima=500,
    opacidade_box=0.7,
    cor_box="rgba(31, 119, 180, 0.60)",
    cor_outliers="rgba(31, 119, 180, 0.95)",
    tamanho_outliers=5,
    cores_por_categoria=False,
    paleta_cores=None,
    cor_linha_media="black",
    largura_linha_media=2,
    estilo_linha_media="dash",
    cor_linha_referencia_global="rgba(120,120,120,0.9)",
    largura_linha_referencia_global=2,
    estilo_linha_referencia_global="dot",
    cor_texto_resumo="rgba(40,40,40,1)",
    bgcolor_texto_resumo="rgba(255,255,255,0.88)",
    bordercolor_texto_resumo="rgba(0,0,0,0.12)",
    borderwidth_texto_resumo=1,
    borderpad_texto_resumo=4,
    tamanho_fonte_geral=14,
    tamanho_fonte_titulo=20,
    tamanho_fonte_subtitulo=13,
    tamanho_fonte_eixos=18,
    tamanho_fonte_resumo=11,
    mostrar_grid_x=True,
    mostrar_grid_y=False,
    cor_grid_x="rgba(0,0,0,0.15)",
    cor_grid_y="rgba(0,0,0,0.0)",
    dtick=None,
    faixa_x=None,
    proporcao_inicio_coluna_resumo=0.80,
    margem_esquerda=180,
    margem_direita=340,
    margem_superior=90,
    margem_inferior=70,
    template="plotly_white",
    remover_nulos=True,
    exibir=True,
):
    """
    Gera um boxplot horizontal por categoria com bloco estatístico alinhado
    em coluna fixa à direita.

    Recursos principais
    -------------------
    - Ordenação por média, mediana, frequência ou nome: Permite organizar as categorias
      no eixo Y de acordo com a métrica escolhida, facilitando a comparação.
    - Linha vertical da média por categoria: Destaca a média de cada categoria,
      fornecendo um ponto de referência visual importante.
    - Linha de referência global: Uma linha vertical opcional que representa uma
      métrica (média ou mediana) do conjunto de dados completo, útil para contextualizar
      as distribuições individuais.
    - Bloco estatístico à direita com coluna fixa: Exibe as principais estatísticas
      (mínimo, máximo, quartis, média, mediana, outliers, contagem) de cada categoria
      em um formato tabular fixo, melhorando a legibilidade e a análise.
    - Opção de usar a mesma cor em todas as categorias ou cores diferentes:
      Permite personalizar a visualização para destacar ou agrupar categorias.
    - Controle fino de layout, espaçamento, fontes e eixos: Oferece flexibilidade
      para adaptar o gráfico a diferentes necessidades de apresentação.

    Parâmetros principais
    ---------------------
    df : pandas.DataFrame
        DataFrame de entrada contendo os dados a serem plotados. Deve ter pelo menos
        uma coluna categórica e uma coluna numérica.

    col_categoria : str
        Nome da coluna no `df` que contém as categorias a serem exibidas no eixo Y.
        Esta coluna será usada para agrupar os dados e criar os boxplots individuais.

    col_valor : str
        Nome da coluna no `df` que contém os valores numéricos a serem analisados e
        exibidos no eixo X de cada boxplot.

    titulo : str, default "Distribuição por categoria"
        Título principal do gráfico. Será exibido no topo da figura.

    subtitulo : str ou None, default None
        Subtítulo do gráfico. Se None, um subtítulo informativo será gerado
        automaticamente com base nas opções de ordenação e linhas de referência.

    ordenar_por : str, default "media"
        Critério de ordenação das categorias no eixo Y. Controla a ordem em que os
        boxplots serão apresentados. As opções válidas são:
        - "media": Ordena as categorias pela média dos valores.
        - "mediana": Ordena as categorias pela mediana dos valores.
        - "frequencia": Ordena as categorias pelo número de observações (contagem).
        - "nome": Ordena as categorias alfabeticamente pelo nome.

    ordem_crescente : bool, default True
        Se True, as categorias serão ordenadas em ordem crescente de acordo com
        o `ordenar_por`. Se False, serão ordenadas em ordem decrescente.

    mostrar_linha_media : bool, default True
        Se True, uma linha vertical pontilhada indicando a média de cada categoria
        será desenhada sobre o boxplot correspondente.

    mostrar_linha_referencia_global : bool, default False
        Se True, uma linha vertical pontilhada será adicionada ao gráfico para
        representar uma métrica global (média ou mediana) de todos os dados.

    referencia_global_tipo : str, default "media"
        Tipo de métrica a ser usada para a linha de referência global. Opções:
        - "media": Usa a média de `col_valor` de todo o `df`.
        - "mediana": Usa a mediana de `col_valor` de todo o `df`.
        Este parâmetro é relevante apenas se `mostrar_linha_referencia_global` for True.

    referencia_global_rotulo : str ou None, default None
        Rótulo opcional para a linha de referência global. Se None, um rótulo
        automático será gerado (ex: "Referência global: média = XX.X").

    mostrar_resumo_estatistico : bool, default True
        Se True, um bloco de texto com as estatísticas descritivas de cada categoria
        será exibido à direita do gráfico.

    resumo_estatistico_linhas : list[list[str]] ou None, default None
        Define o layout e quais estatísticas serão exibidas no bloco de resumo.
        É uma lista de listas, onde cada sublista representa uma linha do resumo.
        Exemplo: `[['media', 'mediana'], ['minimo', 'maximo']]`.
        Se None, um conjunto padrão de estatísticas será usado.
        Estatísticas disponíveis: 'media', 'mediana', 'minimo', 'maximo', 'q1', 'q3',
        'iqr', 'limite_inferior', 'limite_superior', 'outliers_abaixo', 'outliers_acima', 'n'.

    resumo_prefixos : dict ou None, default None
        Dicionário para mapear os nomes das estatísticas (chaves) para prefixos
        personalizados (valores) no bloco de resumo. Exemplo: `{'media': 'Méd'}`.
        Se None, prefixos padrão serão usados.

    casas_decimais_resumo : int, default 1
        Número de casas decimais para formatar os valores numéricos no bloco de resumo
        estatístico.

    boxpoints : str, default "outliers"
        Define quais pontos individuais serão exibidos no boxplot. Opções:
        - "all": Mostra todos os pontos.
        - "outliers": Mostra apenas os pontos identificados como outliers (padrão).
        - False: Não mostra nenhum ponto individual.
        - "suspectedoutliers": Mostra apenas outliers suspeitos.

    categoria_titulo : str, default "Categoria"
        Título do eixo Y (eixo das categorias).

    valor_titulo : str, default "Valor"
        Título do eixo X (eixo dos valores numéricos).

    largura : int, default 1600
        Largura total da figura em pixels.

    altura : int ou None, default None
        Altura total da figura em pixels. Se None, a altura será calculada
        automaticamente com base no número de categorias e `altura_por_categoria`,
        respeitando `altura_minima`.

    altura_por_categoria : int, default 110
        Altura (em pixels) reservada para cada categoria ao calcular a `altura`
        automática. Usado apenas se `altura` for None.

    altura_minima : int, default 500
        Altura mínima (em pixels) para a figura ao calcular a `altura` automática.
        Usado apenas se `altura` for None.

    opacidade_box : float, default 0.7
        Opacidade das caixas dos boxplots (valor entre 0 e 1).

    cor_box : str, default "rgba(31, 119, 180, 0.60)"
        Cor padrão das caixas dos boxplots. Usado se `cores_por_categoria` for False.
        Pode ser um nome de cor CSS, RGB, RGBA ou código hexadecimal.

    cor_outliers : str, default "rgba(31, 119, 180, 0.95)"
        Cor dos pontos de outliers. Usado se `cores_por_categoria` for False.

    tamanho_outliers : int, default 5
        Tamanho dos marcadores de outliers.

    cores_por_categoria : bool, default False
        Se True, cada categoria receberá uma cor diferente da paleta `paleta_cores`.
        Se False, todas as categorias usarão a `cor_box`.

    paleta_cores : list[str] ou None, default None
        Lista de strings de cores (ex: `['#FF0000', '#00FF00']`) para usar quando
        `cores_por_categoria` for True. Se None, a paleta qualitativa padrão do Plotly
        (`px.colors.qualitative.Plotly`) será utilizada.

    cor_linha_media : str, default "black"
        Cor da linha vertical que indica a média de cada categoria.

    largura_linha_media : int, default 2
        Largura da linha de média por categoria.

    estilo_linha_media : str, default "dash"
        Estilo da linha de média por categoria. Opções: "solid", "dot", "dash".

    cor_linha_referencia_global : str, default "rgba(120,120,120,0.9)"
        Cor da linha de referência global.

    largura_linha_referencia_global : int, default 2
        Largura da linha de referência global.

    estilo_linha_referencia_global : str, default "dot"
        Estilo da linha de referência global. Opções: "solid", "dot", "dash".

    cor_texto_resumo : str, default "rgba(40,40,40,1)"
        Cor da fonte do texto no bloco de resumo estatístico.

    bgcolor_texto_resumo : str, default "rgba(255,255,255,0.88)"
        Cor de fundo do bloco de texto de resumo estatístico.

    bordercolor_texto_resumo : str, default "rgba(0,0,0,0.12)"
        Cor da borda do bloco de texto de resumo estatístico.

    borderwidth_texto_resumo : int, default 1
        Largura da borda do bloco de texto de resumo estatístico.

    borderpad_texto_resumo : int, default 4
        Preenchimento interno (padding) da borda do bloco de texto de resumo estatístico.

    tamanho_fonte_geral : int, default 14
        Tamanho da fonte padrão para a maioria dos textos do gráfico.

    tamanho_fonte_titulo : int, default 20
        Tamanho da fonte para o título principal do gráfico.

    tamanho_fonte_subtitulo : int, default 13
        Tamanho da fonte para o subtítulo do gráfico.

    tamanho_fonte_eixos : int, default 18
        Tamanho da fonte para os títulos dos eixos X e Y.

    tamanho_fonte_resumo : int, default 11
        Tamanho da fonte para o texto dentro do bloco de resumo estatístico.

    mostrar_grid_x : bool, default True
        Se True, as linhas de grade verticais no eixo X serão exibidas.

    mostrar_grid_y : bool, default False
        Se True, as linhas de grade horizontais no eixo Y serão exibidas.

    cor_grid_x : str, default "rgba(0,0,0,0.15)"
        Cor das linhas de grade do eixo X.

    cor_grid_y : str, default "rgba(0,0,0,0.0)"
        Cor das linhas de grade do eixo Y.

    dtick : int ou None, default None
        Define o intervalo entre os ticks principais no eixo X. Se None, o Plotly
        decide automaticamente.

    faixa_x : list[int/float] ou None, default None
        Define o intervalo mínimo e máximo para o eixo X. Ex: `[0, 100]`.
        Se None, o Plotly determina o intervalo automaticamente.

    proporcao_inicio_coluna_resumo : float, default 0.80
        Define a posição horizontal (em proporção do papel, de 0 a 1) onde o bloco
        de resumo estatístico começará. Por exemplo, 0.80 significa que o bloco
        começará a 80% da largura do gráfico a partir da esquerda.

    margem_esquerda : int, default 180
        Margem esquerda da figura em pixels. Útil para acomodar rótulos longos
        do eixo Y.

    margem_direita : int, default 340
        Margem direita da figura em pixels. Essencial para acomodar o bloco de
        resumo estatístico.

    margem_superior : int, default 90
        Margem superior da figura em pixels.

    margem_inferior : int, default 70
        Margem inferior da figura em pixels.

    template : str, default "plotly_white"
        Tema de layout do Plotly a ser usado. Ex: "plotly", "plotly_white", "plotly_dark".

    remover_nulos : bool, default True
        Se True, as linhas que contêm valores nulos nas colunas `col_categoria` ou
        `col_valor` serão removidas antes da plotagem.

    exibir : bool, default True
        Se True, a figura será exibida imediatamente após a geração.
        Se False, a figura será apenas retornada sem ser exibida.

    Retorno
    -------
    fig : plotly.graph_objects.Figure
        Objeto Figure do Plotly pronto para ser exibido ou salvo.
    """

    # ==========================================================================
    # Validação e preparação inicial dos dados
    # ==========================================================================
    # Cria uma cópia do DataFrame original para evitar modificações indesejadas
    dados = df.copy()

    # Remove linhas com valores nulos nas colunas essenciais
    # Se remover_nulos for True, quaisquer linhas onde col_categoria ou col_valor são NaN
    # serão descartadas. Isso garante que o boxplot e as estatísticas sejam calculados
    # apenas com dados válidos.
    if remover_nulos:
        dados = dados.dropna(subset=[col_categoria, col_valor])

    # Verifica se o DataFrame resultante está vazio após a remoção de nulos
    # Se estiver vazio, não há dados para plotar, então levanta um erro.
    if dados.empty:
        raise ValueError("O DataFrame ficou vazio após remover valores nulos.")

    # Valida o tipo da coluna de valores, que deve ser numérica
    # Garante que as operações estatísticas (média, mediana, etc.) possam ser realizadas.
    if not pd.api.types.is_numeric_dtype(dados[col_valor]):
        raise TypeError(f"A coluna '{col_valor}' precisa ser numérica.")

    # Converte a coluna de categoria para string para garantir tratamento consistente
    # Isso evita problemas com tipos de dados mistos ou numéricos sendo interpretados
    # como categorias.
    dados[col_categoria] = dados[col_categoria].astype(str)

    # Define as linhas padrão para o resumo estatístico se não forem fornecidas
    # Cada sublista representa uma linha no bloco de resumo, com as chaves das estatísticas.
    if resumo_estatistico_linhas is None:
        resumo_estatistico_linhas = [
            ["media", "mediana"],
            ["minimo", "maximo"],
            ["limite_inferior", "limite_superior"],
            ["outliers_abaixo", "outliers_acima"],
        ]

    # Define os prefixos padrão para as estatísticas no resumo se não forem fornecidos
    # Estes prefixos são usados para tornar o texto do resumo mais conciso (ex: "Méd" para "Média").
    if resumo_prefixos is None:
        resumo_prefixos = {
            "media": "Méd",
            "mediana": "Med",
            "minimo": "Min",
            "maximo": "Máx",
            "limite_inferior": "LI",
            "limite_superior": "LS",
            "outliers_abaixo": "Out↓",
            "outliers_acima": "Out↑",
            "q1": "Q1",
            "q3": "Q3",
            "iqr": "IQR",
            "n": "n",
        }

    # ==========================================================================
    # Funções auxiliares para cálculo e formatação de estatísticas
    # ==========================================================================
    def calcular_estatisticas(serie):
        """
        Calcula um conjunto de estatísticas descritivas para uma série numérica.

        Parâmetros:
        -----------
        serie : pandas.Series
            Série numérica contendo os dados para os quais as estatísticas serão calculadas.

        Retorno:
        --------
        pandas.Series
            Série contendo a média, mediana, mínimo, máximo, quartis (Q1, Q3),
            intervalo interquartil (IQR), limites dos bigodes (inferior e superior),
            contagem de outliers abaixo e acima, e a contagem total de elementos (n).
        """
        # Garante que a série é numérica e remove valores nulos para cálculo preciso.
        s = pd.Series(serie).dropna().astype(float)

        # Calcula os quartis (25º, 50º (mediana) e 75º percentil).
        q1 = s.quantile(0.25)
        mediana = s.quantile(0.50)
        q3 = s.quantile(0.75)

        # Calcula o Intervalo Interquartil (IQR), que é a diferença entre Q3 e Q1.
        iqr = q3 - q1

        # Calcula os limites inferior e superior para identificar outliers.
        # Valores fora desses limites são considerados outliers.
        limite_inferior = q1 - 1.5 * iqr
        limite_superior = q3 + 1.5 * iqr

        # Conta o número de outliers abaixo do limite inferior e acima do limite superior.
        outliers_abaixo = int((s < limite_inferior).sum())
        outliers_acima = int((s > limite_superior).sum())

        # Retorna todas as estatísticas calculadas como uma série do pandas.
        return pd.Series({
            "media": s.mean(),
            "mediana": mediana,
            "minimo": s.min(),
            "maximo": s.max(),
            "q1": q1,
            "q3": q3,
            "iqr": iqr,
            "limite_inferior": limite_inferior,
            "limite_superior": limite_superior,
            "outliers_abaixo": outliers_abaixo,
            "outliers_acima": outliers_acima,
            "n": int(s.size), # Número total de observações na série
        })

    def formatar_metrica(chave, valor):
        """
        Formata uma métrica estatística para exibição no bloco de resumo.
        Aplica prefixos e formatação de casas decimais conforme configurado.

        Parâmetros:
        -----------
        chave : str
            A chave da métrica (ex: 'media', 'n').
        valor : float ou int
            O valor numérico da métrica.

        Retorno:
        --------
        str
            A métrica formatada (ex: 'Méd 45.6', 'n 120').
        """
        # Verifica se a métrica é uma contagem de outliers ou 'n' (número de observações).
        # Para estas, o valor é formatado como inteiro.
        if chave in ["outliers_abaixo", "outliers_acima", "n"]:
            return f"{resumo_prefixos.get(chave, chave)} {int(valor)}"
        # Para outras métricas (como média, mediana), o valor é formatado com casas decimais.
        return f"{resumo_prefixos.get(chave, chave)} {valor:.{casas_decimais_resumo}f}"

    def montar_texto_resumo(est):
        """
        Monta a string de texto completa para o bloco de resumo estatístico de uma categoria.
        Organiza as métricas em linhas e colunas conforme definido por `resumo_estatistico_linhas`.

        Parâmetros:
        -----------
        est : pandas.Series
            Série contendo as estatísticas calculadas para uma única categoria.

        Retorno:
        --------
        str
            String formatada com tags HTML para exibição no Plotly como uma anotação.
        """
        linhas = []
        # Itera sobre cada grupo de estatísticas que forma uma linha no resumo.
        for grupo in resumo_estatistico_linhas:
            partes = []
            # Itera sobre cada chave de estatística dentro do grupo (coluna da linha).
            for chave in grupo:
                # Se a chave da estatística existe nos dados calculados, formata e adiciona.
                if chave in est.index:
                    partes.append(formatar_metrica(chave, est[chave]))
            # Se a linha tem partes (estatísticas formatadas), une-as com " | " e adiciona.
            if partes:
                linhas.append(" | ".join(partes))
        # Une todas as linhas com quebras de linha HTML para formar o texto final.
        return "<br>".join(linhas)

    # ==========================================================================
    # Cálculo das estatísticas por categoria e ordenação
    # ==========================================================================
    # Agrupa o DataFrame pela coluna de categoria e aplica a função
    # calcular_estatisticas a cada grupo. O .unstack() transforma o resultado
    # para que as estatísticas se tornem colunas, e .reset_index() move a
    # coluna de categoria de volta para um índice comum.
    estatisticas = (
        dados.groupby(col_categoria)[col_valor]
        .apply(calcular_estatisticas)
        .unstack()
        .reset_index()
    )

    # Valida o parâmetro 'ordenar_por' para garantir que é uma opção suportada.
    if ordenar_por not in ["media", "mediana", "frequencia", "nome"]:
        raise ValueError("Parâmetro 'ordenar_por' inválido. Use 'media', 'mediana', 'frequencia' ou 'nome'.")

    # Ordena as estatísticas (e, consequentemente, as categorias) com base no critério escolhido.
    if ordenar_por == "nome":
        estatisticas = estatisticas.sort_values(by=col_categoria, ascending=ordem_crescente)
    elif ordenar_por == "frequencia":
        # Para ordenar por frequência, usa a coluna 'n' (contagem) das estatísticas.
        estatisticas = estatisticas.sort_values(by="n", ascending=ordem_crescente)
    else:
        # Para ordenar por média ou mediana, usa a coluna correspondente.
        estatisticas = estatisticas.sort_values(by=ordenar_por, ascending=ordem_crescente)

    # Extrai a lista de categorias ordenadas para uso posterior na criação do gráfico e eixos.
    categorias_ordenadas = estatisticas[col_categoria].tolist()
    # Cria um mapa de dicionários para acesso rápido às estatísticas de cada categoria.
    mapa_est = estatisticas.set_index(col_categoria).to_dict(orient="index")

    # ==========================================================================
    # Cálculo da altura automática da figura
    # ==========================================================================
    # Se a altura não for especificada, calcula uma altura dinâmica.
    # A altura será o maior valor entre a altura mínima definida e o produto
    # do número de categorias pela altura_por_categoria. Isso garante que
    # todas as categorias tenham espaço adequado e que o gráfico não seja muito pequeno.
    if altura is None:
        altura = max(altura_minima, len(categorias_ordenadas) * altura_por_categoria)

    # ==========================================================================
    # Lógica de cores por categoria
    # ==========================================================================
    # Determina o esquema de cores para os boxplots.
    if cores_por_categoria:
        # Se 'cores_por_categoria' for True, usa a paleta fornecida ou a paleta padrão do Plotly.
        paleta = paleta_cores if paleta_cores is not None else px.colors.qualitative.Plotly
        # Mapeia cada categoria a uma cor da paleta, ciclando se houver mais categorias que cores.
        mapa_cores = {
            categoria: paleta[i % len(paleta)]
            for i, categoria in enumerate(categorias_ordenadas)
        }
    else:
        # Se 'cores_por_categoria' for False, todas as categorias usam a mesma cor definida em 'cor_box'.
        mapa_cores = {
            categoria: cor_box
            for categoria in categorias_ordenadas
        }

    # ==========================================================================
    # Criação da figura Plotly
    # ==========================================================================
    # Inicializa um objeto Figure vazio do Plotly.
    fig = go.Figure()

    # Adiciona um boxplot para cada categoria ordenada.
    for categoria in categorias_ordenadas:
        # Filtra os dados para a categoria atual.
        dados_filtrados = dados[dados[col_categoria] == categoria]
        # Obtém a cor para a categoria atual do mapa de cores.
        cor_categoria = mapa_cores[categoria]

        # Define a cor dos outliers. Se 'cores_por_categoria' for True, os outliers
        # terão a mesma cor da caixa da categoria. Caso contrário, usam 'cor_outliers'.
        cor_outlier_categoria = cor_categoria if cores_por_categoria else cor_outliers

        # Adiciona um traço (trace) de boxplot horizontal para a categoria atual.
        fig.add_trace(
            go.Box(
                x=dados_filtrados[col_valor],  # Valores no eixo X
                y=[categoria] * len(dados_filtrados),  # Categoria repetida para o eixo Y
                name=categoria,  # Nome para a legenda (embora a legenda esteja desativada)
                orientation="h",  # Define o boxplot como horizontal
                boxpoints=boxpoints,  # Controla a exibição de pontos individuais (outliers, todos, nenhum)
                showlegend=False,  # Não mostra a legenda para cada boxplot individual
                marker=dict(
                    color=cor_categoria,  # Cor da caixa do boxplot
                    opacity=opacidade_box,  # Opacidade da caixa
                    outliercolor=cor_outlier_categoria,  # Cor dos outliers
                    size=tamanho_outliers,  # Tamanho dos marcadores de outliers
                ),
                line=dict(color=cor_categoria),  # Cor da linha da caixa
                # Configura o texto que aparece ao passar o mouse (tooltip).
                # Usa um f-string para incluir o nome da categoria e o valor do ponto.
                hovertemplate=f"{categoria}<br>{valor_titulo}: %{{x}}<extra></extra>",
            )
        )

    # ==========================================================================
    # Adição de formas (linhas) e anotações (texto) à figura
    # ==========================================================================
    shapes = []  # Lista para armazenar as formas (linhas)
    annotations = []  # Lista para armazenar as anotações (blocos de texto)

    # Itera sobre as categorias ordenadas para adicionar linhas de média e resumos estatísticos.
    for i, categoria in enumerate(categorias_ordenadas):
        est = mapa_est[categoria]  # Obtém as estatísticas para a categoria atual.

        # Adiciona a linha vertical da média por categoria, se 'mostrar_linha_media' for True.
        if mostrar_linha_media:
            shapes.append(
                dict(
                    type="line",  # Tipo da forma: linha
                    x0=est["media"],  # Posição inicial X (média da categoria)
                    x1=est["media"],  # Posição final X (mesma que a inicial para linha vertical)
                    y0=i - 0.4,  # Posição inicial Y (centralizada na categoria com offset)
                    y1=i + 0.4,  # Posição final Y (centralizada na categoria com offset)
                    xref="x",  # Coordenada X relativa ao eixo X do gráfico
                    yref="y",  # Coordenada Y relativa ao eixo Y do gráfico
                    line=dict(
                        color=cor_linha_media,  # Cor da linha de média
                        width=largura_linha_media,  # Largura da linha
                        dash=estilo_linha_media,  # Estilo da linha (sólida, pontilhada, tracejada)
                    ),
                )
            )

        # Adiciona o bloco de resumo estatístico, se 'mostrar_resumo_estatistico' for True.
        if mostrar_resumo_estatistico:
            annotations.append(
                dict(
                    x=proporcao_inicio_coluna_resumo,  # Posição X em proporção do papel (fixa)
                    y=categoria,  # Posição Y na altura da categoria correspondente
                    xref="paper",  # Coordenada X relativa à figura inteira (0 a 1)
                    yref="y",  # Coordenada Y relativa ao eixo Y do gráfico
                    text=montar_texto_resumo(pd.Series(est)),  # Texto do resumo formatado
                    showarrow=False,  # Não mostra a seta da anotação
                    xanchor="left",  # Ancoragem do texto à esquerda na posição X
                    yanchor="middle",  # Ancoragem do texto ao meio na posição Y
                    align="left",  # Alinhamento do texto dentro do bloco
                    font=dict(
                        size=tamanho_fonte_resumo,  # Tamanho da fonte do resumo
                        color=cor_texto_resumo,  # Cor da fonte
                    ),
                    bgcolor=bgcolor_texto_resumo,  # Cor de fundo do bloco de texto
                    bordercolor=bordercolor_texto_resumo,  # Cor da borda do bloco
                    borderwidth=borderwidth_texto_resumo,  # Largura da borda
                    borderpad=borderpad_texto_resumo,  # Preenchimento da borda
                )
            )

    # Adiciona a linha de referência global, se 'mostrar_linha_referencia_global' for True.
    if mostrar_linha_referencia_global:
        valor_referencia = 0  # Inicializa o valor de referência.
        rotulo_auto = ""  # Inicializa o rótulo automático.

        # Calcula o valor da referência global com base no tipo escolhido (média ou mediana).
        if referencia_global_tipo == "media":
            valor_referencia = dados[col_valor].mean()
            rotulo_auto = f"Referência global: média = {valor_referencia:.1f}"
        elif referencia_global_tipo == "mediana":
            valor_referencia = dados[col_valor].median()
            rotulo_auto = f"Referência global: mediana = {valor_referencia:.1f}"
        else:
            # Se o tipo de referência global for inválido, levanta um erro.
            raise ValueError("Parâmetro 'referencia_global_tipo' inválido. Use 'media' ou 'mediana'.")

        # Usa o rótulo fornecido ou o rótulo automático gerado.
        rotulo_final = referencia_global_rotulo if referencia_global_rotulo is not None else rotulo_auto

        # Adiciona a linha de referência global como uma forma.
        shapes.append(
            dict(
                type="line",
                x0=valor_referencia,  # Posição X da linha
                x1=valor_referencia,  # Posição X da linha
                y0=-0.5,  # Posição Y inicial (abaixo da primeira categoria)
                y1=len(categorias_ordenadas) - 0.5,  # Posição Y final (acima da última categoria)
                xref="x",
                yref="y",
                line=dict(
                    color=cor_linha_referencia_global,
                    width=largura_linha_referencia_global,
                    dash=estilo_linha_referencia_global,
                ),
            )
        )

        # Adiciona uma anotação para o rótulo da linha de referência global.
        annotations.append(
            dict(
                x=valor_referencia,  # Posição X na linha de referência
                y=1.03,  # Posição Y acima do gráfico (em proporção do papel)
                xref="x",
                yref="paper",
                text=rotulo_final,
                showarrow=False,
                xanchor="left",
                yanchor="bottom",
                font=dict(size=11, color="rgba(90,90,90,1)"),
                bgcolor="rgba(255,255,255,0.85)",
                bordercolor="rgba(0,0,0,0.15)",
                borderwidth=1,
                borderpad=3,
            )
        )

    # ==========================================================================
    # Geração do subtítulo automático e montagem do título final
    # ==========================================================================
    # Se um subtítulo não foi fornecido, gera um automaticamente com base nas configurações.
    if subtitulo is None:
        partes = []  # Lista para armazenar as partes do subtítulo.

        # Adiciona informações sobre a ordenação das categorias.
        if ordenar_por == "media":
            partes.append("Categorias ordenadas pela média")
        elif ordenar_por == "mediana":
            partes.append("Categorias ordenadas pela mediana")
        elif ordenar_por == "frequencia":
            partes.append("Categorias ordenadas pela frequência")
        elif ordenar_por == "nome":
            partes.append("Categorias ordenadas alfabeticamente")

        # Adiciona informações sobre a linha de média, se mostrada.
        if mostrar_linha_media:
            partes.append("linha tracejada = média")
        # Comentado, mas pode ser incluído se desejado:
        # if mostrar_linha_referencia_global:
        #    partes.append("linha pontilhada = referência global")
        # if mostrar_resumo_estatistico:
        #    partes.append("bloco da direita mostra estatísticas por grupo")
        # if cores_por_categoria:
        #    partes.append("cores diferentes por categoria")

        # Une as partes do subtítulo com " | ".
        subtitulo = " | ".join(partes)

    # Constrói o título final, incorporando o subtítulo se ele existir.
    # Usa tags HTML para controlar o tamanho da fonte do subtítulo.
    titulo_final = titulo
    if subtitulo:
        titulo_final = (
            f"{titulo}"
            f"<br><span style='font-size:{tamanho_fonte_subtitulo}px;'>{subtitulo}</span>"
        )

    # ==========================================================================
    # Configuração do layout da figura
    # ==========================================================================
    # Atualiza o layout da figura com todas as configurações de eixos, títulos, margens, etc.
    fig.update_layout(
        title=dict(
            text=titulo_final,  # Título principal e subtítulo
            font=dict(size=tamanho_fonte_titulo),  # Tamanho da fonte do título
            x=0.08,  # Posição X do título (0.0 a 1.0)
            xanchor="left",  # Ancoragem do título à esquerda
        ),
        xaxis=dict(
            title=valor_titulo,  # Título do eixo X
            title_font=dict(size=tamanho_fonte_eixos),  # Tamanho da fonte do título do eixo X
            showgrid=mostrar_grid_x,  # Exibe ou oculta as grades do eixo X
            gridcolor=cor_grid_x,  # Cor das grades do eixo X
            gridwidth=1,  # Largura das grades do eixo X
            dtick=dtick,  # Intervalo entre os ticks do eixo X
            range=faixa_x,  # Faixa de valores do eixo X
            zeroline=False,  # Não exibe a linha zero do eixo X
        ),
        yaxis=dict(
            title=categoria_titulo,  # Título do eixo Y
            title_font=dict(size=tamanho_fonte_eixos),  # Tamanho da fonte do título do eixo Y
            categoryorder="array",  # Garante que as categorias são ordenadas conforme 'categoryarray'
            categoryarray=categorias_ordenadas,  # A ordem específica das categorias no eixo Y
            showgrid=mostrar_grid_y,  # Exibe ou oculta as grades do eixo Y
            gridcolor=cor_grid_y,  # Cor das grades do eixo Y
            zeroline=False,  # Não exibe a linha zero do eixo Y
        ),
        shapes=shapes,  # Adiciona todas as linhas (média, referência global) como formas
        annotations=annotations,  # Adiciona todos os blocos de texto (resumo estatístico, rótulo de referência global)
        font=dict(size=tamanho_fonte_geral),  # Tamanho da fonte geral do gráfico
        template=template,  # Aplica o tema de layout Plotly
        width=largura,  # Largura total da figura
        height=altura,  # Altura total da figura
        margin=dict(
            l=margem_esquerda,  # Margem esquerda
            r=margem_direita,  # Margem direita
            t=margem_superior,  # Margem superior
            b=margem_inferior,  # Margem inferior
        ),
    )

    # Exibe a figura se 'exibir' for True.
    if exibir:
        fig.show()

    # Retorna o objeto Figure do Plotly.
    return fig
```

```python

def gerar_df_exemplo_boxplot_categoria(seed=42, retornar_parametros=False):
    """
    Gera um DataFrame de exemplo para testar a função de boxplot horizontal.
    """
    np.random.seed(seed)

    parametros = {
        "Asma": {"media": 35, "desvio": 9, "n": 70},
        "Depressão": {"media": 41, "desvio": 10, "n": 75},
        "Obesidade": {"media": 47, "desvio": 11, "n": 85},
        "Diabetes": {"media": 58, "desvio": 9, "n": 80},
        "Hipertensão": {"media": 60, "desvio": 8, "n": 90},
        "Artrite": {"media": 66, "desvio": 7, "n": 65},
    }

    linhas = []

    for categoria, p in parametros.items():
        valores = np.random.normal(
            loc=p["media"],
            scale=p["desvio"],
            size=p["n"]
        )

        valores = np.clip(valores, 18, 95)

        if categoria == "Asma":
            valores = np.append(valores, [77, 80])
        if categoria == "Hipertensão":
            valores = np.append(valores, [35])
        if categoria == "Obesidade":
            valores = np.append(valores, [18, 89])

        for v in valores:
            linhas.append(
                {
                    "patologia_mapeada": categoria,
                    "idade": round(float(v), 1)
                }
            )

    df_exemplo = pd.DataFrame(linhas)

    if retornar_parametros:
        return df_exemplo, parametros

    return df_exemplo
```
