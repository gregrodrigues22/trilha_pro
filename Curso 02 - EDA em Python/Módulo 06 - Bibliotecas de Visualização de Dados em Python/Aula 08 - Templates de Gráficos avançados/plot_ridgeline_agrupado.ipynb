# Função do gráfico

```python
def grafico_ridgeplot(
    df: pd.DataFrame,
    coluna_grupo: str,
    coluna_valor: str,
    titulo: str = "Ridge Plot",
    subtitulo: str | None = None,
    ordem_grupos: list | None = None,
    remover_nulos: bool = True,
    remover_infinitos: bool = True,
    aplicar_filtro_positivo: bool = False,
    minimo_observacoes_por_grupo: int = 1,
    incluir_n_amostral_no_label: bool = False,
    formatar_n_com_ponto: bool = True,
    ordenar_por: str | None = None,
    ascending: bool = True,
    colormap: str | None = None,
    overlap: float | None = None,
    show_lines: bool | None = None,
    width: int = 1200,
    height: int = 600,
    font_size: int = 12,
    plot_bgcolor: str = "white",
    paper_bgcolor: str = "white",
    xaxis_title: str | None = None,
    yaxis_title: str | None = None,
    x_tickprefix: str | None = None,
    x_tickformat: str | None = None,
    mostrar_legenda: bool = False,
    # resumo lateral
    mostrar_resumo_lateral: bool = True,
    resumo_x_paper: float = 1.01,
    resumo_font_size: int = 10,
    margem_direita: int = 280,
    casas_decimais_resumo: int = 1,
    mostrar_media_resumo: bool = True,
    mostrar_mediana_resumo: bool = True,
    mostrar_min_resumo: bool = False,
    mostrar_q1_resumo: bool = True,
    mostrar_q3_resumo: bool = True,
    mostrar_iqr_resumo: bool = True,
    mostrar_max_resumo: bool = False,
    mostrar_n_resumo: bool = True,
    titulo_coluna_resumo: str | None = "Resumo"
):
    """
    Gera um ridge plot padronizado com opção de resumo lateral por grupo.
    O resumo é desenhado em coluna fixa à direita, evitando colisão com legenda.

    Parâmetros:
    -----------
    df : pd.DataFrame
        DataFrame contendo os dados a serem plotados.
    coluna_grupo : str
        Nome da coluna que define os grupos para o ridge plot (eixo Y).
    coluna_valor : str
        Nome da coluna que contém os valores numéricos para as distribuições (eixo X).
    titulo : str, opcional
        Título principal do gráfico. Padrão é "Ridge Plot".
    subtitulo : str | None, opcional
        Subtítulo do gráfico. Padrão é None.
    ordem_grupos : list | None, opcional
        Lista com a ordem desejada dos grupos no eixo Y. Se None, a ordem é determinada por `ordenar_por`.
    remover_nulos : bool, opcional
        Se True, remove linhas com valores nulos em `coluna_grupo` ou `coluna_valor`. Padrão é True.
    remover_infinitos : bool, opcional
        Se True, remove valores infinitos (np.inf, -np.inf) da `coluna_valor`. Padrão é True.
    aplicar_filtro_positivo : bool, opcional
        Se True, filtra apenas valores positivos na `coluna_valor`. Padrão é False.
    minimo_observacoes_por_grupo : int, opcional
        Número mínimo de observações para um grupo ser incluído no plot. Padrão é 1.
    incluir_n_amostral_no_label : bool, opcional
        Se True, inclui o tamanho amostral (n) no rótulo de cada grupo. Padrão é False.
    formatar_n_com_ponto : bool, opcional
        Se True, formata o 'n' amostral com ponto como separador de milhares. Padrão é True.
    ordenar_por : str | None, opcional
        Critério para ordenar os grupos se `ordem_grupos` for None. Opções: 'media', 'mediana', 'min', 'max', 'n'.
        Padrão é None (mantém a ordem original ou alfabética).
    ascending : bool, opcional
        Define se a ordenação deve ser ascendente (True) ou descendente (False). Padrão é True.
    colormap : str | None, opcional
        Nome do colormap a ser usado para colorir as distribuições. Padrão é None (ridgeplot default).
    overlap : float | None, opcional
        Fator de sobreposição entre as distribuições. Padrão é None (ridgeplot default).
    show_lines : bool | None, opcional
        Se True, mostra linhas para a média/mediana nas distribuições. Padrão é None (ridgeplot default).
    width : int, opcional
        Largura do gráfico em pixels. Padrão é 1200.
    height : int, opcional
        Altura do gráfico em pixels. Padrão é 600.
    font_size : int, opcional
        Tamanho da fonte geral do gráfico. Padrão é 12.
    plot_bgcolor : str, opcional
        Cor de fundo da área de plotagem. Padrão é "white".
    paper_bgcolor : str, opcional
        Cor de fundo do papel do gráfico. Padrão é "white".
    xaxis_title : str | None, opcional
        Título do eixo X. Se None, usa `coluna_valor`. Padrão é None.
    yaxis_title : str | None, opcional
        Título do eixo Y. Se None, usa `coluna_grupo`. Padrão é None.
    x_tickprefix : str | None, opcional
        Prefixo para os rótulos do eixo X (ex: "R$"). Padrão é None.
    x_tickformat : str | None, opcional
        Formato dos rótulos do eixo X (ex: ",.0f" para inteiro com separador de milhar). Padrão é None.
    mostrar_legenda : bool, opcional
        Se True, mostra a legenda. Padrão é False.
    mostrar_resumo_lateral : bool, opcional
        Se True, exibe um painel de resumo lateral com estatísticas por grupo. Padrão é True.
    resumo_x_paper : float, opcional
        Posição X (em coordenadas de papel, 0 a 1) do painel de resumo. Padrão é 1.01.
    resumo_font_size : int, opcional
        Tamanho da fonte para o texto do resumo lateral. Padrão é 10.
    margem_direita : int, opcional
        Margem direita do gráfico para acomodar o resumo lateral. Padrão é 280.
    casas_decimais_resumo : int, opcional
        Número de casas decimais para as estatísticas no resumo. Padrão é 1.
    mostrar_media_resumo : bool, opcional
        Se True, mostra a média no resumo lateral. Padrão é True.
    mostrar_mediana_resumo : bool, opcional
        Se True, mostra a mediana no resumo lateral. Padrão é True.
    mostrar_min_resumo : bool, opcional
        Se True, mostra o mínimo no resumo lateral. Padrão é False.
    mostrar_q1_resumo : bool, opcional
        Se True, mostra o primeiro quartil (Q1) no resumo lateral. Padrão é True.
    mostrar_q3_resumo : bool, opcional
        Se True, mostra o terceiro quartil (Q3) no resumo lateral. Padrão é True.
    mostrar_iqr_resumo : bool, opcional
        Se True, mostra o Intervalo Interquartil (IQR) no resumo lateral. Padrão é True.
    mostrar_max_resumo : bool, opcional
        Se True, mostra o máximo no resumo lateral. Padrão é False.
    mostrar_n_resumo : bool, opcional
        Se True, mostra o tamanho amostral (n) no resumo lateral. Padrão é True.
    titulo_coluna_resumo : str | None, opcional
        Título para a coluna do resumo lateral. Padrão é "Resumo".

    Retorna:
    --------
    go.Figure
        Objeto Figure do Plotly contendo o ridge plot.
    """

    # Cria uma cópia do DataFrame original para evitar modificar os dados externos
    base = df.copy()

    # Verifica se as colunas essenciais estão presentes no DataFrame
    colunas_necessarias = [coluna_grupo, coluna_valor]
    colunas_ausentes = [c for c in colunas_necessarias if c not in base.columns]
    if colunas_ausentes:
        raise ValueError(
            f"As seguintes colunas não foram encontradas no DataFrame: {colunas_ausentes}"
        )

    # Converte a coluna de valores para numérico, coercing erros para NaN
    base[coluna_valor] = pd.to_numeric(base[coluna_valor], errors="coerce")

    # Remove valores infinitos (positivos ou negativos) se `remover_infinitos` for True
    if remover_infinitos:
        base[coluna_valor] = base[coluna_valor].replace([np.inf, -np.inf], np.nan)

    # Remove linhas que contêm valores nulos na coluna de grupo ou de valor
    if remover_nulos:
        base = base.dropna(subset=[coluna_grupo, coluna_valor])

    # Aplica filtro para manter apenas valores positivos, se `aplicar_filtro_positivo` for True
    if aplicar_filtro_positivo:
        base = base[base[coluna_valor] > 0].copy()

    # Obtém a lista de grupos únicos presentes nos dados após os filtros
    grupos_unicos = base[coluna_grupo].dropna().unique().tolist()

    # Determina a ordem dos grupos no plot
    if ordem_grupos is not None:
        # Se uma ordem específica for fornecida, usa-a, filtrando apenas grupos presentes nos dados
        grupos_ordenados = [g for g in ordem_grupos if g in grupos_unicos]
    else:
        # Se nenhuma ordem específica for fornecida, ordena com base em uma estatística (média, mediana, etc.)
        if ordenar_por is None:
            # Se não houver critério de ordenação, mantém a ordem natural dos grupos únicos
            grupos_ordenados = grupos_unicos
        else:
            # Agrupa os dados pela coluna de grupo e calcula a estatística escolhida
            agrupado = base.groupby(coluna_grupo)[coluna_valor]

            if ordenar_por == "media":
                estatistica = agrupado.mean()
            elif ordenar_por == "mediana":
                estatistica = agrupado.median()
            elif ordenar_por == "min":
                estatistica = agrupado.min()
            elif ordenar_por == "max":
                estatistica = agrupado.max()
            elif ordenar_por == "n":
                estatistica = agrupado.size() # Contagem de elementos no grupo
            else:
                # Lança um erro se o critério de ordenação for inválido
                raise ValueError(
                    "O parâmetro 'ordenar_por' deve ser: "
                    "'media', 'mediana', 'min', 'max', 'n' ou None."
                )
            # Ordena os grupos com base na estatística calculada
            grupos_ordenados = estatistica.sort_values(ascending=ascending).index.tolist()

    # Função auxiliar para formatar o número 'n' (tamanho amostral)
    def _formatar_n(n: int) -> str:
        # Formata com separador de milhar se `formatar_n_com_ponto` for True
        return f"{n:,}".replace(",", ".") if formatar_n_com_ponto else str(n)

    # Função auxiliar para formatar números decimais no resumo lateral
    def _fmt_num(x) -> str:
        return f"{x:.{casas_decimais_resumo}f}"

    # Listas para armazenar as amostras de dados e os rótulos para o ridgeplot
    my_samples = []
    labels = []
    labels_base = [] # Rótulos sem o 'n' amostral para uso no resumo
    estatisticas_por_grupo = {} # Dicionário para armazenar as estatísticas de cada grupo

    # Itera sobre cada grupo ordenado para preparar os dados e calcular estatísticas
    for grupo in grupos_ordenados:
        # Filtra os valores da coluna_valor para o grupo atual
        valores = base.loc[base[coluna_grupo] == grupo, coluna_valor].dropna().values

        # Ignora grupos que não atingem o mínimo de observações
        if len(valores) < minimo_observacoes_por_grupo:
            continue

        # Calcula as estatísticas descritivas para o grupo
        q1 = np.percentile(valores, 25)
        q3 = np.percentile(valores, 75)
        media = np.mean(valores)
        mediana = np.median(valores)
        minimo = np.min(valores)
        maximo = np.max(valores)
        iqr = q3 - q1
        n = len(valores)

        # Armazena as estatísticas em um dicionário para uso posterior (resumo lateral)
        estatisticas_por_grupo[grupo] = {
            "n": n,
            "media": media,
            "mediana": mediana,
            "min": minimo,
            "q1": q1,
            "q3": q3,
            "iqr": iqr,
            "max": maximo,
        }

        # Adiciona os valores do grupo à lista de amostras para o ridgeplot
        my_samples.append(valores)
        labels_base.append(str(grupo))

        # Cria o rótulo do grupo, opcionalmente incluindo o tamanho amostral
        if incluir_n_amostral_no_label:
            labels.append(f"{grupo} (n={_formatar_n(n)})")
        else:
            labels.append(str(grupo))

    # Se nenhum grupo tiver observações suficientes, levanta um erro
    if len(my_samples) == 0:
        raise ValueError(
            "Nenhum grupo possui observações suficientes após os filtros aplicados."
        )

    # Inspeciona os parâmetros válidos da função `ridgeplot` para compatibilidade
    parametros_validos = inspect.signature(ridgeplot).parameters

    # Prepara os argumentos para a função `ridgeplot`
    kwargs_ridge = {
        "samples": my_samples,
        "labels": labels,
    }

    # Adiciona colormap, overlap e show_lines se forem fornecidos e forem parâmetros válidos
    if colormap is not None and "colormap" in parametros_validos:
        kwargs_ridge["colormap"] = colormap

    if overlap is not None and "overlap" in parametros_validos:
        kwargs_ridge["overlap"] = overlap

    if show_lines is not None and "show_lines" in parametros_validos:
        kwargs_ridge["show_lines"] = show_lines

    # Gera o ridge plot usando a função importada
    fig = ridgeplot(**kwargs_ridge)

    # Define o título final do gráfico, incluindo subtítulo se existir
    if subtitulo:
        titulo_final = f"{titulo}<br><sup>{subtitulo}</sup>"
    else:
        titulo_final = titulo

    # Atualiza o layout do gráfico com as configurações personalizadas
    fig.update_layout(
        title=titulo_final,
        xaxis_title=xaxis_title if xaxis_title is not None else coluna_valor,
        yaxis_title=yaxis_title if yaxis_title is not None else coluna_grupo,
        width=width,
        height=height,
        font=dict(size=font_size),
        plot_bgcolor=plot_bgcolor,
        paper_bgcolor=paper_bgcolor,
        margin=dict(t=80, r=margem_direita, b=60, l=80), # Define as margens, ajustando para o resumo lateral
        showlegend=mostrar_legenda,
    )

    # Configurações opcionais para o eixo X (prefixo, formato de tick)
    xaxis_kwargs = {}
    if x_tickprefix is not None:
        xaxis_kwargs["tickprefix"] = x_tickprefix
    if x_tickformat is not None:
        xaxis_kwargs["tickformat"] = x_tickformat
    if xaxis_kwargs:
        fig.update_xaxes(**xaxis_kwargs)

    # Se o resumo lateral for ativado, adiciona-o ao gráfico
    if mostrar_resumo_lateral:
        grupos_plotados = labels_base # Usa os rótulos base para corresponder às estatísticas
        n_grupos = len(grupos_plotados)

        # Adiciona o título da coluna de resumo, se especificado
        if titulo_coluna_resumo:
            fig.add_annotation(
                x=resumo_x_paper,
                y=1.02, # Posição acima do topo do plot
                xref="paper",
                yref="paper",
                text=f"<b>{titulo_coluna_resumo}</b>",
                showarrow=False,
                xanchor="left",
                yanchor="bottom",
                align="left",
                font=dict(size=resumo_font_size + 1),
            )

        # Adiciona as estatísticas para cada grupo no painel de resumo
        for i, grupo in enumerate(grupos_plotados):
            stats = estatisticas_por_grupo[grupo] # Obtém as estatísticas pré-calculadas

            # Constrói as linhas de texto para o resumo do grupo
            linhas = [f"<b>{grupo}</b>"]

            if mostrar_n_resumo:
                linhas.append(f"n { _formatar_n(stats['n']) }")
            if mostrar_media_resumo:
                linhas.append(f"Méd { _fmt_num(stats['media']) }")
            if mostrar_mediana_resumo:
                linhas.append(f"Med { _fmt_num(stats['mediana']) }")
            if mostrar_min_resumo:
                linhas.append(f"Min { _fmt_num(stats['min']) }")
            if mostrar_q1_resumo:
                linhas.append(f"Q1 { _fmt_num(stats['q1']) }")
            if mostrar_q3_resumo:
                linhas.append(f"Q3 { _fmt_num(stats['q3']) }")
            if mostrar_iqr_resumo:
                linhas.append(f"IQR { _fmt_num(stats['iqr']) }")
            if mostrar_max_resumo:
                linhas.append(f"Max { _fmt_num(stats['max']) }")

            texto = "<br>".join(linhas) # Une as linhas com quebras de linha HTML

            # Calcula a posição vertical para cada bloco de resumo, distribuindo-os
            # A fórmula distribui os blocos uniformemente na área designada do resumo.
            y_paper = 0.88 - i * (0.72 / max(n_grupos - 1, 1)) if n_grupos > 1 else 0.88

            # Adiciona a anotação (bloco de texto) para o resumo do grupo
            fig.add_annotation(
                x=resumo_x_paper,
                y=y_paper,
                xref="paper",
                yref="paper",
                text=texto,
                showarrow=False,
                xanchor="left",
                yanchor="middle",
                align="left",
                font=dict(size=resumo_font_size),
                bgcolor="rgba(255,255,255,0.92)", # Fundo branco semi-transparente
                bordercolor="rgba(180,180,180,0.9)", # Borda cinza semi-transparente
                borderwidth=1,
                borderpad=4,
            )

    return fig

def gerar_df_teste_ridgeplot(seed: int = 42) -> pd.DataFrame:
    np.random.seed(seed)

    configuracoes = [
        {"linha_protocolo": "1ª linha", "doenca": "Artrite Reumatoide", "n": 180, "media": 1800, "desvio": 350},
        {"linha_protocolo": "2ª linha", "doenca": "Artrite Reumatoide", "n": 160, "media": 3200, "desvio": 500},
        {"linha_protocolo": "3ª linha", "doenca": "Artrite Reumatoide", "n": 130, "media": 5200, "desvio": 700},
        {"linha_protocolo": "1ª linha", "doenca": "Asma Grave", "n": 150, "media": 1400, "desvio": 250},
        {"linha_protocolo": "2ª linha", "doenca": "Asma Grave", "n": 120, "media": 2600, "desvio": 450},
        {"linha_protocolo": "3ª linha", "doenca": "Asma Grave", "n": 90, "media": 4100, "desvio": 650},
        {"linha_protocolo": "1ª linha", "doenca": "Dermatite Atópica", "n": 140, "media": 2100, "desvio": 400},
        {"linha_protocolo": "2ª linha", "doenca": "Dermatite Atópica", "n": 110, "media": 3800, "desvio": 600},
        {"linha_protocolo": "Resgate", "doenca": "Angioedema", "n": 50, "media": 9500, "desvio": 1800},
    ]

    registros = []

    for cfg in configuracoes:
        custos = np.random.normal(
            loc=cfg["media"],
            scale=cfg["desvio"],
            size=cfg["n"]
        )
        custos = np.clip(custos, 100, None)

        for custo in custos:
            registros.append(
                {
                    "linha_protocolo": cfg["linha_protocolo"],
                    "doenca": cfg["doenca"],
                    "medicamento_manutencao_custo": float(custo),
                }
            )

    df = pd.DataFrame(registros)

    idx_nulos = np.random.choice(df.index, size=10, replace=False)
    df.loc[idx_nulos, "medicamento_manutencao_custo"] = np.nan

    return df
```

# Dados de demostração

```python
def gerar_df_teste_ridgeplot(seed: int = 42) -> pd.DataFrame:
    """
    Gera um DataFrame sintético para testar a função de ridge plot.

    Retorna
    -------
    pd.DataFrame
        DataFrame com as colunas:
        - linha_protocolo
        - doenca
        - medicamento_manutencao_custo
    """
    np.random.seed(seed)

    configuracoes = [
        {
            "linha_protocolo": "1ª linha",
            "doenca": "Artrite Reumatoide",
            "n": 180,
            "media": 1800,
            "desvio": 350,
        },
        {
            "linha_protocolo": "2ª linha",
            "doenca": "Artrite Reumatoide",
            "n": 160,
            "media": 3200,
            "desvio": 500,
        },
        {
            "linha_protocolo": "3ª linha",
            "doenca": "Artrite Reumatoide",
            "n": 130,
            "media": 5200,
            "desvio": 700,
        },
        {
            "linha_protocolo": "1ª linha",
            "doenca": "Asma Grave",
            "n": 150,
            "media": 1400,
            "desvio": 250,
        },
        {
            "linha_protocolo": "2ª linha",
            "doenca": "Asma Grave",
            "n": 120,
            "media": 2600,
            "desvio": 450,
        },
        {
            "linha_protocolo": "3ª linha",
            "doenca": "Asma Grave",
            "n": 90,
            "media": 4100,
            "desvio": 650,
        },
        {
            "linha_protocolo": "1ª linha",
            "doenca": "Dermatite Atópica",
            "n": 140,
            "media": 2100,
            "desvio": 400,
        },
        {
            "linha_protocolo": "2ª linha",
            "doenca": "Dermatite Atópica",
            "n": 110,
            "media": 3800,
            "desvio": 600,
        },
        {
            "linha_protocolo": "Resgate",
            "doenca": "Angioedema",
            "n": 50,
            "media": 9500,
            "desvio": 1800,
        },
    ]

    registros = []

    for cfg in configuracoes:
        custos = np.random.normal(
            loc=cfg["media"],
            scale=cfg["desvio"],
            size=cfg["n"]
        )

        # Evita custos negativos ou irrealmente baixos
        custos = np.clip(custos, 100, None)

        for custo in custos:
            registros.append(
                {
                    "linha_protocolo": cfg["linha_protocolo"],
                    "doenca": cfg["doenca"],
                    "medicamento_manutencao_custo": float(custo),
                }
            )

    df = pd.DataFrame(registros)

    # Inserir alguns nulos para simular base real
    idx_nulos = np.random.choice(df.index, size=10, replace=False)
    df.loc[idx_nulos, "medicamento_manutencao_custo"] = np.nan

    return df
```
