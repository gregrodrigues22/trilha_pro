üîπ Voc√™ pode usar a fun√ß√£o np.where() para substituir onde tem F por Feminino. √â uma boa escolha de fun√ß√£o quando queremos substituir valores com base em uma √∫nica condi√ß√£o

```python
df_receitas["cliente_sexo"].value_counts()
```

```python
df_receitas["cliente_sexo"] = np.where(df_receitas["cliente_sexo"] == 'F', "Feminino", df_receitas["cliente_sexo"])
```

üîπ No entanto, aqui precisamos substituir dois valores, porque tamb√©m M precisa ser substitu√≠do por Masculino. Nesse caso, √© poss√≠vel utilizar np.select()

```python
# Definir condi√ß√µes
condicoes = [df_receitas["cliente_sexo"] == "F",
             df_receitas["cliente_sexo"] == "M"
             ]

# Definir os valores correspondentes para cada condi√ß√£o. Tem que ter um valor para cada condi√ß√£o
valores = ["Feminino", "Masculino"]

# Aplicar a substitui√ß√£o com np.select() com um valor default
df_receitas["cliente_sexo"] = np.select(condicoes, valores, default= df_receitas["cliente_sexo"])
```

üîπ S√≥ que em programa√ß√£o, existem v√°rios caminhos! Voc√™ ainda pode utilizar fun√ß√µes como replace()

```python
# Substituir diretamente valores espec√≠ficos

df_receitas["cliente_sexo"] = df_receitas["cliente_sexo"].replace({"F": "Feminino", "M": "Masculino"})
```

üîπ Ou a fun√ß√£o map(). S√≥ tome o cuidado, pois se um valor n√£o estiver no dicion√°rio, ele ser√° convertido para NaN.

```python
# Criar um dicion√°rio de mapeamento e aplicar com map().

mapa_sexo = {"F": "Feminino", "M": "Masculino", "Feminino": "Feminino", "Masculino": "Masculino"}

df_receitas["cliente_sexo"] = df_receitas["cliente_sexo"].map(mapa_sexo)
```

üîπ Por fim, utilize a fun√ß√£o apply(), para aplicar uma fun√ß√£o customizada por voc√™!

```python
# Fun√ß√£o personalizada para padronizar a coluna

def padronizar_sexo(valor):
    if valor in ["F", "Feminino"]:
        return "Feminino"
    elif valor in ["M", "Masculino"]:
        return "Masculino"
    return valor #retorna o valor original caso n√£o seja alterado

# Aplicando a fun√ß√£o com apply()
df_receitas["cliente_sexo"] = df_receitas["cliente_sexo"].apply(padronizar_sexo)
```

üîπ Podemos dividir a idade em faixas num√©ricas predefinidas com pd.cut(). O pd.cut() considera o limite inferior fechado e o superior aberto, exceto o √∫ltimo intervalo, que √© fechado.

```python
# Criando categorias de idade com pd.cut()

df_receitas["cliente_faixa_etaria"] = pd.cut(
    df_receitas["cliente_idade"],
    bins=[0, 17, 35, 60, 100],  # Limites das faixas et√°rias
    labels=["0-17", "18-35", "36-60", "61-100"]
)
```

üîπ Existem muitos caminhos. Poder√≠amos ter usado o np.select novamente para isso

```python
# Criando condi√ß√µes para as faixas et√°rias
condicoes = [
    df_receitas["cliente_idade"] < 18,
    (df_receitas["cliente_idade"] >= 18) & (df_receitas["cliente_idade"] <= 35),
    (df_receitas["cliente_idade"] > 35) & (df_receitas["cliente_idade"] <= 60),
    df_receitas["cliente_idade"] > 60
]

# üìå Definindo os r√≥tulos correspondentes
categorias = ["0-17", "18-35", "36-60", "61-100"]

# üìå Criando a nova coluna com np.select()
df_receitas["cliente_faixa_etaria"] = np.select(condicoes, categorias, default="Desconhecido")
```

```python
df_receitas["cliente_faixa_etaria"].value_counts(dropna=False)
```

Isso √© o que chamamos de **feature engeniier ou engenharia de features**, uma habilidade especialmente importante

### üí° Aplicando fun√ß√µes

No entanto, pode ser que a cria√ß√£o de features necessite de algum c√°lculo mais complexo. Um m√©todo muito usado na manipula√ß√£o de dados √© aplicar uma fun√ß√£o nos dados ou em parte deles, para isso o m√©todo <font color = "orange">**`.apply()`</font>** √© muito utilizado.

Importante dizer que esse m√©todo **aplica uma determinada fun√ß√£o em apenas uma coluna ou linha do DF.**

```python
# Fun√ß√£o para categorizar a idade
def classificar_faixa_etaria(idade):
    if idade < 18:
        return "0-17"
    elif idade <= 35:
        return "18-35"
    elif idade <= 60:
        return "36-60"
    elif idade > 60:
        return "61-100"
    else:
        return "Desconhecido"
```

```python
# Aplicando a fun√ß√£o com apply()
df_receitas["cliente_faixa_etaria"] = df_receitas["cliente_idade"].apply(classificar_faixa_etaria)
```

| üõ†Ô∏è **Engenharia de features** | ‚úÖ Ideal para regras simples de classifica√ß√£o            | ‚úÖ Ideal para regras complexas e condicionais encadeadas |
| üîç **Depura√ß√£o e testes**      | ‚ùå Mais dif√≠cil de testar e isolar erros                 | ‚úÖ Fun√ß√£o independente facilita testes                   |

A primeira fun√ß√£o a ser trabalhada para jun√ß√µes chama-se <font color = 'orange'> **`.merge()`</font>** e tem por objetivo cruzar informa√ß√µes de dois DataFrames relacionados, unindo suas informa√ß√µes em um √∫nico DataFrame, possibilitando o acesso √†s informa√ß√µes de maneira pr√°tica.

##### üö© **INNER JOIN**

A partir desses 2 DataFrames, podemos notar que existe uma coluna em comum - a coluna 'chave' - que relaciona os dois.
A nossa inten√ß√£o √© unir as informa√ß√µes de cada DF cuja 'chave' tenha o mesmo valor. Para isso, usaremos o <font color = 'orange'> inner join</font>.

>
Lembra de conjuntos? Esse comando √© como a **intersec√ß√£o** de dois conjuntos - apenas as informa√ß√µes que pertecem aos dois!

```python
pd.merge(esquerda, direita, on='chave', how='inner')

#Os √∫nicos valores comuns nas chaves s√£o 2, 3 e 4. Portanto, nosso merge s√≥ ter√° essas linhas.
```

##### üö© **LEFT OUTER JOIN**

A diferen√ßa dessa fun√ß√£o para a anterior √© que agora o programa vai pegar e retornar todas as chaves do DataFrame da esquerda, j√° no da direita, s√≥ vai retornar os valores que existirem no da esquera. Caso n√£o exista, retornar√° com `NaN`.

```python
pd.merge(esquerda, direita, on='chave', how='left')
```

##### üö© **RIGHT OUTER JOIN**

An√°logo ao `merge` anterior, a ideia desse √© pegar as chaves do DF da direita, verificar quais chaves existem na esquerda e, assim, devolver os respectivos valores.

```python
pd.merge(esquerda, direita, on='chave', how='right')
```

***
<font color = 'grey'> Talvez voc√™ seja ligeiro(a) e j√° tenha entendido esse neg√≥cio de left e right, mas talvez n√£o. E como n√≥s somos muuuuuuuuito did√°ticos, vamos explicar:
  
<font color = 'grey'> **Como que o programa sabe quem √© `left` e quem √© `right`, ali como par√¢metro do `how`?**
  
 <font color = 'grey'> √â bem √≥bvio, mesmo: o DataFrame referenciado primeiro, dentro de `pd.merge`, est√° √† esquerda do DF refer√™nciado em seguida. Ou seja, o `how='left'` est√° se referindo ao primeiro DF (Por isso, no nosso exemplo, chamamos o DataFrame de 'esquerda')
   
<font color = 'grey'>Perceba que se trocarmos a ordem e colocarmos o DataFrame 'direita' √† esquerda do DataFrame 'esquerda', teremos exatamente o contr√°rio e, portanto, ao usar `how='left'` na verdade ele estar√° referenciando o DataFrame 'direita'
   
   ****
  

```python
pd.merge(direita, esquerda, on='chave', how='right')
```

##### **üö© FULL OUTER JOIN**

Por fim, essa √∫ltima forma de <font color = 'orange'> `join` </font>basicamente pega todas as chaves existentes em um **ou** outro DataFrame e, com isso, constr√≥i um terceiro com todas elas, havendo conex√£o entre os DFs ou n√£o.
>
Esse aqui pode ser comparado com a **uni√£o** de dois conjuntos!


```python
pd.merge(esquerda, direita, on='chave', how='outer')
```

```python
pd.concat([esquerda, direita])
```

### üí° Agrega√ß√µes

üîπ Podemos utilizar a fun√ß√£o groupby, muito conhecida do SQL. Passe as colunas pelas quais quer agregar (como dimens√£o). As colunas que quer agregar (como fato) e a forma de agrega√ß√£o

#### üßê Teoria

Presta aten√ß√£o aqui! Para usar essa fun√ß√£o, precisamos entender algumas coisinhas, pois o groupby() funciona em tr√™s etapas ‚Äî pense como um processo de f√°brica

| Etapa        | O que acontece                                                   | Exemplo                              |
| ------------ | ---------------------------------------------------------------- | ------------------------------------ |
| **Dividir**  | Divide os dados em grupos, com base em uma ou mais colunas       | Agrupar por sexo, curso, cidade...   |
| **Aplicar**  | Aplica uma fun√ß√£o em cada grupo (ex: m√©dia, soma, contagem, etc) | M√©dia de notas, soma de vendas, etc. |
| **Combinar** | Junta os resultados em uma nova tabela, organizada por grupo     | Mostra os resultados lado a lado     |


```python
import pandas as pd

# Criando a tabela (DataFrame)
dados = pd.DataFrame({
    'Aluno': ['Ana', 'Bruno', 'Carla', 'Daniel'],
    'Sexo': ['F', 'M', 'F', 'M'],
    'Turma':['Turma A', 'Turma A', 'Turma B', 'Turma B'],
    'Nota': [8.5, 7.0, 9.0, 6.5]
})
```

```python
dados.groupby('Sexo')['Nota'].mean()
```

Explica√ß√£o:
* dados.groupby('Sexo')   # Agrupa as linhas por valores iguais na coluna 'Sexo'
* ['Nota']                # Escolhe a coluna num√©rica onde ser√° feita a opera√ß√£o
* .mean()                 # Aplica a m√©dia a cada grupo

**Se voc√™ quiser agrupar por mais de uma coisa:**

```python
dados.groupby(['Sexo', 'Turma'])['Nota'].mean()
```

**Se voc√™ quiser calcular mais de uma medida:**

```python
dados.groupby("Sexo")["Nota"].agg(["mean", "min"])
```

```python
dados.groupby("Sexo")["Nota"].aggregate(["mean", "min"])
```

No Pandas, .agg() e .aggregate() s√£o exatamente a mesma coisa. A diferen√ßa est√° somente no nome ‚Äî ambas as fun√ß√µes fazem a mesma opera√ß√£o.

**Se voc√™ quiser agregar calculando uma fun√ß√£o tamb√©m √© poss√≠vel:**

üë©‚Äçüè´ √Äs vezes, queremos fazer mais do que s√≥ calcular uma m√©dia, soma ou contagem. A fun√ß√£o .apply() permite aplicar qualquer l√≥gica personalizada em cada grupo!

```python
def classificar_grupo(grupo):
    media = grupo["Nota"].mean()
    if media > 8:
        return "Excelente"
    elif media >= 6:
        return "Regular"
    else:
        return "Ruim"

# Aplicando por grupo de Sexo
dados.groupby("Sexo").apply(classificar_grupo)
```

### üí° Stack & Unstack

Acabamos de aprender uma forma de organizar o DataFrame em rela√ß√£o ao seu √≠ndice, agora vamos dar uma olhadinha em outra maneira de modelar nosso `DataFrame`.

Os m√©todos <font color = "orange">**`.stack()`</font>** e <font color = "orange">**`.unstack()`</font>** podem ser usados tanto em `Series` quanto em `DataFrames` e s√£o usados juntamente com o conceito de `Multi-index` que acabamos de ver.

#### `Stack`

Em portugu√™s, "empilhar", √© um m√©todo que, basicamente, empilha os dados do DataFrame em uma ou mais linhas. Ou seja, move colunas para o √≠ndice (linha) ‚Üí deixa o DataFrame mais longo e estreito

```python
#Primeiro vamos criar um DataFrame

colunas = pd.MultiIndex.from_product([['2017', '2018'],['Vendas', 'Clientes']])
df_exemplo=([[12, 45, 67, 56],[78, 89, 45, 67],[45, 67, 89, 90],[67, 44, 56, 55]])


df_exemplo = pd.DataFrame(df_exemplo,
                  index=['Ana', 'Bruno', 'Cristina', 'Jackson'],
                  columns=colunas)
df_exemplo
```

Aqui temos uma estrutura de colunas com dois n√≠veis:
* N√≠vel 0: ano (2017, 2018)
* N√≠vel 1: tipo de dado (Clientes, Vendas)

Vamos ent√£o transformar as colunas do n√≠vel 0 em linhas:

```python
empilhado = df_exemplo.stack(0)
empilhado
```

E agora vamos ent√£o transformar as colunas do n√≠vel 1 em linhas:

```python
empilhado = df_exemplo.stack(1)
empilhado
```

Se n√£o passar o par√¢metro, ser√° adotado o n√≠vel interno (maior)

```python
empilhado = df_exemplo.stack()
empilhado
```

#### `Unstack`

√â o m√©todo inverso ao `stack`, ou seja, ele pega os √≠ndices interiores de linhas e transforma em √≠ndices interiores de colunas. Ou seja, move linhas para colunas ‚Üí deixa o DataFrame mais curto e longo.

```python
empilhado
```

Aqui temos uma estrutura de √≠ndices (linhas) com dois n√≠veis:
* N√≠vel 0: nomes
* N√≠vel 1: tipo de dado (Clientes, Vendas)

Vamos ent√£o transformar as linhas do n√≠vel 0 em linhas:

```python
desempilhado = empilhado.unstack(0)
desempilhado
```

E agora vamos ent√£o transformar os √≠ndices do n√≠vel 1 em linhas:

```python
desempilhado = empilhado.unstack(1)
desempilhado
```

Se n√£o passar o par√¢metro, ser√° adotado o n√≠vel interno (maior)

```python
desempilhado = empilhado.unstack()
desempilhado
```

Perceba que o "desempilhado" √© o mesmo DataFrame que o original "df". Portanto, o `unstack` desfez o que o `stack` tinha feito, por isso esses dois m√©todos s√£o como duas opera√ß√µes inversas.

Basicamente, usamos o `stack` quando queremos os dados agrupados em menos colunas e mais linhas, e, caso contr√°rio, usamos o `unstack`.

#### üßê Teoria

Em planilhas, utilizamos as boas e velhas Tabelas Din√¢micas, com um suporte do mouse

Basicamente, uma `pivot table` √© uma tabela com dados retirados de um banco de dados maior, que foram agrupados de maneira significativa para a busca de informa√ß√µes.
Ela nos permite organizar e reorganizar os dados para podermos visualiz√°-los mais facilmente, e tamb√©m, resumir os dados de forma a retirar as informa√ß√µes realmente importantes e √∫teis.


Vamos a um exemplo:

```python
sales = [{'conta': 'Jota LLC', 'gerente': 'Artur Nascimento', 'cliente': 'Diego Lima' ,'saldo': 150, 'movimenta√ß√µes': 1, 'tipo': 'C'},
         {'conta': 'Alfa Co','gerente': 'Artur Nascimento','cliente': 'C√©sar Pereira', 'saldo': 200, 'movimenta√ß√µes': 4, 'tipo': 'A'},
         {'conta': 'Novo Inc','gerente': 'D√©bora Souza', 'cliente': 'Alana Madeiros', 'saldo': 260, 'movimenta√ß√µes': 2, 'tipo': 'C' },
         {'conta': 'Delta A','gerente': 'Artur Nascimento', 'cliente': 'Jos√© Thompson', 'saldo': 50, 'movimenta√ß√µes': 2, 'tipo': 'B' },
         {'conta': 'Lib Inc','gerente': 'D√©bora Souza', 'cliente': 'Lara Nandi', 'saldo': 195, 'movimenta√ß√µes': 3, 'tipo': 'A'},
         {'conta': 'Alfa Co','gerente': 'Artur Nascimento','cliente': 'C√©sar Pereira', 'saldo': 105, 'movimenta√ß√µes': 2, 'tipo': 'A'},
         {'conta': 'Novo Inc','gerente': 'D√©bora Souza', 'cliente': 'Alana Madeiros', 'saldo': 130, 'movimenta√ß√µes': 1, 'tipo': 'C' }]
sales = pd.DataFrame(sales)
```

```python
sales
```

Algumas funcionalidades do `pivot table` n√≥s j√° vimos anteriormente com o m√©todo `groupby()`, por√©m veremos como esse novo m√©todo oferece algumas facilidades no conceito de multi-indexing.

Tudo o que precisamos para come√ßar a us√°-lo √© um `DataFrame` e um `index`:

```python
#Vamos come√ßar definindo a "conta" como nosso index e os valores como o saldo

sales_pivot = sales.pivot_table(
    index="gerente",
    values='saldo'
)

sales_pivot
```

Uma dica para usar esse m√©todo, √© come√ßar com poucos argumentos e testar cada vez, assim, a medida que aumentar, verificar o resultado, at√© chegar na informa√ß√£o que voc√™ precisa.

Usando o conceito de `multi-index`, podemos t√°mbem usar v√°rios √≠ndices no `pivot table`:

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values='saldo'
)

sales_pivot
```

Aqui, podemos ver que como n√£o indicamos o que queremos fazer com os dados, o m√©todo simplismente faz uma m√©dia dos valores das colunas, mas e se quisermos uma realizar uma opera√ß√£o diferente para as colunas?
Lembra da fun√ß√£o `agg`? Us√°vamos com o groupby para realizar alguma opera√ß√£o nos dados agrupados. Aqui podemos usar o m√©todo `aggfunc`


```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values='saldo',
    aggfunc="sum",
)

sales_pivot
```

E se quisermos mais de uma fun√ß√£o? Temos!

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values='saldo',
    aggfunc=['sum', 'mean']
)

sales_pivot
```

Ainda, se quisermos ver colunas diferentes e aplicar fun√ß√µes diferentes em cada uma? Temos!

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values=['movimenta√ß√µes', 'saldo'],
    aggfunc={'movimenta√ß√µes': 'sum','saldo': ['sum','max']}
)

sales_pivot
```

Al√©m disso, lembra do unstack? Aqui tamb√©m podemos "desempilhar" os dados, por exemplo, podemos selecionar dados que estavam em formato coluna e transform√°-los em linhas, com o m√©todo columns:

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values=['movimenta√ß√µes', 'saldo'],
    columns=['tipo'],
    aggfunc={'movimenta√ß√µes': 'sum','saldo': ['sum','max']}
)

sales_pivot
```

Ficou meio ruim de visualizar com todos esses `NaNs` n√©? Vamos substitui-los com o m√©todo `fill_value`

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values=['movimenta√ß√µes', 'saldo'],
    columns=['tipo'],
    aggfunc={'movimenta√ß√µes': 'sum','saldo': ['sum','max']},
    fill_value=0
)

sales_pivot
```

O que queremos mostrar √© que o m√©todo `pivot table` √© muito √∫til e vers√°til para visualizarmos os dados da maneira em que desejamos! Podemos brincar com ele e descobrir diversas maneiras eficientes para entendermos melhor os dados que temos.

Por exemplo, podemos mover outros itens para o `index` para termos diferentes modos de visualiza√ß√£o. Como remover o 'tipo' de `columns` e mov√™-lo para o `index`

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta', 'tipo'],
    values=['movimenta√ß√µes', 'saldo'],
    aggfunc={'movimenta√ß√µes': 'sum','saldo': ['sum','max']},
    fill_value=0
)

sales_pivot
```

E se quisermos ver um total de tudo isso? D√° uma olhadinha nesse m√©todo: `margins=True`, ele nos mostra a soma das colunas de valores:


```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta', 'tipo'],
    values=['movimenta√ß√µes', 'saldo'],
    aggfunc='sum',
    fill_value=0,
    margins=True
)
sales_pivot
```

E mais, podemos dar um nome para essa nova linha com `margins_name`:

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta', 'tipo'],
    values=['movimenta√ß√µes', 'saldo'],
    aggfunc='sum',
    fill_value=0,
    margins=True,
    margins_name='Total'
)
sales_pivot
```

**Compara√ß√£o Pivot Table e Groupby**

Conseguiu ver a diferen√ßa entre os m√©todos? O importante √© saber que podemos obter informa√ß√µes de v√°rias formas diferentes, por isso, para usar o Pandas √© essencial saber claramente o que voc√™ est√° procurando, assim fica mais f√°cil escolher o m√©todo mais adequado.

Para resumir o que acabamos de discutir:

| Caracter√≠stica                        | `pivot_table` / `stack` / `unstack`                         | `groupby`                                      |
| ------------------------------------- | ----------------------------------------------------------- | ---------------------------------------------- |
| üß† **Objetivo**                       | Reorganizar, modelar e transformar a visualiza√ß√£o dos dados | Resumir os dados por grupos                    |
| üì¶ **Ret√©m todos os dados?**          | Sim. Os dados n√£o s√£o perdidos, apenas reorganizados        | N√£o. Retorna apenas os dados agregados         |
| üìê **Formato de sa√≠da**               | DataFrame reorganizado, geralmente com multi-√≠ndices        | DataFrame ou Series com menos informa√ß√µes      |
| üîß **Funcionalidade principal**       | Mudan√ßa de forma: colunas viram linhas, ou vice-versa       | C√°lculo de agrega√ß√µes, estat√≠sticas por grupo  |
| üîÅ **Reversibilidade**                | `stack/unstack` s√£o revers√≠veis (`df.stack().unstack()`)    | N√£o revers√≠vel diretamente                     |
| üí° **Uso t√≠pico**                     | Tabelas din√¢micas, dashboards, visualiza√ß√µes tabulares      | Sum√°rios estat√≠sticos, an√°lise de agrupamentos |
| üßÆ **Agrega√ß√µes simult√¢neas**         | Sim, com `pivot_table(aggfunc={...})`                       | Sim, mas separadamente ou com `.agg({...})`    |
| üß± **Facilidade de m√∫ltiplos n√≠veis** | Muito boa (MultiIndex nativo)                               | Tamb√©m boa, mas mais manual                    |


üéØ Quando usar cada um?
* Use groupby quando quiser resumos estat√≠sticos simples.
* Use pivot_table para an√°lises multidimensionais, como em tabelas din√¢micas.
* Use stack e unstack para transformar a estrutura de linhas/colunas ‚Äî √≥timo para visualiza√ß√µes e pr√©-processamento.

```python
- index: Coluna que vai para o eixo das linhas (√≠ndice da tabela)
- columns: Coluna que ser√° transformada em colunas
- values: Coluna cujos valores ser√£o agregados
- aggfunc: Fun√ß√£o de agrega√ß√£o (ex: 'mean', 'sum', 'count', etc.)
```
