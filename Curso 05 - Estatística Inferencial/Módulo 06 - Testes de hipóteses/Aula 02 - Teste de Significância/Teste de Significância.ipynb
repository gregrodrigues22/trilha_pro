 O n√≠vel de signific√¢ncia √© um valor arbitr√°rio (ou seja, escolhido por n√≥s) que definimos antes do estudo. A grosso modo, √© a probabilidade de voc√™ dizer que h√° um efeito quando n√£o h√° efeito.

N√≠veis de signific√¢ncia mais baixos indicam que voc√™ precisa de evid√™ncias mais substanciais antes de rejeitar o nulo ‚Äî ou seja, quanto menor o alpha, mais dificilmente voc√™ rejeita a hip√≥tese nula.

üìå Exemplo: um n√≠vel de signific√¢ncia de 0,05 significa um risco de 5% de decidir que existe um efeito quando n√£o existe.



#üéØ  Como entender isso na pr√°tica

#üìà Visualizando o n√≠vel de signific√¢ncia

üìâ O n√≠vel de signific√¢ncia define o qu√£o longe o valor da amostra deve estar para que possamos rejeitar a hip√≥tese nula.

A √°rea sombreada sob a curva representa o quanto estamos dispostos a aceitar que um resultado raro (se a hip√≥tese nula for verdadeira) ainda assim aconte√ßa por acaso.


No exemplo acima, temos um alpha de 0,1, ent√£o 10% da curva est√° sombreada (5% de cada lado).

Isso define a regi√£o de rejei√ß√£o da hip√≥tese nula.

Se nosso valor amostral cair ali, dizemos que est√° estatisticamente longe demais para ser compat√≠vel com a hip√≥tese nula.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√¢metros
media_hipotese_nula = 0
desvio_padrao = 1
alpha = 0.1  # n√≠vel de signific√¢ncia total (5% para cada lado)
z_critico = norm.ppf(1 - alpha / 2)  # valor cr√≠tico para regi√£o de rejei√ß√£o bilateral

# Gera√ß√£o dos dados da curva normal
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, media_hipotese_nula, desvio_padrao)

# Plot da curva normal
plt.plot(x, y, color='black')

# Preenchendo as regi√µes cr√≠ticas
x_rejeita_esquerda = np.linspace(-4, -z_critico, 100)
y_rejeita_esquerda = norm.pdf(x_rejeita_esquerda, media_hipotese_nula, desvio_padrao)
plt.fill_between(x_rejeita_esquerda, y_rejeita_esquerda, color='blue', alpha=0.4)

x_rejeita_direita = np.linspace(z_critico, 4, 100)
y_rejeita_direita = norm.pdf(x_rejeita_direita, media_hipotese_nula, desvio_padrao)
plt.fill_between(x_rejeita_direita, y_rejeita_direita, color='blue', alpha=0.4)

# Anota√ß√µes
plt.text(-z_critico, 0.02, '5%', ha='center')
plt.text(z_critico, 0.02, '5%', ha='center')
plt.text(0, 0.42, '90%', ha='center')
plt.axvline(x=-z_critico, color='black', linestyle='dashed')
plt.axvline(x=z_critico, color='black', linestyle='dashed')

# T√≠tulos e r√≥tulos
plt.title('Visualiza√ß√£o do N√≠vel de Signific√¢ncia (Œ± = 0,1)')
plt.xlabel('Estat√≠stica de Teste (z)')
plt.ylabel('Densidade de Probabilidade')

# Exibir gr√°fico
plt.grid(True)
plt.show()
```

Aqui est√° o gr√°fico que representa visualmente o n√≠vel de signific√¢ncia Œ± = 0,1 em um teste bicaudal:

* As √°reas azuis representam as regi√µes cr√≠ticas de rejei√ß√£o da hip√≥tese nula (H‚ÇÄ).
* Cada regi√£o cr√≠tica tem 5% da √°rea total da distribui√ß√£o (Œ±/2).
* A √°rea central (entre os dois limites cr√≠ticos) representa os 90% de confian√ßa, onde n√£o rejeitamos H‚ÇÄ.

Se a estat√≠stica de teste (z) calculada cair em qualquer uma das √°reas sombreadas, rejeitamos H‚ÇÄ com 10% de signific√¢ncia. Caso contr√°rio, n√£o rejeitamos H‚ÇÄ. Isso √© usado, por exemplo, para verificar se uma m√©dia observada est√° "longe o suficiente" da m√©dia sob a hip√≥tese nula. ‚Äã
