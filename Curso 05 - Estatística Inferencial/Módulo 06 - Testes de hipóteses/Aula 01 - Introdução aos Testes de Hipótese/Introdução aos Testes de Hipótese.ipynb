#‚úÖ Exemplo

Imagine que um pesquisador est√° estudando os n√≠veis m√©dios de colesterol LDL em uma popula√ß√£o de pacientes adultos. No ano passado, a m√©dia populacional registrada era de 130 mg/dL. Neste ano, com o objetivo de verificar se houve alguma mudan√ßa, o pesquisador seleciona aleatoriamente 25 pacientes para medir seus n√≠veis atuais de colesterol LDL.

**Dados da amostra**

O pesquisador encontrou uma m√©dia amostral de 165 mg/dL. No entanto, √© poss√≠vel que, devido ao erro amostral, a m√©dia real da popula√ß√£o ainda seja 130 mg/dL (mesmo com a diferen√ßa aparente). Como o desvio padr√£o √© alto, outras amostras poderiam gerar m√©dias diferentes. Portanto, n√£o √© poss√≠vel afirmar com certeza a mudan√ßa apenas com base em uma √∫nica m√©dia.

Logo, o pesquisador decide fazer um teste de hip√≥tese. Para isso ele tra√ßa as 2 hip√≥teses necess√°rias

‚ú≥Ô∏è Formula√ß√£o das Hip√≥teses
* Hip√≥tese nula (H‚ÇÄ): A m√©dia do colesterol LDL na popula√ß√£o permanece igual a 130 mg/dL.
* Hip√≥tese alternativa (H‚ÇÅ): A m√©dia n√£o √© igual a 130 mg/dL.

Ou seja:

H‚ÇÄ: Œº = 130

H‚ÇÅ: Œº ‚â† 130

Para responder a essa pergunta, simulamos a distribui√ß√£o amostral com base na suposi√ß√£o de que a m√©dia populacional n√£o mudou e continua sendo 130 mg/dL.

O grande ponto aqui √©: **todos os testes de hip√≥tese come√ßam assumindo que a hip√≥tese nula est√° correta**. N√£o sabemos se de fato ela est√°, mas precisamos de um ponto de partida ‚Äî e ele √©: suponha que H‚ÇÄ est√° certa.



A partir da suposi√ß√£o de que a m√©dia continua sendo 130 mg/dL, podemos gerar a distribui√ß√£o de todas as m√©dias poss√≠veis (caso muitas amostras de 25 pacientes fossem coletadas). Essa distribui√ß√£o nos permite ver o qu√£o improv√°vel seria obter uma m√©dia amostral como 165 mg/dL se a m√©dia real da popula√ß√£o ainda fosse 130 mg/dL.

Imagine uma curva normal centrada em 130 mg/dL. Podemos colocar a m√©dia observada (165 mg/dL) nesse gr√°fico. Se ela estiver muito distante do centro da curva, indicar√° que tal m√©dia seria muito improv√°vel ‚Äî sugerindo que H‚ÇÄ pode estar errada.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Dados do problema
media_populacional = 130  # mg/dL
media_amostral = 165  # mg/dL
tamanho_amostra = 25
desvio_padrao_populacional = 20 # exemplo

# Calculando o desvio padr√£o da m√©dia amostral
desvio_padrao_media_amostral = desvio_padrao_populacional / np.sqrt(tamanho_amostra)

# Criando a curva normal
x = np.linspace(media_populacional - 4 * desvio_padrao_media_amostral, media_populacional + 4 * desvio_padrao_media_amostral, 100)
y = norm.pdf(x, loc=media_populacional, scale=desvio_padrao_media_amostral)

# Plotando a curva normal
plt.plot(x, y)

# Marcando a m√©dia amostral no gr√°fico
plt.axvline(x=media_amostral, color='red', linestyle='--', label='M√©dia Amostral (165 mg/dL)')

# Configurando o gr√°fico
plt.title('Distribui√ß√£o Amostral da M√©dia de Colesterol LDL')
plt.xlabel('N√≠veis de Colesterol LDL (mg/dL)')
plt.ylabel('Probabilidade')
plt.legend()
plt.grid(True)

# Exibindo o gr√°fico
plt.show()

# Calculando o p-valor (exemplo)
z_score = (media_amostral - media_populacional) / desvio_padrao_media_amostral
p_valor = 2 * (1 - norm.cdf(abs(z_score))) # teste bilateral

print(f"z-score: {z_score}")
print(f"p-valor: {p_valor}")
```

Se a nossa m√©dia amostral est√° em uma regi√£o extremamente improv√°vel da curva, dizemos que temos evid√™ncia estat√≠stica para rejeitar a hip√≥tese nula. Isso significa que a diferen√ßa entre os valores (165 vs. 130) √© estatisticamente significativa.

No entanto, n√£o existe um ponto exato na curva que indique quando a hip√≥tese nula deve ser rejeitada. O que h√° √© uma regi√£o de cauda da distribui√ß√£o que nos permite calcular a probabilidade de observar um valor t√£o extremo quanto 165 se H‚ÇÄ for verdadeira. Essa probabilidade √© chamada de p-valor ‚Äî e se for pequena o suficiente (menor que 0,05, por exemplo), rejeitamos H‚ÇÄ.

üîç Conclus√£o:
A hip√≥tese nula funciona como a teoria padr√£o: partimos do pressuposto de que nada mudou. A partir da evid√™ncia da amostra (neste caso, uma m√©dia de 165 mg/dL), utilizamos a estat√≠stica inferencial para decidir se devemos manter ou rejeitar essa hip√≥tese inicial. Esse processo nos ajuda a tomar decis√µes mais objetivas e baseadas em probabilidade ‚Äî o que √© essencial em an√°lises cient√≠ficas e cl√≠nicas na sa√∫de.

#üîç O que √© o Valor-P?

#‚ùå Erros comuns de interpreta√ß√£o (Misconceptions)



> "O valor-p √© provavelmente o √≠ndice mais onipresente e, ao mesmo tempo, o mais mal compreendido, mal interpretado e ocasionalmente mal calculado em toda a pesquisa biom√©dica." Garcia-Berthou E, Alcaraz C. Incongruence between test statistics and P values in medical papers. BMC Medical Research Methodology, volume 4, p√°gina 13, ano 2004.



**Se o p-valor √© pequeno, ent√£o o resultado n√£o √© devido ao acaso. Mas o p-valor n√£o mede diretamente: Se o acaso foi a √∫nica explica√ß√£o. Se outra hip√≥tese alternativa √© verdadeira. **P <= 0,05 significa que H‚ÇÄ tem s√≥ 5% de chance de ser verdadeira**: Falso.  Esta √© uma das concep√ß√µes err√¥neas mais difundidas. O valor P √© calculado assumindo que a hip√≥tese nula √© verdadeira e, portanto, n√£o pode ser a probabilidade de a hip√≥tese nula ser falsa.

**P > 0,05 prova que n√£o h√° diferen√ßa**: 	Falso. Apenas indica que n√£o h√° evid√™ncia suficiente para rejeitar H‚ÇÄ. Na verdade, quer dizer apenas que voc√™ n√£o encontrou evid√™ncia suficiente para afirmar que existe uma diferen√ßa. **Estudos com p <= 0,05 e p > 0,05 s√£o conflitantes**: Falso. Um estudo pode ter p < 0,05 e o outro p > 0,05 mesmo que o efeito observado (a diferen√ßa entre tratamentos) seja praticamente o mesmo.
