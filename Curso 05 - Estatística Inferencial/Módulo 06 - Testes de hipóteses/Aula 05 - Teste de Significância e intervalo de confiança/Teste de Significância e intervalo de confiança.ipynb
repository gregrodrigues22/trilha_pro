* O teste de hip√≥tese mede a probabilidade de observar os dados se H0 fosse verdadeira.

#üéØ Rela√ß√£o entre Intervalo de Confian√ßa e Testes de Signific√¢ncia

* Se o intervalo de confian√ßa de 95% N√ÉO cont√©m o valor de interesse (ex: 0 para diferen√ßa de m√©dias, 1 para risco relativo) ‚Üí o teste ser√° estatisticamente significativo (valor-p < 0,05).
* Se o intervalo de confian√ßa de 95% CONT√âM o valor de interesse ‚Üí o teste ser√° n√£o significativo (valor-p > 0,05).

Ou seja, valor-p e intervalo de confian√ßa chegam √† mesma conclus√£o:
* Ambos dizem se rejeitamos ou n√£o H0.
* S√≥ s√£o formas diferentes de mostrar a mesma hist√≥ria.

```python
# Novo gr√°fico baseado no seu pedido:
# Mostrar valor-p como uma faixa, com estimativa pontual e intervalo de confian√ßa

# Configura√ß√µes
p_value = 0.03  # valor-p hipot√©tico (significativo)
estimate = 2.5  # estimativa pontual
ci_lower = 1.2  # limite inferior do IC
ci_upper = 3.8  # limite superior do IC

# Criar o gr√°fico
plt.figure(figsize=(10, 3))

# Desenhar a linha do intervalo de confian√ßa
plt.hlines(1, ci_lower, ci_upper, color='black', linewidth=2, label='Intervalo de Confian√ßa 95%')

# Marcar a estimativa pontual
plt.plot(estimate, 1, 'o', color='red', label='Estimativa Pontual')

# Adicionar faixa ilustrando o p-valor
plt.fill_betweenx([0.9, 1.1], ci_lower, ci_upper, color='green', alpha=0.3, label=f'√Årea Relacionada ao p-valor (p={p_value})')

# Ajustes visuais
plt.axvline(x=0, color='gray', linestyle='--', label='Valor de Refer√™ncia (0)')
plt.yticks([])
plt.xlabel('Valor da Estat√≠stica')
plt.title('Visualizando Estimativa, IC e √Årea de p-valor')
plt.legend()
plt.grid(True)
plt.show()

```

* Linha preta: √© o Intervalo de Confian√ßa (IC 95%) ‚Äî vai de 1.2 a 3.8.
* Ponto vermelho: √© a Estimativa Pontual (neste exemplo, 2.5).
* √Årea verde clara: representa a regi√£o associada ao p-valor.
* Linha tracejada no 0: √© o valor de refer√™ncia da hip√≥tese nula H0 (sem efeito).

üìä Interpreta√ß√£o:
* Como o intervalo n√£o inclui o 0, rejeitamos H0.
* A estimativa pontual est√° centralizada dentro do intervalo de confian√ßa.
* O p-valor de 0,03 confirma essa rejei√ß√£o para um n√≠vel de ùõº = 0,05

#üìö Detalhe importante

* Para diferen√ßas absolutas (ex: m√©dias, propor√ß√µes): o valor cr√≠tico √© 0.

* Para indicadores relativos (ex: risco relativo, odds ratio): o valor cr√≠tico √© 1.

```python
import matplotlib.pyplot as plt

# Criar uma figura com dois gr√°ficos lado a lado
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# Gr√°fico 1: Diferen√ßa de m√©dias (analisando 0)
axs[0].hlines(0, -5, 15, colors='black', linestyles='--', label='Zero (Sem diferen√ßa)')
axs[0].hlines(1, 2, 11, colors='blue', linewidth=6, label='Intervalo de Confian√ßa [2,11]')
axs[0].vlines(0, -1, 2, colors='red', linestyles=':', label='Linha no 0 (ponto cr√≠tico)')
axs[0].set_xlim(-5, 15)
axs[0].set_ylim(-1, 2)
axs[0].set_title('Diferen√ßa de M√©dias (Analisando 0)')
axs[0].set_xlabel('Diferen√ßa')
axs[0].set_yticks([])
axs[0].legend(loc='upper right')

# Gr√°fico 2: Risco Relativo (analisando 1)
axs[1].hlines(1, 0.5, 2.5, colors='black', linestyles='--', label='Um (Sem diferen√ßa)')
axs[1].hlines(1.2, 1.1, 2.0, colors='green', linewidth=6, label='Intervalo de Confian√ßa [1.1,2.0]')
axs[1].vlines(1, 0.8, 1.4, colors='red', linestyles=':', label='Linha no 1 (ponto cr√≠tico)')
axs[1].set_xlim(0.5, 2.5)
axs[1].set_ylim(0.8, 1.4)
axs[1].set_title('Risco Relativo (Analisando 1)')
axs[1].set_xlabel('Risco Relativo')
axs[1].set_yticks([])
axs[1].legend(loc='upper right')

plt.tight_layout()
plt.show()
```
