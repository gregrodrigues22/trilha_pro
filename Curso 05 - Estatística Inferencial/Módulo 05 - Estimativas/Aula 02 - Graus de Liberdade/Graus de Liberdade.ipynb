# üìä Exemplo em Python

Vamos calcular a vari√¢ncia de uma amostra com e sem ajuste de graus de liberdade.

```python
import numpy as np

# Exemplo: notas de 5 alunos
notas = np.array([7, 8, 6, 9, 10])

# M√©dia da amostra
media = np.mean(notas)

# Vari√¢ncia com n no denominador (vi√©sada)
variancia_vi√©sada = np.var(notas, ddof=0)

# Vari√¢ncia com n - 1 no denominador (n√£o vi√©sada)
variancia_corrigida = np.var(notas, ddof=1)

print(f"üéØ M√©dia: {media}")
print(f"‚ö†Ô∏è Vari√¢ncia com GL = 5 (vi√©sada): {variancia_vi√©sada:.2f}")
print(f"‚úÖ Vari√¢ncia com GL = 4 (corrigida): {variancia_corrigida:.2f}")
```
