#üîç O que √© um Intervalo de Confian√ßa?

Um intervalo de confian√ßa √© um intervalo de valores plaus√≠veis para o par√¢metro de interesse (como a m√©dia populacional), baseado em uma amostra.

Exemplo: "Com 95% de confian√ßa, a m√©dia do colesterol da popula√ß√£o est√° entre 180 e 200 mg/dL."

#üî¢ Como interpretar?

Um IC de 95% significa que, se repet√≠ssemos o estudo muitas vezes, cerca de 95% dos intervalos constru√≠dos conteriam o verdadeiro par√¢metro.

Ou seja, a m√©dia amostral √© uma estimativa imparcial da m√©dia populacional, ou seja, em m√©dia, ela n√£o tende nem a subestimar nem a superestimar. Por isso, os intervalos de confian√ßa, baseados na m√©dia amostral, s√£o corretos em m√©dia ‚Äì se o processo for repetido muitas vezes.

#‚ùå Erros comuns de interpreta√ß√£o (Misconceptions)

**Erro: Ignorar a margem de erro**

Um erro comum √© ignorar a margem de erro ou a largura do intervalo de confian√ßa (estimativa intervalar) e focar apenas na estimativa pontual ou no centro do intervalo. No entanto, a margem de erro reflete a precis√£o e a variabilidade da estimativa e √© afetada pelo tamanho da amostra, pelo n√≠vel de confian√ßa e pela variabilidade populacional. Um intervalo de confian√ßa estreito indica uma estimativa mais precisa, enquanto um intervalo de confian√ßa amplo indica uma estimativa mais incerta. Portanto, voc√™ deve sempre relatar e comparar os intervalos de confian√ßa com suas margens de erro, e n√£o apenas as estimativas pontuais.

eaea21212.png

**Erro: Confundir confian√ßa com probabilidade**

Um dos erros mais comuns √© confundir o n√≠vel de confian√ßa com a probabilidade de o par√¢metro populacional estar dentro do intervalo de confian√ßa. Por exemplo, se voc√™ tem um intervalo de confian√ßa de 95% para o peso m√©dio de uma popula√ß√£o, isso n√£o significa que h√° uma chance de 95% de que o peso m√©dio verdadeiro esteja dentro desse intervalo. Em vez disso, isso significa que, se voc√™ repetir o processo de amostragem muitas vezes, 95% dos intervalos de confian√ßa obtidos conter√£o o verdadeiro peso m√©dio. O n√≠vel de confian√ßa reflete a confiabilidade do m√©todo de amostragem, e n√£o a incerteza sobre o par√¢metro populacional.

**Erro: Comunicar mal os resultados obtidos com base nos intervalos de confian√ß**

Um erro comum √© interpretar ou comunicar mal os resultados com base nos intervalos de confian√ßa, como fazer afirma√ß√µes causais, generalizar al√©m do escopo dos dados ou usar linguagem vaga ou enganosa. Um bom analista deve ser claro, honesto e t√©cnico, evitando exageros. O IC √© uma ferramenta poderosa, mas deve ser interpretado com respeito √†s limita√ß√µes da amostra, suposi√ß√µes do modelo e objetivos da an√°lise.

Isso porque h√° incerteza inerente quando voc√™ usa dados de amostra para fazer infer√™ncias sobre uma popula√ß√£o inteira. Os intervalos de confian√ßa reconhecem essa incerteza, ao inv√©s de tentar ocult√°-la ‚Äî e isso √© uma das grandes for√ßas da estat√≠stica.

Se voc√™ tem um intervalo de confian√ßa n√£o deve usar termos como "garantido", "comprovado" ou "exato" para descrever o resultado, pois eles implicam certeza e ignoram a incerteza e a variabilidade da estimativa. Para evitar esse erro, voc√™ deve usar uma linguagem clara, precisa e cautelosa para comunicar os resultados com base nos intervalos de confian√ßa e reconhecer as limita√ß√µes e suposi√ß√µes da an√°lise. Exemplo:
* ‚ùå ‚ÄúEst√° comprovado que a m√©dia √© X.‚Äù
* ‚úÖ ‚ÄúEstimamos que a m√©dia esteja entre Y e Z, com 95% de confian√ßa.‚Äù
* ‚úÖ ‚ÄúA estimativa da m√©dia √© de X, com um intervalo de confian√ßa de 95% entre Y e Z.‚Äù
* ‚úÖ ‚ÄúCom base nessa amostra, podemos inferir com 95% de confian√ßa que...‚Äù

**Erro padr√£o**

Ou seja, a f√≥rmula do intervalo de confian√ßa usa o erro padr√£o para definir a faixa de valores plaus√≠veis para a m√©dia populacional. A Precis√£o da estimativa melhora com menor erro padr√£o, o que acontece se:
* O desvio padr√£o da popula√ß√£o for menor
* O tamanho da amostra for maior

**N√≠vel de confian√ßa**

Al√©m do erro padr√£o, o tamanho do intervalo de confian√ßa tamb√©m depende do valor cr√≠tico (ùëß ùõº/2), que est√° relacionado ao n√≠vel de confian√ßa que voc√™ escolhe. O n√∫mero ùëß ùõº/2 √© retirado da distribui√ß√£o normal padr√£o que representa quantos desvios padr√£o queremos considerar para "capturar" a m√©dia populacional com uma certa confian√ßa. Em outras palavras, o valor de nos diz quantos erros padr√£o para cada lado da m√©dia voc√™ precisa incluir para capturar a √°rea desejada sob a curva normal (a sua confian√ßa).

eaeaea211.png

Se voc√™ usa 95% de confian√ßa, o valor cr√≠tico ùëß ùõº/2 = 1.96, ou seja, voc√™ est√° dizendo: ‚ÄúQuero um intervalo que fique 1.96 desvios padr√£o para cima e para baixo da m√©dia ‚Äî isso cobre 95% das poss√≠veis amostras‚Äù.

üéØ Por que n√£o usar 100% de confian√ßa?

Porque o intervalo teria que ser infinitamente largo. Para garantir 100% de certeza de que o verdadeiro valor est√° inclu√≠do, o intervalo teria que conter todos os valores poss√≠veis ‚Äî e isso n√£o √© √∫til para an√°lises pr√°ticas. Exemplo do texto: Para garantir que a altura de toda mulher brasileira esteja dentro de um intervalo com 100% de confian√ßa, precisar√≠amos de um intervalo gigante ‚Äî o que n√£o ajuda na tomada de decis√£o.

‚ùó E se usarmos um n√≠vel muito baixo, como 5%?

Nesse caso, o intervalo ser√° estreito, mas a probabilidade de conter o par√¢metro verdadeiro ser√° muito baixa. Isso significa que n√£o podemos confiar nesse intervalo para representar bem a popula√ß√£o. Exemplo: Se o intervalo de 5% for de 1,62 a 1,64m para altura de mulheres, isso √© muito estreito e n√£o representar√° a maioria da popula√ß√£o, mesmo que pare√ßa preciso.



‚úÖ Ent√£o, qual o n√≠vel de confian√ßa ideal?

A pr√°tica mais comum √© usar 95% de confian√ßa, podendo variar entre 90% e 99%, dependendo da √°rea de estudo e do grau de precis√£o necess√°rio.



**Margem de Erro**

Assim, √© que temos a margem de erro. Ela representa o quanto estamos dispostos a "errar" para mais ou para menos em torno da m√©dia amostral, ao construir um intervalo de confian√ßa.

#üë®‚Äçüíª Demonstra√ß√£o do c√°lculo do Intervalo de Confian√ßa da M√©dia quando desvio padr√£o √© conhecido

Quando voc√™ calcula um intervalo de confian√ßa na m√©dia, voc√™ calcula a m√©dia de uma amostra para estimar a m√©dia da popula√ß√£o. √â claro que se voc√™ j√° soubesse a m√©dia da popula√ß√£o, n√£o haveria necessidade de um intervalo de confian√ßa. No entanto, para explicar como os intervalos de confian√ßa s√£o constru√≠dos, vamos trabalhar de tr√°s para frente e come√ßar assumindo caracter√≠sticas da popula√ß√£o. Ent√£o mostraremos como os dados da amostra podem ser usados ‚Äã‚Äãpara construir um intervalo de confian√ßa.

Suponha que os pesos de crian√ßas de 10 anos sejam normalmente distribu√≠dos com uma m√©dia de 90 e um desvio padr√£o de 36. Qual √© a distribui√ß√£o amostral da m√©dia para um tamanho de amostra de 9?

```python
from scipy.stats import norm
from numpy import linspace
import plotly.graph_objects as go
```

```python
mu = 90 #m√©dia da popula√ß√£o
sigma = 36 #desvio padr√£o da popula√ß√£o
N = 9 #tamanho da amostra
```

```python
X = linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
go.Figure(
    {
        'data': [
            {'x': X, 'y': norm(mu, sigma).pdf(X), 'name': 'population'},
            {'x': X, 'y': norm(mu, sigma / (N ** 0.5)).pdf(X), 'name': 'sample means (n=9)'}
        ]
    }
)
```

Lembre-se da aula sobre a distribui√ß√£o amostral da m√©dia que a m√©dia da distribui√ß√£o amostral √© $\mu$ e o erro padr√£o da m√©dia √©

$\sigma_M=\frac{\sigma}{\sqrt{N}}$

Para o exemplo presente, a distribui√ß√£o amostral da m√©dia tem uma m√©dia de 90 e um desvio padr√£o de 36/3 = 12.

```python
sigma / (N ** 0.5)
```

Note que o desvio padr√£o de uma distribui√ß√£o amostral √© seu erro padr√£o. A Figura abaixo mostra essa distribui√ß√£o. A √°rea sombreada representa os 95% do meio da distribui√ß√£o e se estende de 66,48 a 113,52.

```python
norm(mu, sigma / (N ** 0.5)).interval(0.95)
```

Esses limites foram calculados adicionando e subtraindo 1,96 desvios-padr√£o de/para a m√©dia de 90, como segue:

$90 - (1,96)(12) = 66,48$

$90 + (1,96)(12) = 113,52$

O valor de 1,96 √© baseado no fato de que 95% da √°rea de uma distribui√ß√£o normal est√° dentro de 1,96 desvios-padr√£o da m√©dia; 12 √© o erro-padr√£o da m√©dia.

```python
 z = norm().interval(0.95)[1]
 z
```

```python
lb = mu - z * sigma / (N ** 0.5)
ub = mu + z * sigma / (N ** 0.5)
lb, ub
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, t, uniform
import pandas as pd
```

```python
def area_between(mean, std, between):
    dist = norm(mean, std)
    fig, ax = plt.subplots(1, 1, figsize=(15, 5))
    x_axis = np.arange(mean - 5 * std, mean + 5 * std, 0.01)
    n = pd.Series(dist.pdf(x_axis), index=x_axis)
    _ = n.plot(ax=ax, legend=False)
    lb = mean - between * std
    ub = mean + between * std
    where = np.logical_and(x_axis > lb, x_axis < ub)
    _ = ax.fill_between(x_axis, n, 0, where=where, interpolate=True)
    area = dist.cdf(ub) - dist.cdf(lb)
    title = ('The shaded area accounts for {:.4f}% of total. '
             'Mean: {:.2f}; Std: {:.2f}').format(area * 100, mean, std)
    _ = plt.title(title)

area_between(0, 1, 1.96)
```

A Figura acima mostra que 95% das m√©dias n√£o est√£o a mais de 23,52 unidades (1,96 desvios padr√£o) da m√©dia de 90.

Agora considere a probabilidade de que uma m√©dia amostral calculada em uma amostra aleat√≥ria esteja dentro de 23,52 unidades da m√©dia populacional de 90. Como 95% da distribui√ß√£o est√° dentro de 23,52 de 90, a probabilidade de que a m√©dia de qualquer amostra dada esteja dentro de 23,52 de 90 √© 0,95.

Isso significa que se calcularmos repetidamente a m√©dia ($M$) de uma amostra e criarmos um intervalo variando de $M - 23,52$ a $M + 23,52$, esse intervalo conter√° a m√©dia populacional 95% do tempo. Em geral, voc√™ calcula o intervalo de confian√ßa de 95% para a m√©dia com a seguinte f√≥rmula:

Limite inferior = $M - Z_{.95}\sigma_M$

Limite superior = $M + Z_{.95}\sigma_M$

onde $Z_{.95}$ √© o n√∫mero de desvios-padr√£o que se estendem da m√©dia de uma distribui√ß√£o normal necess√°ria para conter 0,95 da √°rea e $\sigma_M$ √© o erro-padr√£o da m√©dia.

```python
mean = 90
std = 12
z = 1.96

dist = norm(90, 12)
lower_limit = mean - z * std
upper_limit = mean + z * std
area = dist.cdf(upper_limit) - dist.cdf(lower_limit)

print(f'{area*100:.2f}% da probabilidade est√° dentro do intervalo {lower_limit:.2f} to {upper_limit:.2f}.')
```

```python
mean = 90
std = 12
area = 0.95

dist = norm(mean, std)
lower_limit = dist.ppf((1 - area) / 2)
upper_limit = dist.ppf(area + (1 - area) / 2)
z = (upper_limit - mean) / std

msg = 'mean +/- {:.2f} stds (Z={:.2f}) define the range {:.2f} to {:.2f}.'
print(msg.format(z, z, lower_limit, upper_limit))
```

Se voc√™ observar atentamente esta f√≥rmula para um intervalo de confian√ßa, notar√° que **voc√™ precisa saber o desvio padr√£o da popula√ß√£o ($\sigma$) para estimar a m√©dia**. Isso pode parecer irreal, e √©. No entanto, calcular um intervalo de confian√ßa quando $\sigma$ √© conhecido √© mais f√°cil do que quando $\sigma$ tem que ser estimado, e serve a um prop√≥sito pedag√≥gico. Mais adiante nesta se√ß√£o, mostraremos como calcular um intervalo de confian√ßa para a m√©dia quando $\sigma$ tem que ser estimado.

Suponha que os cinco n√∫meros a seguir foram amostrados de uma distribui√ß√£o normal com um desvio padr√£o de 2,5: 2, 3, 5, 6 e 9. Para calcular o intervalo de confian√ßa de 95%, comece calculando a m√©dia e o erro padr√£o:

$M = \frac{(2 + 3 + 5 + 6 + 9)}{5} = 5$.

$\sigma_M = \frac{2,5}{\sqrt{5}} = 1,118$.

```python
ser = pd.Series([2, 3, 5, 6, 9])
m = ser.mean()
sigma_m = 2.5 / (m ** 0.5)
m, sigma_m
```

$Z_{.95}$ pode ser encontrado usando a calculadora de distribui√ß√£o normal e especificando que a √°rea sombreada √© 0,95 e indicando que voc√™ deseja que a √°rea esteja entre os pontos de corte.

```python
upper_limit = norm(m, sigma_m).ppf(0.975)
z = (upper_limit - m) / sigma_m
z
```

#üî¢ Simula√ß√£o

Esta simula√ß√£o ilustra intervalos de confian√ßa.

Para cada execu√ß√£o da simula√ß√£o, 100 experimentos de amostra s√£o conduzidos e um intervalo de confian√ßa na m√©dia √© computado para cada experimento. Em cada experimento, as pontua√ß√µes s√£o amostradas de uma popula√ß√£o com uma m√©dia de 50 e um desvio padr√£o de 10.

Portanto, o par√¢metro sendo estimado √© sempre 50. Ambos os intervalos de confian√ßa de 95% e 99% s√£o computados para cada experimento.

Se o intervalo de confian√ßa de 95% contiver a m√©dia populacional de 50, ent√£o o intervalo de confian√ßa √© mostrado como uma linha laranja.

Se o intervalo n√£o contiver 50, ele √© mostrado como uma linha vermelha.

Voc√™ pode escolher o tamanho da amostra para cada experimento.

```python
import matplotlib.pylab as plt
import numpy as np
from scipy.stats import norm
import pandas as pd
```

```python
figure, ax = plt.subplots(2, figsize=(15, 8), sharex=True)

mean, std = 50, 10
n = 2 #Aqui voc√™ tirou uma √∫nica amostra com n = 2 (tamanho pequeno).

x_axis = np.arange(mean - 4 * std, mean + 4 * std, 0.01)

this_norm = norm(mean, std)

ser = pd.Series(this_norm.pdf(x_axis),
                index=x_axis,
                name='Population')

_ = ser.plot(ax=ax[0], legend=True) #Distribui√ß√£o azul em forma de sino: representa a popula√ß√£o original, que √© uma distribui√ß√£o normal com m√©dia 50 e desvio padr√£o 10.

ax[0].axvline(x=this_norm.ppf(0.5), color='k', linestyle='--', linewidth=2)
sample = pd.Series(norm(mean, std).rvs(n))
mean_M = sample.mean()
s_M = sample.std() / (n ** 0.5)
this_norm = norm(mean_M, s_M)
ser = pd.Series(this_norm.pdf(x_axis),
                index=x_axis,
                name='Sampling Distribution of the Sample Means') #A curva mostra a distribui√ß√£o normal simulada com a m√©dia da amostra e seu erro padr√£o.

_ = ser.plot(ax=ax[1], legend=True)

#Intervalo de Confian√ßa de 95% (vermelho ou laranja)
#S√£o as duas linhas vermelhas ou laranja tracejadas ao redor da m√©dia da amostra.
#Se a m√©dia real da popula√ß√£o (50) cai fora do intervalo ‚Üí vermelho (erro de cobertura).
#Se ela cai dentro ‚Üí laranja (acerto).
#Como n √© pequeno (n = 2), o erro padr√£o √© grande, e o intervalo pode n√£o cobrir o valor verdadeiro.
lb1, ub1 = this_norm.ppf(0.025), this_norm.ppf(0.975)
color1 = 'red' if mean < lb1 or mean > ub1 else 'orange'

#Intervalo de Confian√ßa de 99% (azul ou branco)
#S√£o as linhas azuis/brancas ‚Äî intervalo mais largo.
#Maior confian√ßa ‚Üí maior margem de erro.
#Novamente, depende se o valor 50 cai dentro do intervalo.
lb2, ub2 = this_norm.ppf(0.005), this_norm.ppf(0.995)
color2 = 'white' if mean < lb1 or mean > ub1 else 'blue'

ax[1].axvline(x=lb1, color=color1, linestyle='--', linewidth=2)
ax[1].axvline(x=ub1, color=color1, linestyle='--', linewidth=2)
ax[1].axvline(x=mean_M, color='k', linestyle='--', linewidth=2)  #Linha preta tracejada: √© a m√©dia da amostra.
ax[1].axvline(x=lb2, color=color2, linestyle='--', linewidth=2)
ax[1].axvline(x=ub2, color=color2, linestyle='--', linewidth=2)
plt.show()
pd.DataFrame([sample.count(), sample.mean(), sample.std(),
              lb1, ub1, lb2, ub2],
             index=['count', 'mean', 'std',
                    '95% LB', '95% UB', '99% LB', '99% UB']).T
```

* M√©dia amostral ‚âà 50.97 ‚Üí pr√≥xima de 50
* Mas com n = 2, o desvio √© muito pequeno para capturar a variabilidade real. Amostras pequenas (n = 2) produzem intervalos inst√°veis.
* Os limites do IC s√£o baseados nesse erro padr√£o pequeno, o que pode gerar intervalos enganadores com n pequeno.
* Os intervalos de 99% s√£o mais largos e tendem a cobrir mais frequentemente o valor verdadeiro.



Agora faremos mais 100 experimentos simulados

```python
m = 1000
n = 100

mean, std = 50, 10 #uma popula√ß√£o normal com m√©dia 50 e desvio padr√£o 10
df = []

#calcular a m√©dia da amostra, o desvio padr√£o da amostra, Intervalos de confian√ßa de 95% e 99%

for _ in range(m):
    sample = pd.Series(norm(mean, std).rvs(n)) #Cria uma amostra de tamanho n da distribui√ß√£o normal N(50, 10)
    mean_M = sample.mean() #Calcula a m√©dia da amostra
    s_M = sample.std() / (n ** 0.5) #Calcula o erro padr√£o da m√©dia
    this_norm = norm(mean_M, s_M) #Cria uma nova distribui√ß√£o normal centrada na m√©dia da amostra com o erro padr√£o calculado
    lb1, ub1 = this_norm.ppf(0.025), this_norm.ppf(0.975) #O .ppf() fornece os percentis da distribui√ß√£o normal
    color1 = 'red' if mean < lb1 or mean > ub1 else 'orange' #se a m√©dia verdadeira (50) est√° fora do intervalo de 95%, o IC √© vermelho (erro). caso contr√°rio, √© laranja (acerto).
    lb2, ub2 = this_norm.ppf(0.005), this_norm.ppf(0.995) #O .ppf() forn  ece os percentis da distribui√ß√£o normal
    color2 = 'white' if mean < lb2 or mean > ub2 else 'blue' #mesma l√≥gica vale para o IC de 99% (azul e branco).

    df.append([sample.count(), sample.mean(), sample.std(),
               lb1, ub1, color1, lb2, ub2, color2])


df = pd.DataFrame(df, columns=['count', 'mean', 'std',
                               '95% LB', '95% UB', '95%parecer', '99% LB', '99% UB', '99%parecer'])

df.head()
```

#‚úÖ Quando usar z ou t?

Para usar f√≥rmulas padr√£o de intervalo de confian√ßa, a distribui√ß√£o dos dados precisa ser aproximadamente normal.

Entretanto, gra√ßas ao Teorema do Limite Central, quando o tamanho da amostra for n ‚â• 30, a distribui√ß√£o das m√©dias amostrais ser√° aproximadamente normal, mesmo que a distribui√ß√£o original dos dados n√£o seja. Isso permite o uso de intervalos de confian√ßa sim√©tricos.

No entanto, **para usar a distribui√ß√£o Z precisamos saber previamente do desvio padr√£o populacional. Ocorre que, na pr√°tica, quase nunca conhecemos o desvio padr√£o da popula√ß√£o.**

Alternativamente, devemos usar a distribui√ß√£o t em vez da distribui√ß√£o normal quando a vari√¢ncia n√£o for conhecida e tiver que ser estimada a partir de dados de amostra.

Quando o tamanho da amostra √© grande, digamos 30 ou mais (100 em algumas literaturas, por seguran√ßa), a distribui√ß√£o t √© muito semelhante √† distribui√ß√£o normal padr√£o. No entanto, com tamanhos de amostra menores, a distribui√ß√£o t √© leptoc√∫rtica, o que significa que tem relativamente mais pontua√ß√µes em suas caudas do que a distribui√ß√£o normal (Ela √© mais larga - tem caudas mais pesadas-, ou seja, mais conservadora)

```python
def compare_normal_and_t(deg_frees):
    x_axis = np.arange(-5, 5, 0.01)
    df = [t(deg_free).pdf(x_axis) for deg_free in deg_frees] + \
         [norm().pdf(x_axis)]
    columns = ['df={}'.format(deg_free) for deg_free in deg_frees] + \
              ['norm']
    df = pd.DataFrame(list(map(list, zip_longest(*df))),
                      columns=columns,
                      index=x_axis)
    _ = df.plot()

compare_normal_and_t([2, 4, 10])
print('Uma compara√ß√£o entre distribui√ß√µes t com 2, 4 e 10 df \n'
'e a distribui√ß√£o normal padr√£o. A distribui√ß√£o com o \n'
'pico mais baixo √© a distribui√ß√£o de 2 df, o pr√≥ximo mais baixo √© o de 4 df, \n'
'o menor depois disso √© o de 10 df, e o mais alto √© a distribui√ß√£o normal padr√£o.')
```

#üë®‚Äçüíª Demonstra√ß√£o do c√°lculo do Intervalo de Confian√ßa da M√©dia quando desvio padr√£o √© desconhecido

Suponha que os cinco n√∫meros a seguir sejam amostrados de uma distribui√ß√£o normal: 2, 3, 5, 6 e 9, e que o desvio padr√£o seja desconhecido. Os primeiros passos s√£o calcular a m√©dia amostral e o desvio padr√£o:

```python
import pandas as pd
```

```python
ser = pd.Series([2, 3, 5, 6, 9])
sample_mean = ser.mean()
sample_std = ser.std()
sample_mean, sample_std
```

O pr√≥ximo passo √© estimar o erro padr√£o da m√©dia. Se soub√©ssemos a vari√¢ncia populacional, poder√≠amos usar a seguinte f√≥rmula:

$\sigma_m = \frac{\sigma}{\sqrt{N}}$

Em vez disso, calculamos uma estimativa do erro padr√£o ($s_M$):

$s_m = \frac{s}{\sqrt{N}}=1,225$

```python
sigma_m = sample_std / (ser.size ** 0.5)
sigma_m
```

O pr√≥ximo passo √© encontrar o valor de t. Como voc√™ pode ver na Tabela 1, o valor para o intervalo de 95% para df = N - 1 = 4 √© 2,776.

```python
t_stat = t(4).ppf(0.975)
t_stat
```

O intervalo de confian√ßa √© ent√£o calculado da mesma forma que para $\sigma_M$. As √∫nicas diferen√ßas s√£o que $s_M$ e $t$ s√£o usados ‚Äã‚Äãem vez de $\sigma_M$ e $Z$. Abaixo voc√™ tem a f√≥rmula da margem de erro:

eaea2112eae.png

E ent√£o, de forma mais geral, a f√≥rmula para o intervalo de confian√ßa de 95% na m√©dia √©:

Limite inferior = $M - (t_{CL})(s_M)$

Limite superior = $M + (t_{CL})(s_M)$

onde $M$ √© a m√©dia amostral, $t_{CL}$ √© o $t$ para o n√≠vel de confian√ßa desejado (0,95 no exemplo acima) e $s_M$ √© o erro padr√£o estimado da m√©dia.

Dessa forma, calcule para o nosso exemplo:

Limite inferior = $5 - (2,776)(1,225) = 1,60$

Limite superior = $5 + (2,776)(1,225) = 8,40$

```python
df = ser.size - 1
t_stat = 2.776

lower_limit = sample_mean - t_stat * sigma_m
upper_limit = sample_mean + t_stat * sigma_m
area = t(df, loc=sample_mean, scale=sigma_m).cdf(upper_limit) - \
       t(df, loc=sample_mean, scale=sigma_m).cdf(lower_limit)

msg = '{:.2f}% da probabilidade est√° dentro do intervalo {:.2f} to {:.2f}.'
print(msg.format(area*100, lower_limit, upper_limit))
```
