##üè• Permuta√ß√µes: Ordem dos Pacientes para Cirurgia

üí° Exemplos Pr√°ticos em Sa√∫de

Imagine que temos uma lista com **4 pacientes** aguardando cirurgia. Queremos saber em quantas **ordens diferentes** podemos organizar a sequ√™ncia das cirurgias

O n√∫mero de permuta√ß√µes √© dado por:

üßÆ F√≥rmula:
$P(n) = n!$

Exemplo:
$4! = 4 \times 3 \times 2 \times 1 = 24$

```python
from math import factorial

# N√∫mero de pacientes
n = 4

# C√°lculo da permuta√ß√£o
permutacoes = factorial(n)
print(f"N√∫mero de permuta√ß√µes para {n} pacientes: {permutacoes}")
```

## üóÉÔ∏è Combina√ß√µes: Escolha de Procedimentos

Agora, considere que temos **6 tipos de exames** para realizar em cada paciente, mas s√≥ podemos escolher **3**. Quantas combina√ß√µes diferentes podemos formar?

üßÆ F√≥rmula:
$C(n, r) = \frac{n!}{(n - r)! \times r!}$

```python
from itertools import combinations

# N√∫mero de tipos de exames
n = 6
r = 3

# Gera todas as combina√ß√µes
comb = list(combinations(range(1, n+1), r))
print(f"N√∫mero de combina√ß√µes de {r} exames entre {n} dispon√≠veis: {len(comb)}")
print(comb)
```

## üîÑ Multiplica√ß√£o de Op√ß√µes: Sequ√™ncia de Procedimentos

Suponha que cada paciente passe por **3 etapas de triagem**:
1. Coleta de dados
2. Exame f√≠sico
3. Testes laboratoriais

Se houverem **3 op√ß√µes para cada etapa**, quantas combina√ß√µes completas de atendimento podemos ter?

üßÆ F√≥rmula:
$Total = n_1 \times n_2 \times n_3$

```python
# Op√ß√µes em cada etapa
opcoes = [3, 3, 3]

total = 1
for opcao in opcoes:
    total *= opcao

print(f"N√∫mero total de combina√ß√µes de atendimento: {total}")
```
