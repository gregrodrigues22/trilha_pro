As distribui√ß√µes estat√≠sticas est√£o profundamente conectadas ao conceito de **probabilidade** porque cada vari√°vel aleat√≥ria est√° associada a uma **distribui√ß√£o de probabilidade**, que descreve como os valores poss√≠veis da vari√°vel est√£o distribu√≠dos e qual √© a probabilidade associada a cada valor.

O conceito de vari√°vel aleat√≥ria √© particularmente √∫til em situa√ß√µes nas quais se disp√µe de um n√≠vel de conhecimento parcial ou incompleto do comportamento de uma grandeza que est√° sendo estudada. Essa incerteza pode ser ent√£o introduzida sob a forma de um modelo probabil√≠stico.

üîó Conectando Distribui√ß√µes e Probabilidade

Distribui√ß√µes s√£o, essencialmente, representa√ß√µes gr√°ficas da **probabilidade**. Dependendo do tipo de vari√°vel (discreta ou cont√≠nua), a forma de calcular e interpretar a probabilidade varia.


## üé≤ Distribui√ß√µes Discretas: Fun√ß√£o de Massa de Probabilidade (FMP)

Dizemos que X √© uma vari√°vel aleat√≥ria discreta se o n√∫mero de valores que ela pode admitir √© finito. Por exemplo, em uma cl√≠nica m√©dica, o n√∫mero de pacientes atendidos por dia √© anotado para fins de monitoramento da demanda. Suponha que, em determinado per√≠odo, o n√∫mero de pacientes atendidos diariamente varia entre 0 e 20. Nesse caso, a v.a. X √© o n√∫mero de pacientes atendidos por dia. Assim, X pode admitir os valores 0, 2, 3, 4, ..., 10, 11, 12, 13, 14, ..., 20. Portanto, claramente X √© uma v.a. aleat√≥ria discreta.

Uma distribui√ß√£o discreta descreve a probabilidade de ocorr√™ncia de cada valor de uma vari√°vel aleat√≥ria discreta. Com uma distribui√ß√£o de probabilidade discreta, cada valor poss√≠vel da vari√°vel aleat√≥ria discreta pode ser associado a uma probabilidade diferente de zero. Deste modo, uma distribui√ß√£o de probabilidade discreta √©, por vezes, apresentada em forma de tabela. Com uma distribui√ß√£o discreta √© poss√≠vel calcular a probabilidade de que X seja exatamente igual ou maior ou menor que algum valor.

A **Fun√ß√£o de Massa de Probabilidade (FMP)** atribui probabilidades a valores espec√≠ficos de uma vari√°vel aleat√≥ria discreta.

P(X = x) = f(x)

Essa f√≥rmula indica que a probabilidade da vari√°vel \(X\) assumir um valor espec√≠fico \(x\) √© dada pela fun√ß√£o de massa de probabilidade \(f(x)\).

Para esse caso, a distribui√ß√£o que melhor se aplica √© a Distribui√ß√£o de Poisson, pois estamos lidando com a contagem de eventos (n√∫mero de pacientes atendidos por dia) em um intervalo de tempo fixo (um dia).

A Distribui√ß√£o de Poisson √© usada para modelar a probabilidade de um n√∫mero espec√≠fico de eventos ocorrer em um intervalo de tempo ou espa√ßo, quando os eventos ocorrem de forma independente e com uma taxa m√©dia conhecida.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

media = 10  # M√©dia de pacientes por hora
x = np.arange(0, 30)
y = stats.poisson.pmf(x, media)

plt.bar(x, y, color='skyblue')
plt.title('Distribui√ß√£o Poisson - Pacientes por Dia')
plt.xlabel('N√∫mero de Pacientes')
plt.ylabel('Probabilidade')
plt.legend()
plt.show()
```

Qual a probabilidade de atender mais de 15 pacientes?

```python
import scipy.stats as stats

media = 10  # M√©dia de pacientes por hora
x = np.arange(0, 30)
y = stats.poisson.pmf(x, media)

plt.bar(x, y, color='skyblue')
plt.title('Distribui√ß√£o Poisson - Pacientes por Dia')
plt.xlabel('N√∫mero de Pacientes')
plt.ylabel('Probabilidade')
plt.fill_between(x, y, where=(x >= 15), color='red', alpha=0.7, label='P(X >= 15)')
plt.legend()
plt.show()
```

A fun√ß√£o sf do pacote scipy.stats significa Survival Function (fun√ß√£o de sobreviv√™ncia). Ela retorna a probabilidade complementar da fun√ß√£o de distribui√ß√£o acumulada (CDF). Em outras palavras, a sf calcula a probabilidade de que a vari√°vel aleat√≥ria seja maior ou igual a um certo valor.

stats.poisson.sf(k, mu)

Onde:

* k √© o valor a partir do qual queremos calcular a probabilidade (no nosso caso, 14).
* mu √© a m√©dia da distribui√ß√£o de Poisson (no nosso caso, vari√°vel media).

```python
probabilidade_x_maior_igual_15 = stats.poisson.sf(14, media)

print(f"A probabilidade de X >= 15 √©: {probabilidade_x_maior_igual_15:.4f}")
```

A altura das barras somam 0,0835; por conseguinte, a probabilidade de que o n√∫mero de pacientes atendidos por dia seja de 15 ou mais √© 8,35%.

Ao longo desse m√≥dulo, apresentaremos a alguns dos modelos probabil√≠sticos discretos que costumam ser mais utilizados nas aplica√ß√µes pr√°ticas da Estat√≠stica em sa√∫de. Existem diversas modelagens para prever a probabilidade de vari√°veis discretas, como os modelos que envolvem ensaios de Bernoulli (Bernoulli, Binomial, Geom√©trico) e o modelo de Poisson.

## üåä Distribui√ß√µes Cont√≠nuas: Fun√ß√£o Densidade de Probabilidade (FDP)

A distribui√ß√£o cont√≠nua descreve as probabilidades dos poss√≠veis valores de uma vari√°vel aleat√≥ria cont√≠nua. Uma vari√°vel aleat√≥ria cont√≠nua √© uma vari√°vel aleat√≥ria com um conjunto de valores poss√≠veis (conhecidos como intervalos) que √© infinito e incont√°vel.

As probabilidades de vari√°veis aleat√≥rias cont√≠nuas (
ùëã) s√£o definidas como a √°rea sob a curva da sua distribui√ß√£o. Assim, apenas as faixas de valores podem ter uma probabilidade diferente de zero. A probabilidade de que uma vari√°vel aleat√≥ria cont√≠nua seja igual a algum valor √© sempre zero. Isso acontece porque a probabilidade de uma vari√°vel cont√≠nua assumir exatamente um valor espec√≠fico √© sempre zero.

Normal (Cont√≠nua): "Qual √© a probabilidade de um paciente pesar 70 kg?" ‚Üí Sempre zero.

A **Fun√ß√£o Densidade de Probabilidade (FDP)** n√£o atribui a probabilidade a um √∫nico valor, mas sim √† √°rea sob a curva em um intervalo.

A f√≥rmula da FDP mostra que a probabilidade de a vari√°vel \(X\) estar em um intervalo \([a, b]\) √© dada pela integral da fun√ß√£o densidade \(f(x)\) nesse intervalo. A √°rea sob a curva representa a probabilidade acumulada dentro do intervalo.


Exemplo de Distribui√ß√£o Normal:
A distribui√ß√£o normal √© uma das mais comuns e segue a famosa curva em sino.

O peso de pacientes em um hospital pode ser modelado como uma distribui√ß√£o normal.

```python
media = 70  # M√©dia de peso em kg
desvio = 10  # Desvio padr√£o
x = np.linspace(40, 100, 100) #A fun√ß√£o np.linspace(40, 100, 100) cria uma s√©rie de 100 valores igualmente espa√ßados entre 40 e 100.
y = stats.norm.pdf(x, media, desvio) #A fun√ß√£o stats.norm.pdf(x, media, desvio) calcula a fun√ß√£o densidade de probabilidade (FDP) da Distribui√ß√£o Normal para cada valor em x
#Isso retorna uma lista de valores de densidade para cada peso no intervalo.

plt.plot(x, y, color='darkblue')
plt.title('Distribui√ß√£o Normal - Peso de Pacientes')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidade de Probabilidade')
plt.fill_between(x, y, alpha=0.5)
plt.show()
```

## üìä Fun√ß√£o de distribui√ß√£o cumulativa (CFD)

A Fun√ß√£o de Distribui√ß√£o Acumulada (CDF) pode ser calculada tanto para vari√°veis discretas quanto para vari√°veis cont√≠nuas. No entanto, a forma de interpreta√ß√£o e o c√°lculo espec√≠fico variam conforme o tipo de vari√°vel.

Em todo caso, a interpreta√ß√£o final √© a mesma: probabilidade acumulada at√© o ponto

### üìä CDF para Vari√°veis Discretas

A CDF para vari√°veis discretas √© a soma acumulada das probabilidades de todos os valores menores ou iguais a um valor espec√≠fico.

üöÄ Exemplo: Distribui√ß√£o Binomial
Vamos calcular a probabilidade acumulada de obter at√© 3 sucessos em 10 tentativas com uma probabilidade de sucesso de 20%.

```python
from scipy.stats import binom

# Par√¢metros
n = 10  # N√∫mero de tentativas
p = 0.20  # Probabilidade de sucesso
x = 3  # N√∫mero m√°ximo de sucessos

# C√°lculo da CDF
prob_acumulada = binom.cdf(x, n, p)
print(f"üî¢ A probabilidade acumulada de obter at√© 3 sucessos √©: {prob_acumulada:.4%}")
```

###üåä CDF para Vari√°veis Cont√≠nuas

A CDF para vari√°veis cont√≠nuas √© a integra√ß√£o da fun√ß√£o densidade de probabilidade (FDP) da esquerda at√© o valor desejado.

üöÄ Exemplo: Distribui√ß√£o Normal
Calcular a probabilidade acumulada de um paciente ter peso at√© 80 kg, sabendo que a m√©dia √© 70 kg e o desvio padr√£o √© 10 kg.

```python
from scipy.stats import norm

# Par√¢metros
media = 70  # M√©dia de peso
desvio = 10  # Desvio padr√£o
x = 80  # Peso m√°ximo

# C√°lculo da CDF
prob_acumulada = norm.cdf(x, media, desvio)
print(f"üìè A probabilidade acumulada de pesar at√© 80 kg √©: {prob_acumulada:.4%}")
```

```python
probabilidade = stats.norm.cdf(80, media, desvio) - stats.norm.cdf(60, media, desvio)
print(f'A probabilidade de um paciente ter peso entre 60 e 80 kg √©: {probabilidade:.4f}')
```

###üåü SF: A Fun√ß√£o de Sobreviv√™ncia (Survival Function) - O Complemento da CDF

A Fun√ß√£o de Sobreviv√™ncia (SF) √© simplesmente o complemento da CDF (Fun√ß√£o de Distribui√ß√£o Acumulada). Enquanto a CDF calcula a probabilidade acumulada at√© um determinado valor, a SF calcula a probabilidade acumulada a partir desse valor.

üè• Exemplo na Sa√∫de: N√∫mero de Interna√ß√µes
Vamos considerar um exemplo:

Um hospital tem uma m√©dia de 5 interna√ß√µes por dia.
Queremos saber a probabilidade de ter mais de 7 interna√ß√µes em um dia

```python
from scipy.stats import poisson

# Par√¢metros
media = 5  # M√©dia de interna√ß√µes por dia
x = 7  # N√∫mero de interna√ß√µes desejado

# C√°lculo da SF (Survival Function)
prob_mais_que_7 = poisson.sf(x, media)

print(f"üöë A probabilidade de ter mais de 7 interna√ß√µes em um dia √©: {prob_mais_que_7:.4%}")
```

Por Que Usar a SF ao Inv√©s da CDF?
* Quando queremos a probabilidade de ultrapassar um valor espec√≠fico.
* Para facilitar o c√°lculo de valores extremos sem ter que somar manualmente todas as probabilidades superiores.
* Para vari√°veis cont√≠nuas ou discretas, funciona da mesma forma, pois sempre calcula o complemento da CDF.

Quando Usar Cada Uma:
* CDF: Quando queremos a probabilidade menor ou igual a um valor espec√≠fico.
* SF: Quando queremos a probabilidade maior que um valor espec√≠fico.
