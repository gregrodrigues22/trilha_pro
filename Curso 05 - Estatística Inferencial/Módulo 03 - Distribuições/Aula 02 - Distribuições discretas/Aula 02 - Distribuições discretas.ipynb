## üí° Conceito de Vari√°veis Aleat√≥rias Discretas


Uma vari√°vel aleat√≥ria discreta √© aquela que pode assumir um n√∫mero **finito ou cont√°vel de valores**. Cada valor poss√≠vel est√° associado a uma probabilidade, e a soma de todas as probabilidades √© igual a **1**.

Exemplos na Sa√∫de:
- üè• **N√∫mero de pacientes atendidos por hora** em uma emerg√™ncia.
- ü¶† **N√∫mero de infec√ß√µes** em uma ala hospitalar por dia.
- üìù **N√∫mero de interna√ß√µes** em um m√™s.

As vari√°veis aleat√≥rias discretas s√£o usadas para representar contagens ou eventos que ocorrem em um intervalo de tempo ou espa√ßo. Elas s√£o extremamente √∫teis para modelar situa√ß√µes cl√≠nicas e administrativas no contexto da sa√∫de.

## üéØ Modelos Probabil√≠sticos Discretos Dicot√¥micos ou Bin√°rios

Quando a vari√°vel aleat√≥ria s√≥ pode assumir dois valores poss√≠veis, geralmente representados como "sucesso" e "fracasso", "1" e "0", ou "verdadeiro" e "falso".

Os modelos de Bernoulli, binomial, geom√©trico s√£o usados para modelar fen√¥menos probabil√≠sticos com duas possibilidades (sucesso ou fracasso).

### üü¢ Modelo de Bernoulli


#### üìñ Descri√ß√£o

A Distribui√ß√£o de Bernoulli √© uma das distribui√ß√µes discretas mais simples e fundamentais na teoria das probabilidades. Ela modela um experimento aleat√≥rio que possui apenas dois resultados poss√≠veis.  

Caracter√≠sticas da Distribui√ß√£o de Bernoulli

* Eventos Independentes: Cada tentativa √© independente das demais.
* Dois Resultados Poss√≠veis: Sucesso (1) ou Fracasso (0).
* Par√¢metro: p = Probabilidade de sucesso. 1 ‚àí ùëù = Probabilidade de fracasso.
* N√∫mero de Sucessos (X): X = 0 OU ùëã = 1

####üî¢ Fun√ß√£o Massa de Probabilidade (FMP)

√â utilizada para representar a ocorr√™ncia de um evento bin√°rio. O **Modelo de Bernoulli** representa um experimento com **dois resultados poss√≠veis**: **sucesso** ou **fracasso**. A probabilidade de sucesso √© denotada por **p**, enquanto a probabilidade de fracasso √© **(1 - p)**.

Calcular a probabilidade de um paciente ser curado (sucesso) em um tratamento experimental com 70% de chance de efic√°cia.

```python
from scipy.stats import bernoulli

p = 0.7  # Probabilidade de sucesso (cura)
x = 1  # Sucesso

probabilidade = bernoulli.pmf(x, p)
print(f"Probabilidade de cura: {probabilidade:.4%}")
```

#### üìù Propriedades da Distribui√ß√£o de Bernoulli

22221211.png

A moda da distribui√ß√£o de Bernoulli √©:
* 1, se ùëù > 0.5
* 0, se ùëù < 0.5
* Ambos, se ùëù = 0.5

* O desvio padr√£o √© a raiz quadrada da vari√¢ncia e representa a dispers√£o dos resultados em torno da m√©dia.

* A m√©dia indica o valor esperado da vari√°vel aleat√≥ria, ou seja, a probabilidade de sucesso.

* A vari√¢ncia mede o grau de dispers√£o dos valores em torno da m√©dia.

**Exemplo:** Um teste r√°pido para detec√ß√£o de uma doen√ßa √© realizado em um paciente, e a probabilidade de um resultado positivo √© de 70%. Vamos calcular as propriedades da distribui√ß√£o (m√©dia, vari√¢ncia, desvio padr√£o e mediana) para um √∫nico teste.



```python
from scipy.stats import binom

# Par√¢metros da distribui√ß√£o binomial
n = 10  # N√∫mero de ensaios (ex: n√∫mero de pacientes testados)
p = 0.7 # Probabilidade de sucesso (ex: probabilidade de testar positivo)

# C√°lculo das propriedades
media = binom.mean(n, p)
variancia = binom.var(n, p)
desvio_padrao = binom.std(n, p)
mediana = binom.median(n, p)

print(f"M√©dia: {media:.2f}")
print(f"Vari√¢ncia: {variancia:.2f}")
print(f"Desvio Padr√£o: {desvio_padrao:.2f}")
print(f"Mediana: {mediana:.2f}")

```

* M√©dia (0.70): A m√©dia representa a probabilidade esperada de sucesso (resultado positivo). Como a probabilidade de um teste positivo √© de 70%, a m√©dia tamb√©m √© 0.70.

* Vari√¢ncia (0.21): A vari√¢ncia indica a dispers√£o dos valores ao redor da m√©dia. Quanto maior a vari√¢ncia, maior a incerteza sobre o resultado do teste.

* Desvio Padr√£o (0.46): O desvio padr√£o √© a raiz quadrada da vari√¢ncia e representa a varia√ß√£o t√≠pica dos resultados.

* Mediana (1.00): A mediana representa o valor que divide a distribui√ß√£o em duas partes iguais. Como a probabilidade de sucesso √© maior que 0.5, a mediana √© 1 (resultado positivo).

#### üìä Gr√°fico da Fun√ß√£o Massa de Probabilidade (FMP)

**Exemplo na Sa√∫de:**
Imagine que 80% dos pacientes de uma cl√≠nica testem positivo para uma infec√ß√£o viral. O sucesso √© considerado como o teste positivo.

**Gr√°fico: Distribui√ß√£o de Bernoulli**
- Sucesso (presen√ßa de v√≠rus): ‚úÖ
- Fracasso (aus√™ncia de v√≠rus): ‚ùå

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import binom
```

```python
p = 0.8  # Probabilidade de sucesso (ex: presen√ßa de v√≠rus em pacientes)
x = [0, 1]  # 0 = fracasso, 1 = sucesso
y = [1 - p, p]
plt.bar(x, y, color='lightblue')
plt.title('Distribui√ß√£o de Bernoulli - Presen√ßa de V√≠rus')
plt.xlabel('Resultado')
plt.ylabel('Probabilidade')
plt.xticks([0, 1], ['Ausente', 'Presente'])
plt.show()
```

####üìà Fun√ß√£o de Distribui√ß√£o Cumulativa (CDF)

```python
# Fun√ß√£o de Distribui√ß√£o Cumulativa
cdf = bernoulli.cdf(x, p)

# Exibindo os valores
for valor, prob in zip(x, cdf):
    print(f'P(X ‚â§ {valor}) = {prob:.2f}')
```

####üõë Fun√ß√£o de Sobreviv√™ncia (SF)

A Fun√ß√£o de Sobreviv√™ncia (SF) √© o complemento da CDF e representa a probabilidade de o sucesso n√£o ter ocorrido at√© o valor x:

```python
# Fun√ß√£o de Sobreviv√™ncia
sf = bernoulli.sf(x, p)

# Exibindo os valores
for valor, prob in zip(x, sf):
    print(f'P(X > {valor}) = {prob:.2f}')
```

### üîµ Modelo Binomial

#### üìñ Descri√ß√£o

A Distribui√ß√£o Binomial √© uma distribui√ß√£o de probabilidade discreta que descreve o n√∫mero de sucessos em uma sequ√™ncia de n experimentos de Bernoulli independentes, onde cada experimento possui:

* Dois resultados poss√≠veis: Sucesso (1) ou Fracasso (0).
* Probabilidade constante de sucesso (p).
* O n√∫mero de experimentos (n) √© fixo.




A vari√°vel aleat√≥ria X representa o n√∫mero de sucessos obtidos em n tentativas.

#### üî¢ Fun√ß√£o Massa de Probabilidade

A Fun√ß√£o Massa de Probabilidade (FMP) da distribui√ß√£o binomial descreve a probabilidade de obter exatamente um n√∫mero espec√≠fico de sucessos em um n√∫mero fixo de tentativas, quando cada tentativa tem a mesma probabilidade de sucesso.

a212121.png

```python
# Par√¢metros da distribui√ß√£o binomial
n = 10  # N√∫mero de ensaios (ex: n√∫mero de pessoas em um estudo)
p = 0.2  # Probabilidade de sucesso em cada ensaio (ex: probabilidade de uma pessoa ter uma certa doen√ßa)

# Criando a distribui√ß√£o binomial
binom_dist = binom(n, p)

# Calculando a probabilidade de obter exatamente k sucessos
k = 3  # N√∫mero de sucessos (ex: n√∫mero de pessoas com a doen√ßa)
prob_k_sucessos = binom_dist.pmf(k)
print(f"Probabilidade de exatamente {k} pessoas terem a doen√ßa: {prob_k_sucessos}")
```

####üìê Propriedades da Distribui√ß√£o Binomial

21211221.png

* M√©dia (Esperan√ßa): Representa o valor m√©dio esperado de sucessos em uma s√©rie de experimentos

* Vari√¢ncia: Representa a dispers√£o em torno da m√©dia

* Desvio padr√£o: Representa a varia√ß√£o t√≠pica em torno da m√©dia

* Moda: Representa o valor mais prov√°vel (ou que ocorre com mais frequ√™ncia) de sucessos

**Exemplo:** Um hospital realiza um estudo para avaliar a efic√°cia de uma nova vacina contra uma doen√ßa. Sabe-se que a vacina tem uma efic√°cia de 70%. O estudo foi realizado com 50 pacientes, e o objetivo √© verificar a expectativa e a variabilidade no n√∫mero de pacientes imunizados.

```python
from scipy.stats import binom

# Par√¢metros
n = 50  # N√∫mero de pacientes vacinados
p = 0.70  # Probabilidade de sucesso (imuniza√ß√£o)

# C√°lculo das propriedades usando scipy.stats
media = binom.mean(n, p)
variancia = binom.var(n, p)
desvio_padrao = binom.std(n, p)
moda = binom.median(n, p)

print(f"M√©dia (Esperan√ßa): {media}")
print(f"Vari√¢ncia: {variancia}")
print(f"Desvio Padr√£o: {desvio_padrao:.2f}")
print(f"Moda: {moda}")

```

* M√©dia (35): Espera-se que, em m√©dia, 35 pacientes desenvolvam imunidade.
* Vari√¢ncia (10.5): A dispers√£o em torno da m√©dia indica uma variabilidade moderada no n√∫mero de pacientes imunizados.
* Desvio Padr√£o (3.24): A varia√ß√£o t√≠pica em torno da m√©dia √© de aproximadamente 3 pacientes.
* Moda (35): O valor mais prov√°vel de pacientes imunizados tamb√©m √© 35.




Na distribui√ß√£o binomial, a moda e a m√©dia tendem a ser iguais ou muito pr√≥ximas, especialmente quando:
* O n√∫mero de tentativas (n) √© grande.
* A probabilidade de sucesso (p) est√° pr√≥xima de 0,5.

Para valores extremos de  p (muito pr√≥ximos de 0 ou 1) e poucos experimentos, a moda pode diferir da m√©dia, mas, em geral, em situa√ß√µes pr√°ticas de sa√∫de, essas diferen√ßas s√£o pequenas.

####üìä Gr√°fico da Fun√ß√£o Massa de Probabilidade (FMP)

Vamos plotar a fun√ß√£o de massa de probabilidade para observar como a probabilidade se comporta

**Exemplo na Sa√∫de:**
Um teste r√°pido para detec√ß√£o de uma doen√ßa √© realizado em 10 pacientes, e a probabilidade de um resultado positivo √© de 70%. Queremos calcular a distribui√ß√£o de resultados positivos.

```python
n = 10  # N√∫mero de tentativas (ex: pacientes testados)
p = 0.7  # Probabilidade de sucesso (ex: teste positivo para doen√ßa)
x = np.arange(0, n+1)
y = stats.binom.pmf(x, n, p) #C√≥digo para calcular a fun√ß√£o massa de probabilidade para fun√ß√£o binomial
plt.bar(x, y, color='orange')
plt.title('Distribui√ß√£o Binomial - Testes Positivos para Doen√ßa')
plt.xlabel('N√∫mero de Sucessos')
plt.ylabel('Probabilidade')
plt.show()
```

O gr√°fico apresentado mostra uma Distribui√ß√£o Binomial que representa a probabilidade de obter um determinado n√∫mero de sucessos (n√∫mero de testes positivos para uma doen√ßa) em um conjunto de 10 pacientes testados, considerando que a probabilidade de sucesso (teste positivo) √© de 70%.

* Eixo X (N√∫mero de Sucessos): Varia de 0 a 10, representando a quantidade de pacientes que tiveram resultado positivo para a doen√ßa em um grupo de 10 pacientes testados.

* Eixo Y (Probabilidade): Representa a probabilidade de ocorr√™ncia de cada n√∫mero de sucessos.

* Barras de Probabilidade: Cada barra mostra a probabilidade espec√≠fica de obter exatamente aquele n√∫mero de sucessos.

* Moda e Valor mais Prov√°vel: O pico do gr√°fico est√° no valor de 7 sucessos, indicando que a maior probabilidade √© que 7 pacientes testados apresentem teste positivo. Isso faz sentido, pois a m√©dia te√≥rica da distribui√ß√£o √©, como veremos abaixo: n‚ãÖp=10‚ãÖ0,7=7
*  gr√°fico apresenta uma leve assimetria √† esquerda, o que √© esperado, j√° que a probabilidade de sucesso (ùëù=0,7) √© relativamente alta.
* Quando a probabilidade √© pr√≥xima de 0,5, o gr√°fico √© mais sim√©trico.

####üìà Fun√ß√£o de Distribui√ß√£o Cumulativa (CDF)

A CDF da distribui√ß√£o binomial calcula a probabilidade acumulada de obter at√© k sucessos em n tentativas:

Suponha que, em um hospital, existe uma probabilidade de 10% de um paciente ser internado ap√≥s uma consulta de emerg√™ncia. Se houver uma alta probabilidade de ter 5 ou mais interna√ß√µes por semana, ser√° necess√°rio acionar um plant√£o extra de profissionais.

Naquela semana, o hospital realizou 50 atendimentos de emerg√™ncia. Qual √© a probabilidade de termos que acionar o plant√£o extra? ü§î

* X = quantidade de interna√ß√µes
* n = 50 atendimentos (n√∫mero de ensaios)
* p = probabilidade de interna√ß√£o = 10% (0,10)
* q = probabilidade de n√£o interna√ß√£o = 1 - 0,10 = 90% (0,90)

Precisamos calcular a probabilidade de ter 5 ou mais interna√ß√µes. Isso pode ser feito somando as probabilidades de 5, 6, 7,... at√© 50 interna√ß√µes, usando a Distribui√ß√£o Binomial

```python
from scipy.stats import binom

# Par√¢metros
n = 50  # N√∫mero de atendimentos
x = 5   # N√∫mero de interna√ß√µes desejadas
p = 0.10  # Probabilidade de interna√ß√£o

# C√°lculo da FMP da distribui√ß√£o binomial
probabilidade = binom.pmf(x, n, p)

print(f"A probabilidade de exatamente 5 interna√ß√µes na semana √©: {probabilidade:.4%}")
```

####üõë Fun√ß√£o de Sobreviv√™ncia (SF)

A Fun√ß√£o de Sobreviv√™ncia (SF) √© o complemento da CDF e calcula a probabilidade de obter mais de k sucessos:

aeeaea212.png

Para calcular a probabilidade de haver 5 ou mais interna√ß√µes, precisamos somar as probabilidades de todos os valores poss√≠veis de sucessos a partir de 5 at√© o n√∫mero total de tentativas (50):

P(X‚â•5)=P(X=5)+P(X=6)+P(X=7)+‚ãØ+P(X=50)

A soma dessas probabilidades pode ser simplificada utilizando a fun√ß√£o de sobreviv√™ncia (SF), que √© o complemento da fun√ß√£o de distribui√ß√£o acumulada (CDF):

P(X‚â•5)=1‚àíP(X<5)

Ou seja:

P(X‚â•5)=1‚àíCDF(4)

Podemos utilizar a fun√ß√£o binom.sf() do pacote scipy.stats para calcular diretamente essa probabilidade:

```python
from scipy.stats import binom

# Par√¢metros
n = 50  # N√∫mero de atendimentos
p = 0.10  # Probabilidade de interna√ß√£o

# Probabilidade de 5 ou mais interna√ß√µes
probabilidade = binom.sf(4, n, p)  # Complemento da CDF para x >= 5

print(f"A probabilidade de ter 5 ou mais interna√ß√µes na semana √©: {probabilidade:.4%}")
```

### üîµ Modelo Geom√©trica

#### üìñ Descri√ß√£o

A Distribui√ß√£o Geom√©trica √© uma distribui√ß√£o discreta que modela o n√∫mero de tentativas at√© o primeiro sucesso em uma sequ√™ncia de experimentos de Bernoulli independentes.



Caracter√≠sticas da Distribui√ß√£o Geom√©trica
* Eventos Independentes: Cada tentativa √© independente das demais.
* Dois Resultados Poss√≠veis: Cada tentativa resulta em sucesso (1) ou fracasso (0).
* N√∫mero de Tentativas: A vari√°vel aleat√≥ria
X representa o n√∫mero de tentativas at√© o primeiro sucesso.
* Par√¢metro:
p: Probabilidade de sucesso em cada tentativa.

####üî¢ Fun√ß√£o massa de Probabilidade (FMP)

A FMP da distribui√ß√£o geom√©trica calcula a probabilidade de que o primeiro sucesso ocorra na
ùë•
x-√©sima tentativa:

eaeeaae211.png

Onde:

* x = N√∫mero de tentativas at√© o primeiro sucesso (ùë•=1, 2, 3,‚Ä¶)
* p = Probabilidade de sucesso em uma tentativa
* 1 ‚àí ùëù = Probabilidade de fracasso

**Exemplo:** Suponha que um tratamento experimental tem 30% de chance de cura em cada aplica√ß√£o. Queremos calcular a probabilidade de que a primeira cura aconte√ßa na quarta aplica√ß√£o.



```python
from scipy.stats import geom

# Par√¢metros
p = 0.30  # Probabilidade de sucesso (cura)
x = 4     # N√∫mero de tentativas at√© o primeiro sucesso

# C√°lculo da probabilidade
prob = geom.pmf(x, p)
print(f"üåü A probabilidade de a primeira cura ocorrer na 4¬™ tentativa √©: {prob:.4%}")
```

####üìê Propriedades da Distribui√ß√£o Geom√©trica

2111221.png

Na distribui√ß√£o geom√©trica, a m√©dia nos diz o n√∫mero m√©dio de tentativas at√© o primeiro sucesso. A vari√¢ncia mede a dispers√£o dos valores em torno da m√©dia, ou seja, o qu√£o espalhados os resultados est√£o. Se a vari√¢ncia √© alta, significa que o n√∫mero de tentativas at√© o primeiro sucesso pode variar bastante. Se a vari√¢ncia √© baixa, as tentativas at√© o primeiro sucesso estar√£o mais pr√≥ximas da m√©dia.

A distribui√ß√£o geom√©trica modela o n√∫mero de tentativas at√© o primeiro sucesso. Isso significa que o valor mais prov√°vel (a moda) √© o n√∫mero 1, pois a primeira tentativa tem a maior probabilidade de ser um sucesso.

**Exemplo:** Imagine que um tratamento tem 20% de chance de sucesso a cada aplica√ß√£o.

```python
from scipy.stats import geom

# Par√¢metro da distribui√ß√£o geom√©trica
p = 0.2 # Probabilidade de sucesso

# C√°lculo das propriedades
media = geom.mean(p)
variancia = geom.var(p)
desvio_padrao = geom.std(p)
mediana = geom.median(p)

print(f"M√©dia: {media:.2f}")
print(f"Vari√¢ncia: {variancia:.2f}")
print(f"Desvio Padr√£o: {desvio_padrao:.2f}")
print(f"Mediana: {mediana:.2f}")
```

Isso significa que, em m√©dia, voc√™ precisar√° de 5 aplica√ß√µes para obter o primeiro sucesso. Isso significa que o n√∫mero de tentativas at√© o primeiro sucesso pode variar consideravelmente em torno da m√©dia (5).


####üìä Gr√°fico da Fun√ß√£o Massa de Probabilidade (FMP)

Vamos plotar a fun√ß√£o de massa de probabilidade para observar como a probabilidade decai conforme o n√∫mero de tentativas aumenta.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 20)  # N√∫mero de tentativas (1 a 20)
pmf = geom.pmf(x, p)

plt.stem(x, pmf, basefmt=" ")
plt.title("Distribui√ß√£o Geom√©trica - Probabilidade de Primeira Cura")
plt.xlabel("N√∫mero de Tentativas")
plt.ylabel("Probabilidade")
plt.show()
```

O valor da probabilidade diminui exponencialmente conforme o n√∫mero de tentativas aumenta. Portanto: A maior probabilidade ocorre na primeira tentativa.

####üìà Fun√ß√£o de distribui√ß√£o cumulativa (CDF)

**Exemplo:**  Um medicamento tem 30% de chance de cura por aplica√ß√£o. Qual √© a probabilidade de que a primeira cura ocorra at√© a 4¬™ aplica√ß√£o?

### üîµ Modelo Hipergeom√©trica

#### üìñ Descri√ß√£o

A Distribui√ß√£o Hipergeom√©trica √© uma distribui√ß√£o de probabilidade discreta que descreve o n√∫mero de sucessos em uma amostra extra√≠da sem reposi√ß√£o de uma popula√ß√£o finita que cont√©m um n√∫mero fixo de sucessos e fracassos.

Caracter√≠sticas:
* Popula√ß√£o Finita: Cont√©m um total de N elementos.
* N√∫mero de Sucessos na Popula√ß√£o: K (por exemplo, pessoas doentes em um hospital).
* Tamanho da Amostra: n (quantidade de indiv√≠duos extra√≠dos).
* N√∫mero de Sucessos na Amostra: x (sucessos observados na amostra).

####üî¢ Fun√ß√£o massa de Probabilidade (FMP)

A FMP da distribui√ß√£o hipergeom√©trica √© dada por:

**Exemplo:** Em um hospital com 100 pacientes, sendo 30 positivos para uma doen√ßa, selecionamos 10 pacientes para testes r√°pidos. Qual a probabilidade de encontrar exatamente 5 positivos?

```python
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
import numpy as np

# Par√¢metros
N = 100  # Tamanho da popula√ß√£o
K = 30   # N√∫mero de sucessos na popula√ß√£o (positivos)
n = 10   # Tamanho da amostra
x = 5

# C√°lculo da probabilidade
prob = hypergeom.pmf(x, N, K, n)
print(f"üåü A probabilidade √©: {prob:.4%}")
```

####üìê Propriedades da Distribui√ß√£o Hipergeom√©trica

* M√©dia (Esperan√ßa):

Interpreta√ß√£o: O valor esperado representa o n√∫mero m√©dio de sucessos ao realizar uma amostragem de tamanho

n da popula√ß√£o.

* Vari√¢ncia:

eaee2112.png

Interpreta√ß√£o: A vari√¢ncia mede o grau de dispers√£o dos sucessos ao longo da amostra, considerando a corre√ß√£o para amostragem sem reposi√ß√£o.



**Exemplo:** Em um hospital, temos um estoque de 100 kits de diagn√≥stico para uma determinada doen√ßa. Sabemos que 30 desses kits possuem defeito de fabrica√ß√£o. Vamos realizar uma inspe√ß√£o retirando 10 kits ao acaso para verificar a qualidade. Queremos calcular as propriedades da distribui√ß√£o (m√©dia, vari√¢ncia, desvio padr√£o e mediana) para a quantidade de kits defeituosos encontrados na amostra.



```python
from scipy.stats import hypergeom

# Par√¢metros
N = 100  # Tamanho da popula√ß√£o
K = 30   # N√∫mero de sucessos na popula√ß√£o
n = 10   # Tamanho da amostra

# C√°lculo das propriedades
media = hypergeom.mean(N, K, n)
variancia = hypergeom.var(N, K, n)
desvio_padrao = hypergeom.std(N, K, n)
mediana = hypergeom.median(N, K, n)

print(f"M√©dia: {media:.2f}")
print(f"Vari√¢ncia: {variancia:.2f}")
print(f"Desvio Padr√£o: {desvio_padrao:.2f}")
print(f"Mediana: {mediana:.2f}")
```

M√©dia (3.00): Em m√©dia, esperamos encontrar 3 kits defeituosos em cada inspe√ß√£o de 10 kits.

Vari√¢ncia (1.91): A dispers√£o em torno da m√©dia √© relativamente pequena, indicando que os valores n√£o variam muito.

Desvio Padr√£o (1.38): A quantidade t√≠pica de kits defeituosos em torno da m√©dia √© aproximadamente 1.38, indicando uma varia√ß√£o moderada.

Mediana (3.00): O valor central da distribui√ß√£o tamb√©m √© 3, refor√ßando que a maioria das inspe√ß√µes deve encontrar 3 kits defeituosos.



####üìä Gr√°fico da Fun√ß√£o Massa de Probabilidade (FMP)

O gr√°fico exibe a probabilidade de encontrar um n√∫mero espec√≠fico de positivos na amostra. As barras mais altas indicam os valores com maior probabilidade.

```python
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
import numpy as np

# Par√¢metros
N = 100  # Tamanho da popula√ß√£o
K = 30   # N√∫mero de sucessos na popula√ß√£o (positivos)
n = 10   # Tamanho da amostra
x = np.arange(0, n+1)

# Fun√ß√£o Massa de Probabilidade
pmf = hypergeom.pmf(x, N, K, n)

# Gr√°fico da FMP
plt.bar(x, pmf, color='purple')
plt.title('Distribui√ß√£o Hipergeom√©trica - Testes Positivos para Doen√ßa')
plt.xlabel('N√∫mero de Sucessos (Positivos)')
plt.ylabel('Probabilidade')
plt.show()
```

* A maioria dos valores de sucesso (testes positivos) se concentra nos valores intermedi√°rios (entre 2 e 5). Isso indica que a maior probabilidade est√° em obter um n√∫mero intermedi√°rio de testes positivos dentro da amostra.

* Pico da Distribui√ß√£o (Moda): O valor de moda (mais prov√°vel) ocorre pr√≥ximo de 4 sucessos positivos, indicando que esse √© o n√∫mero de testes positivos mais frequente na amostra.

* Cauda Direita: √Ä medida que o n√∫mero de sucessos aumenta (a partir de 6), a probabilidade cai rapidamente. Isso mostra que √© improv√°vel obter um n√∫mero muito elevado de positivos.

* Cauda Esquerda: A probabilidade de obter poucos sucessos (0, 1 ou 2) tamb√©m √© consideravelmente menor do que os valores pr√≥ximos da moda. Isso significa que h√° uma chance pequena de obter poucos positivos, mas n√£o √© nula.

####üìà Fun√ß√£o de distribui√ß√£o cumulativa (CDF)

A CDF calcula a probabilidade acumulada de obter at√© x sucessos:

21eaeaea.png

```python
# Fun√ß√£o de Distribui√ß√£o Acumulada (CDF)
cdf = hypergeom.cdf(x, N, K, n)

# Gr√°fico da CDF
plt.plot(x, cdf, marker='o', linestyle='-', color='blue')
plt.title('Distribui√ß√£o Acumulada - Testes Positivos para Doen√ßa')
plt.xlabel('N√∫mero de Sucessos (Positivos)')
plt.ylabel('Probabilidade Acumulada')
plt.show()
```

####üõë Fun√ß√£o sobreviv√™ncia (SF)

A SF calcula a probabilidade complementar da CDF, ou seja:

eaeaeaeaeaeaeaea.png

```python
# Fun√ß√£o de Sobreviv√™ncia (SF)
sf = hypergeom.sf(x, N, K, n)

# Gr√°fico da SF
plt.plot(x, sf, marker='o', linestyle='-', color='red')
plt.title('Fun√ß√£o de Sobreviv√™ncia - Testes Positivos para Doen√ßa')
plt.xlabel('N√∫mero de Sucessos (Positivos)')
plt.ylabel('Probabilidade de N√£o Atingir')
plt.show()
```

### üü£ Modelo de Poisson

#### üìñ Descri√ß√£o



O **Modelo de Poisson** √© usado para modelar o n√∫mero de ocorr√™ncias de um evento em um intervalo de tempo fixo, dado que as ocorr√™ncias s√£o independentes e a m√©dia de ocorr√™ncias por intervalo √© conhecida.


* Eventos Independentes: A ocorr√™ncia de um evento n√£o afeta a ocorr√™ncia de outro.
* Taxa Constante (ùúÜ): A taxa de eventos por intervalo √© fixa.
* Probabilidade Proporcional: A probabilidade depende diretamente da m√©dia e do n√∫mero de eventos desejados.

Para garantir que haja profissionais suficientes no plant√£o, o gestor deseja saber:

üìä Qual √© a probabilidade de atender exatamente 15 pacientes em uma hora?

```python
from scipy.stats import poisson

# Par√¢metros
media = 12  # M√©dia de pacientes por hora
x = 15  # N√∫mero de pacientes desejado

# 1. Probabilidade de exatamente 15 pacientes
prob_15 = poisson.pmf(x, media)
print(f"üìä A probabilidade de atender exatamente 15 pacientes em uma hora √©: {prob_15:.4%}")
```
