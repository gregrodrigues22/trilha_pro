## üòê O que √© Distribui√ß√£o Normal Padr√£o?

A **Distribui√ß√£o Normal Padr√£o** √© uma distribui√ß√£o normal com m√©dia 0 e desvio padr√£o 1.
Ela √© usada para padronizar vari√°veis e comparar diferentes conjuntos de dados.

Na √°rea da sa√∫de, podemos usar a distribui√ß√£o normal padr√£o para analisar dados de exames laboratoriais, como n√≠veis de glicose, colesterol e press√£o arterial.

```python
# C√≥digo para plotagem da distribui√ß√£o normal padr√£o
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Par√¢metros da distribui√ß√£o normal padr√£o
mean = 0
std = 1

# Gera√ß√£o de valores Z
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mean, std)

# Plotagem da distribui√ß√£o normal padr√£o
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribui√ß√£o Normal Padr√£o', color='blue')
plt.title('Distribui√ß√£o Normal Padr√£o üìä')
plt.xlabel('Z')
plt.ylabel('Densidade de Probabilidade')
plt.axvline(x=0, color='red', linestyle='--', label='M√©dia (Z=0)')
plt.legend()
plt.show()
```

## üîÑ Transforma√ß√£o de Dados para o Escore Z

A f√≥rmula para calcular o escore Z √©:

 $Z = \frac{(X - \mu)}{\sigma}$



Onde:
- \( X \): Valor observado.
- \( \mu \): M√©dia da popula√ß√£o.
- \( \sigma \): Desvio padr√£o da popula√ß√£o.

### Exemplo na √Årea da Sa√∫de üè•
Um hospital coletou dados sobre o n√≠vel de glicose em jejum de 1.000 pacientes. A m√©dia foi de 100 mg/dL, com um desvio padr√£o de 15 mg/dL. Qual √© o escore Z para um paciente com glicose de 130 mg/dL?


```python
# Dados do exemplo
X = 130  # N√≠vel de glicose do paciente
mu = 100  # M√©dia da popula√ß√£o
sigma = 15  # Desvio padr√£o

# C√°lculo do escore Z
Z = (X - mu) / sigma
print(f"Escore Z para glicose de 130 mg/dL: {Z:.2f}")
```

## üìù Uso da Tabela Z

As tabelas padronizadas foram criadas para facilitar a obten√ß√£o dos valores das √°reas sob a curva normal e eliminar a necessidade de solucionar integrais definidas.

Para consultarmos os valores em uma tabela padronizada basta transformarmos nossa vari√°vel em uma vari√°vel padronizada $Z$.

Esta vari√°vel $Z$ representa o afastamento em desvios padr√µes de um valor da vari√°vel original em rela√ß√£o √† m√©dia.

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

tabela_normal_padronizada = pd.DataFrame(
    [],
    index=["{0:0.2f}".format(i / 100) for i in range(0, 400, 10)],
    columns = ["{0:0.2f}".format(i / 100) for i in range(0, 10)])

for index in tabela_normal_padronizada.index:
    for column in tabela_normal_padronizada.columns:
        Z = np.round(float(index) + float(column), 2)
        tabela_normal_padronizada.loc[index, column] = "{0:0.4f}".format(norm.cdf(Z))

tabela_normal_padronizada.rename_axis('Z', axis = 'columns', inplace = True)

tabela_normal_padronizada
```

A Tabela Z mostra a √°rea acumulada √† esquerda de um valor Z espec√≠fico na distribui√ß√£o normal padr√£o. Isso significa que a √°rea representada na tabela √© a probabilidade acumulada de encontrar um valor menor ou igual ao valor de Z.

A tabela acima fornece a √°rea sob a curva entre  ‚àí‚àû  e  Z  desvios padr√£o acima da m√©dia. Lembrando que por se tratar de valores padronizados temos  Œº=0 .

Cada valor da tabela representa a probabilidade acumulada para um determinado escore Z. A probabilidade indica a √°rea sob a curva normal padr√£o √† esquerda do valor Z. Portanto, ao usar a tabela, voc√™ est√° sempre respondendo √† pergunta: "Qual a probabilidade de obter um valor menor ou igual a Z?"

Al√©m disso, um conceito fundamental ligado √† tabela Z √© o percentil.
O percentil indica a posi√ß√£o relativa de um valor dentro de um conjunto de dados, mostrando quanto da distribui√ß√£o est√° abaixo daquele valor.

O valor Z = 1,28 corresponde a um percentil de aproximadamente 90 (ou 89,97% para ser exato). Isso significa que 90% da popula√ß√£o tem valores menores ou iguais ao valor associado ao escore Z de 1,28. Em termos de sa√∫de, se o n√≠vel de colesterol de um paciente est√° no percentil 90, ele tem um n√≠vel de colesterol maior que 90% da popula√ß√£o.

Ou seja, a tabela Z nos permite n√£o apenas calcular probabilidades, mas tamb√©m entender como um valor se posiciona em rela√ß√£o √† popula√ß√£o inteira. Isso √© fundamental para interpreta√ß√µes estat√≠sticas e compara√ß√µes na pr√°tica cl√≠nica e epidemiol√≥gica. üí°

## üìù Conclus√£o

A principal diferen√ßa entre usar o cdf diretamente com os dados n√£o padronizados ou com os Z-scores est√° no contexto de aplica√ß√£o e na facilidade de interpreta√ß√£o. Vamos detalhar isso:



aeae212eaea.png

* Usando cdf com dados originais: A interpreta√ß√£o √© direta: a probabilidade acumulada at√© o valor X dentro da distribui√ß√£o normal espec√≠fica. Exemplo: "Qual a probabilidade de um paciente ter colesterol menor que 220 mg/dL?"

* Usando cdf com Z-score: A interpreta√ß√£o √© baseada na dist√¢ncia da m√©dia em termos de desvios padr√£o. Exemplo: "Qual a probabilidade de um valor estar a 1 desvio padr√£o da m√©dia?"

A abordagem padronizada (Z-score) permite:
* Comparar diferentes conjuntos de dados com m√©dias e desvios padr√£o distintos.
* Facilitar a an√°lise estat√≠stica e a generaliza√ß√£o dos resultados.
* A abordagem direta com cdf n√£o oferece essa flexibilidade, pois os resultados s√£o espec√≠ficos para uma distribui√ß√£o com par√¢metros definidos.

üí° Exemplo Pr√°tico: Compara√ß√£o de Altura e Colesterol

* Altura: M√©dia 1,70 m e desvio padr√£o 0,1 m. Probabilidade de ter altura maior que 1,80 m.

* Colesterol: M√©dia 200 mg/dL e desvio padr√£o 20 mg/dL. Probabilidade de ter colesterol maior que 240 mg/dL.



```python
from scipy.stats import norm

# Par√¢metros da altura
media_altura = 1.70
desvio_padrao_altura = 0.1
altura = 1.80

# Par√¢metros do colesterol
media_colesterol = 200
desvio_padrao_colesterol = 20
colesterol = 240

# C√°lculo direto com cdf
prob_altura_direto = 1 - norm.cdf(altura, media_altura, desvio_padrao_altura)
prob_colesterol_direto = 1 - norm.cdf(colesterol, media_colesterol, desvio_padrao_colesterol)

# C√°lculo usando Z-score
z_altura = (altura - media_altura) / desvio_padrao_altura
print(f"Z-score de altura > 1,80 m: {z_altura:.2f}")
z_colesterol = (colesterol - media_colesterol) / desvio_padrao_colesterol
print(f"Z-score de colesterol > 240 mg/dL: {z_colesterol:.2f}")

prob_altura_z = norm.sf(z_altura)  # Usando sf para a cauda direita
prob_colesterol_z = norm.sf(z_colesterol)  # Usando sf para a cauda direita

print(f"Probabilidade direta de altura > 1,80 m: {prob_altura_direto * 100:.2f}%")
print(f"Probabilidade direta de colesterol > 240 mg/dL: {prob_colesterol_direto * 100:.2f}%")

print(f"Probabilidade com Z-score de altura > 1,80 m: {prob_altura_z * 100:.2f}%")
print(f"Probabilidade com Z-score de colesterol > 240 mg/dL: {prob_colesterol_z * 100:.2f}%")
```

Usando os Z-scores, percebemos que: A altura de 1,80 m est√° a 1 desvio padr√£o acima da m√©dia. O colesterol de 240 mg/dL est√° a 2 desvios padr√µes acima da m√©dia. Portanto, n√≠veis de colesterol acima de 240 mg/dL s√£o muito mais raros do que alturas acima de 1,80 m.

No entanto, na pr√°tica, n√£o h√° muita diferen√ßa entre usar o cdf diretamente com os dados originais ou usar com o Z-score, especialmente quando temos ferramentas de programa√ß√£o √† disposi√ß√£o, como Python. Ambos os m√©todos chegam ao mesmo resultado, desde que os c√°lculos estejam corretos.

A grande vantagem da Tabela Z √© justamente em situa√ß√µes onde n√£o se tem acesso a um computador ou uma ferramenta de programa√ß√£o. Por exemplo:

* C√°lculos √† M√£o ou com Calculadora Simples: A tabela fornece rapidamente a probabilidade acumulada para diferentes valores de Z, sem precisar fazer c√°lculos complexos.

* Materiais Did√°ticos e Provas: Em contextos acad√™micos ou provas, como em vestibulares ou exames estat√≠sticos, a tabela Z √© uma ferramenta pr√°tica e tradicional.

* Situa√ß√µes de Campo ou Sem Computador: Profissionais de sa√∫de, estat√≠sticos ou cientistas de dados podem usar a tabela quando est√£o longe de ferramentas computacionais.

Mas √© importante lembrar que:

* A tabela Z facilita o entendimento conceitual sobre a normal padr√£o. Mesmo que n√£o usemos tanto em programa√ß√£o, ela ainda √© uma ferramenta valiosa para o aprendizado e para situa√ß√µes pr√°ticas sem computador.
