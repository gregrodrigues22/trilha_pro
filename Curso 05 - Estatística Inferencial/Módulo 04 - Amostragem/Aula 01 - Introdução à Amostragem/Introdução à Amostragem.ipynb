## üí≠ Vi√©s de amostragem

O vi√©s de amostragem ocorre quando a sele√ß√£o de uma amostra n√£o representa adequadamente a popula√ß√£o alvo, comprometendo a validade dos resultados de uma pesquisa ou estudo. Esse vi√©s pode surgir de diversas formas, afetando a precis√£o das conclus√µes obtidas.

**Causas Comuns do Vi√©s de Amostragem**

- Vi√©s de Subcobertura: Descri√ß√£o: Parte da popula√ß√£o n√£o tem chance de ser selecionada na amostra. - Vi√©s de Auto-Sele√ß√£o: Descri√ß√£o: Os participantes escolhem se querem ou n√£o participar, o que atrai perfis espec√≠ficos. Exemplo: Enquetes online sobre prefer√™ncias de produto. Problema: Quem participa pode ter uma opini√£o mais forte ou interesse no tema, enquanto pessoas indiferentes n√£o respondem. - Vi√©s de Voluntariado: A amostra √© formada por indiv√≠duos que se voluntariam para participar. Exemplo: Pesquisas sobre qualidade de atendimento em que apenas clientes insatisfeitos respondem. - Vi√©s de N√£o-Resposta: Descri√ß√£o: Parte dos indiv√≠duos selecionados para a amostra n√£o responde √† pesquisa. Exemplo: Uma pesquisa eleitoral em que apenas eleitores de um determinado grupo pol√≠tico respondem. Problema: O vi√©s ocorre porque os que n√£o respondem podem ter opini√µes significativamente diferentes. Impacto: A amostra n√£o representa a opini√£o geral da popula√ß√£o.

- Vi√©s de Conveni√™ncia: Descri√ß√£o: Seleciona os indiv√≠duos mais facilmente acess√≠veis. Exemplo: Pesquisas realizadas apenas com estudantes universit√°rios de uma √∫nica institui√ß√£o. - Vi√©s Geogr√°fico: Descri√ß√£o: A amostra se concentra em uma regi√£o geogr√°fica espec√≠fica, ignorando outras. Exemplo: Pesquisas sobre h√°bitos alimentares feitas apenas em regi√µes metropolitanas. Problema: Habitantes de √°reas rurais podem ter dietas e h√°bitos completamente diferentes. Impacto: A an√°lise n√£o reflete a realidade de √°reas n√£o amostradas.

- Vi√©s de Frequ√™ncia: Descri√ß√£o: A amostragem ocorre em intervalos que coincidem com padr√µes repetitivos. Exemplo: Coletar dados sobre tr√°fego apenas em hor√°rios de pico. - Vi√©s de Sobreviv√™ncia: Descri√ß√£o: Focar apenas em casos bem-sucedidos ou sobreviventes, ignorando os que falharam. Exemplo: Estudos sobre longevidade que analisam apenas idosos vivos. ## üóÇÔ∏è Tipos de Amostragem

O vi√©s de amostragem pode distorcer gravemente os resultados de uma an√°lise estat√≠stica, levando a decis√µes equivocadas. Por isso, √© fundamental planejar cuidadosamente a coleta e garantir a representatividade da amostra.

üßÆ **Amostragem Probabil√≠stica:**

Cada elemento da popula√ß√£o tem uma chance conhecida e n√£o nula de ser selecionado. √â considerada mais robusta e minimiza vieses, pois garante aleatoriedade.

Exemplos:

- Amostragem Estratificada: A popula√ß√£o √© dividida em grupos homog√™neos.
- Amostragem Sistem√°tica: Escolhe-se elementos em intervalos fixos.
- Amostragem por Conglomerado: Grupos heterog√™neos s√£o selecionados aleatoriamente.

üö© **Amostragem N√£o Probabil√≠stica**:

A escolha dos elementos n√£o segue um crit√©rio aleat√≥rio, o que pode introduzir vi√©s significativo. N√£o podemos garantir que a amostra represente adequadamente a popula√ß√£o. Problema: A sele√ß√£o pode ser influenciada por preconceitos ou vieses pessoais.

Exemplos:

- Amostragem por Conveni√™ncia: Escolhe-se o que est√° mais f√°cil ou acess√≠vel.

- Amostragem por Julgamento: O especialista decide quais elementos s√£o relevantes.

```python
import pandas as pd
import numpy as np

# Criando um DataFrame de exemplo com 1000 elementos
np.random.seed(42)  # Para reprodutibilidade
populacao = pd.DataFrame({
    'ID': range(1, 1001),
    'Idade': np.random.randint(18, 65, size=1000),
    'G√™nero': np.random.choice(['M', 'F'], size=1000),
    'Renda': np.random.normal(2500, 800, size=1000)
})

print("Dados de Popula√ß√£o:")
display(populacao.head())
```

## üßÆ Amostragem Probabil√≠stica

### üßÆ Amostragem Aleat√≥ria Simples

Amostragem Aleat√≥ria Simples (SRS):

Seleciona elementos aleatoriamente da popula√ß√£o, garantindo que todos tenham a mesma chance de serem escolhidos.

**Vantagens:**

* F√°cil Implementa√ß√£o:
Usando algoritmos simples (como a fun√ß√£o sample() do Python), √© poss√≠vel realizar a sele√ß√£o com poucos passos.

* Representatividade Estat√≠stica:
Por ser totalmente aleat√≥ria, caracteriza muito bem a popula√ß√£o, garantindo infer√™ncias robustas.

**Desvantagens:**

* Aus√™ncia de Categorias Raras: Se houver categorias pouco frequentes na popula√ß√£o, elas podem n√£o aparecer na amostra, prejudicando an√°lises espec√≠ficas.

* Custo e Log√≠stica: Para popula√ß√µes grandes, pode ser caro e demorado garantir que todos os elementos tenham a mesma chance.



```python
# 1. üìä Amostragem Aleat√≥ria Simples
amostra_aleatoria = populacao.sample(n=100, random_state=42)
print("\nüìä Amostragem Aleat√≥ria Simples:")
display(amostra_aleatoria.head())
```

### üßÆ Amostragem Sistem√°tica

Amostragem Sistem√°tica:

A amostragem sistem√°tica consiste em selecionar um elemento inicial de forma aleat√≥ria e, a partir dele, escolher elementos a cada intervalo fixo (en√©simo elemento) na popula√ß√£o. √ötil quando os dados est√£o ordenados.

**Vantagens:**

* F√°cil Implementa√ß√£o: Simples de realizar e pouco sujeita a erros operacionais.

* Distribui√ß√£o Uniforme: Garante que os elementos estejam distribu√≠dos ao longo da popula√ß√£o.

* Caracteriza Bem a Popula√ß√£o: Se a popula√ß√£o for homog√™nea ou ordenada, representa bem o conjunto total.

* Mitiga o Problema de Amostras Raras: Quando o intervalo √© bem definido, permite capturar grupos menos comuns.

**Desvantagens**:

* Padr√£o na Popula√ß√£o: Se houver um padr√£o c√≠clico nos dados, o m√©todo pode capturar apenas um subconjunto enviesado.

* Perda de Representatividade: Se algum elemento selecionado n√£o puder ser usado (por exemplo, o quinto elemento n√£o responder a uma pesquisa), pode prejudicar a representatividade.

* Rigidez na Escolha: Uma vez definido o ponto de partida e o intervalo, n√£o h√° flexibilidade para ajustes.

```python
# 2. üìà Amostragem Sistem√°tica
# Define o intervalo de sele√ß√£o: total da popula√ß√£o dividido pelo tamanho da amostra
intervalo = len(populacao) // 100  # Tamanho da amostra: 100 elementos
# Seleciona elementos a cada 'intervalo' posi√ß√µes
indices = np.arange(0, len(populacao), intervalo)
amostra_sistematica = populacao.iloc[indices]
print("\nüìà Amostragem Sistem√°tica:")
display(amostra_sistematica.head())  # Mostra as primeiras linhas da amostra
```

### üßÆ Amostragem Estratificada

Divide a popula√ß√£o em estratos homog√™neos e realiza a amostragem dentro de cada grupo. No exemplo, a estratifica√ß√£o foi feita por g√™nero.

**Vantagem:**

* Mitiga o Problema de Amostras Raras: Garante que grupos minorit√°rios ou menos frequentes estejam representados na amostra.

* Maior Precis√£o Estat√≠stica: Aumenta a precis√£o das estimativas ao considerar a variabilidade dentro de cada estrato.

* Redu√ß√£o de Vi√©s: Evita que algum grupo fique sub-representado ou completamente ausente.

**Desvantagem:**

* Complexidade na Estratifica√ß√£o: Muitas vezes, √© dif√≠cil definir os estratos corretamente. Exemplo: Cargo e Senioridade em uma empresa podem se sobrepor, dificultando a segmenta√ß√£o.

* Grupos M√∫ltiplos ou Cruzados: Alguns indiv√≠duos podem pertencer a mais de um estrato, o que gera ambiguidade na sele√ß√£o

```python
# 3. üóÇÔ∏è Amostragem Estratificada
from sklearn.model_selection import train_test_split

# Divide a popula√ß√£o em estratos com base no g√™nero e seleciona 10% de cada grupo
amostra_estratificada = populacao.groupby('G√™nero', group_keys=False).apply(lambda x: x.sample(frac=0.1))
print("\nüóÇÔ∏è Amostragem Estratificada:")
display(amostra_estratificada.head())  # Mostra as primeiras linhas da amostra
```

### üßÆ Amostragem por Conglomerado

## ‚úÖ Como Mitigar o Vi√©s de Amostragem

**Planejamento Cuidadoso: Definir claramente a popula√ß√£o-alvo e as estrat√©gias de coleta.**
